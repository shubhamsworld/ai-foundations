# Why Do We Need Hypothesis Testing?

Suppose you observe something interesting.

Examples:

- Coin gave 65 heads in 100 tosses.
- New website version increased sales.
- New medicine seems better.
- Model A gives higher accuracy than Model B.
- Women appear to spend more than men.
- New recommendation system increases CTR.

Question:

```text
Is this a real effect?
OR
Did it happen by chance?
```

Hypothesis Testing provides a mathematical framework to answer this.

---

# Core Idea

Whenever we observe something unusual:

```text
Observation
     ↓
Could be Random Chance
     OR
Could be Real Effect
```

Hypothesis testing measures:

> "How surprising is the observation if nothing special is happening?"

# Final Intuition

Hypothesis Testing is nothing more than:

```text
1. Assume the boring explanation is true (H₀)

2. Measure how surprising your observation is.

3. If the observation is extremely unlikely under H₀,
   reject H₀.

4. Otherwise,
   fail to reject H₀.
```

For the coin example:

```text
Observed:
65 heads in 100 tosses

If coin were fair:
Probability = 0.17%

This is extremely rare.

Therefore:
Reject H₀

Evidence suggests the coin is biased toward heads.
```

Hypothesis Testing is a mathematical framework used to decide whether an observed result happened:

> "Given the data we observed, is there enough evidence against our default assumption?"

# Complete Mental Model

Hypothesis Testing is simply:

```text
1. Assume the boring explanation is true.

2. Measure how surprising the observation is.

3. Compute:
   P(Observed or More Extreme | H₀)

4. This probability is called the P-value.

5. Compare it with α.

6. If p-value is very small:
      Reject H₀

7. Otherwise:
      Fail to Reject H₀
```

# Why Do We Need Hypothesis Testing?

In real life we usually do not have access to the entire population.

Examples:

```text
All customers of Amazon
All voters in India
All website visitors
All coin tosses that will ever happen
```

Instead we collect a:

```text
Sample
```

and use that sample to make conclusions about the population.

The problem is:

> Was the observed result produced by random chance, or is there actually something unusual happening?

Hypothesis Testing is the mathematical framework that answers this question.

A random variable follows Binomial Distribution if:

p^k
(1-p)^{n-k}

This gives probability of exactly:

# One Sentence Summary

Hypothesis testing measures how surprising the observed data would be if the null hypothesis were true, and uses either a p-value or critical value to decide whether that surprise is large enough to reject H₀.

# Hypothesis Testing — Complete Framework (Intuition → Errors → Power → p-value → α → Decision Making)

---

# Big Picture

Hypothesis Testing is a formal way of answering:

> "Is the observed evidence strong enough to doubt our current belief?"

---

Every hypothesis test starts with two competing claims:

## Null Hypothesis

$$
H_0
$$

Current belief, status quo, default assumption.

---

## Alternative Hypothesis

$$
H_a
$$

The claim we are trying to gather evidence for.

---

# Running Example Throughout

Suppose we have a coin.

Current belief:

```text
Coin is fair.
```

Therefore

$$
H_0 : p = 0.5
$$

where

$$
p
=
P(\text{Head})
$$

---

Possible alternatives:

### Right-Tailed

$$
H_a:p>0.5
$$

Coin favors heads.

---

### Left-Tailed

$$
H_a:p<0.5
$$

Coin favors tails.

---

### Two-Tailed

$$
H_a:p\neq0.5
$$

Coin is biased.

Could favor heads OR tails.

---

# Why Do We Assume H₀ First?

Imagine a court case.

Current assumption:

```text
Person is innocent.
```

Only strong evidence can change that belief.

---

Hypothesis testing works exactly the same way.

We start by assuming:

$$
H_0
$$

is true.

Then we ask:

> If H₀ were true, how surprising would our observed data be?

---

# Framework of Hypothesis Testing

```text
Define H₀
      ↓
Define Hₐ
      ↓
Choose α
      ↓
Collect Sample
      ↓
Calculate Test Statistic
      ↓
Find Sampling Distribution under H₀
      ↓
Compute p-value
      ↓
Compare with α
      ↓
Reject H₀
or
Fail To Reject H₀
```

---

# Test Statistic

A test statistic is a number extracted from data.

It summarizes evidence.

---

Examples:

### Coin Toss

$$
T
=
\text{Number of Heads}
$$

---

### Mean Test

$$
T
=
\bar X
$$

Sample Mean.

---

### Z-Test

$$
Z
=
\frac{\bar X-\mu_0}
{\sigma/\sqrt n}
$$

---

### t-Test

$$
t
=
\frac{\bar X-\mu_0}
{s/\sqrt n}
$$

---

The test statistic is what we compare against the null hypothesis.

---

# Sampling Distribution Under H₀

This is one of the most important concepts.

---

Assume:

$$
H_0
$$

is true.

Now imagine repeating the experiment thousands of times.

The distribution of the resulting test statistic is called:

```text
Sampling Distribution Under H₀
```

---

Example:

Coin tossed 100 times.

Under

$$
H_0:p=0.5
$$

Expected heads:

$$
50
$$

Most outcomes occur around:

```text
50 heads
```

Few occur at:

```text
20 heads
80 heads
```

Extremely few occur at:

```text
5 heads
95 heads
```

---

This distribution lets us measure:

```text
How unusual our observation is.
```

---

# Significance Level (α)

Before collecting data we choose:

$$
\alpha
$$

---

Definition:

> Maximum probability of incorrectly rejecting H₀.

---

Common choices:

| α | Meaning |
|----|----------|
| 0.10 | 10% |
| 0.05 | 5% |
| 0.01 | 1% |
| 0.001 | 0.1% |

---

Most common:

$$
\alpha = 0.05
$$

---

Interpretation

If

$$
\alpha = 5\%
$$

then:

```text
I am willing to incorrectly reject H₀
at most 5% of the time.
```

---

# Critical Region

Critical Region is also called:

```text
Rejection Region
```

---

Definition:

> Region containing values that are so extreme that we reject H₀.

---

Visual

```text
Normal values      Extreme values

---------|██████████
         Tcr
```

---

If test statistic enters this region:

```text
Reject H₀
```

---

# Acceptance Region

Region outside critical region.

---

Better wording:

```text
Fail To Reject H₀
```

instead of

```text
Accept H₀
```

---

Why?

Because we never prove H₀.

We only say:

```text
Not enough evidence against it.
```

---

# Critical Value

Critical value is the boundary.

---

Visual

```text
Accept H₀ | Reject H₀
----------|------------
          ^
         Tcr
```

---

Depends on:

1. α
2. Distribution
3. Test type

---

# One-Tailed Tests

Used when direction matters.

---

# Right-Tailed Test

Question:

```text
Is value significantly larger?
```

---

Hypotheses

$$
H_0:p=0.5
$$

$$
H_a:p>0.5
$$

---

Visual

```text
                 /\
               /    \
             /        \
-----------50----Tcr------
                    ██████
                    ██████
```

---

Reject H₀ if:

$$
T>T_{cr}
$$

---

Example

Medicine increases recovery rate.

$$
H_a:\mu>\mu_0
$$

---

Example

New algorithm increases accuracy.

$$
H_a:\mu_{new}>\mu_{old}
$$

---

# Left-Tailed Test

Question:

```text
Is value significantly smaller?
```

---

Hypotheses

$$
H_0:p=0.5
$$

$$
H_a:p<0.5
$$

---

Visual

```text
      /\
    /    \
  /        \
Tcr----50----------
██████
██████
```

---

Reject H₀ if:

$$
T<T_{cr}
$$

---

Example

Drug lowers blood pressure.

$$
H_a:\mu<\mu_0
$$

---

Example

New algorithm reduces latency.

$$
H_a:\mu_{new}<\mu_{old}
$$

---

# Two-Tailed Test

Question:

```text
Is value significantly different?
```

---

Hypotheses

$$
H_0:p=0.5
$$

$$
H_a:p\neq0.5
$$

---

Visual

```text
      ████        ████
-----T1---50---T2-----
```

---

Reject H₀ if:

$$
T<T_1
$$

or

$$
T>T_2
$$

---

Example

Checking whether a machine calibration changed.

Not interested in direction.

Only interested in:

```text
Any change.
```

---

# Why α Splits In Two-Tailed Test

Suppose

$$
\alpha=0.05
$$

---

Then

$$
\frac{0.05}{2}
=
0.025
$$

goes to left tail.

---

and

$$
0.025
$$

goes to right tail.

---

Visual

```text
2.5%      95%      2.5%

████------████------████
```

---

# P-value

Definition:

> Probability of observing data at least as extreme as what we observed assuming H₀ is true.

---

Most Important Formula

$$
p\text{-value}
=
P(\text{Data at least this extreme}\mid H_0)
$$

---

NOT

$$
P(H_0|Data)
$$

This is a common mistake.

---

# Coin Example

Suppose:

100 tosses

Observed:

$$
T_{obs}=65
$$

---

Under

$$
H_0:p=0.5
$$

Compute:

$$
P(T\ge65|H_0)
$$

---

Suppose result is:

$$
0.002
$$

Meaning:

```text
Only 0.2% chance of seeing
65 or more heads if coin is fair.
```

---

Very surprising.

Therefore:

```text
Reject H₀.
```

---

# Decision Rule Using P-value

Reject H₀ if:

$$
p\text{-value}<\alpha
$$

---

Fail To Reject H₀ if:

$$
p\text{-value}\ge\alpha
$$

---

Example

$$
\alpha=0.05
$$

---

Case 1

$$
p=0.002
$$

Since

$$
0.002<0.05
$$

Reject H₀.

---

Case 2

$$
p=0.12
$$

Since

$$
0.12>0.05
$$

Fail To Reject H₀.

---

# Type-I Error

Actual Reality:

$$
H_0
\text{ is true}
$$

But we reject it.

---

Visual

| Actual | Decision |
|----------|----------|
| H₀ True | Reject H₀ |

---

This is called:

```text
False Positive
```

---

Probability:

$$
\alpha
$$

---

Example

Coin is actually fair.

But experiment concludes:

```text
Coin is biased.
```

---

# Type-II Error

Actual Reality:

$$
H_a
\text{ is true}
$$

But we fail to reject H₀.

---

Visual

| Actual | Decision |
|----------|----------|
| Hₐ True | Fail To Reject H₀ |

---

This is called:

```text
False Negative
```

---

Probability:

$$
\beta
$$

---

Example

Coin is actually biased.

But test says:

```text
Coin appears fair.
```

---

# Complete Error Matrix

| Actual Reality | Decision: Fail To Reject H₀ | Decision: Reject H₀ |
|----------------|----------------------------|--------------------|
| H₀ True | Correct | Type-I Error |
| Hₐ True | Type-II Error | Correct |

---

# False Positive

Same as:

```text
Type-I Error
```

---

Example

Healthy patient diagnosed sick.

---

# False Negative

Same as:

```text
Type-II Error
```

---

Example

Sick patient diagnosed healthy.

---

# Relationship Between α and Type-I Error

Very Important

$$
\alpha
=
P(\text{Reject }H_0 \mid H_0)
$$

---

Read as:

> Probability of rejecting H₀ when H₀ is actually true.

---

# Confidence Level

Confidence Level and α are complements.

$$
Confidence
=
1-\alpha
$$

---

Example

$$
\alpha=0.05
$$

Then

$$
Confidence=95\%
$$

---

Example

$$
\alpha=0.01
$$

Then

$$
Confidence=99\%
$$

---

# Power of a Test

One of the most important concepts.

Definition:

> Probability that the test correctly detects a real effect.

---

Formula

$$
Power
=
1-\beta
$$

---

Visual

| Quantity | Meaning |
|-----------|----------|
| α | Type-I Error |
| β | Type-II Error |
| 1-β | Power |

---

Example

Power = 90%

Means:

```text
If a real effect exists,
the test will detect it
90% of the time.
```

---

# Why Power Matters

Imagine:

New medicine actually works.

Low power means:

```text
We may fail to detect it.
```

---

High power means:

```text
We are likely to discover it.
```

---

# Factors Affecting Power

Increase Power:

### Larger Sample Size

$$
n \uparrow
\Rightarrow
Power \uparrow
$$

---

### Larger Effect Size

Bigger difference is easier to detect.

---

### Lower Variance

Less noise.

---

### Larger α

More willingness to reject H₀.

---

# Trade-Off Between α and β

Making α very small:

```text
Harder to reject H₀
```

---

Therefore:

```text
β increases
```

---

Therefore:

```text
Power decreases
```

---

This is why hypothesis testing is a balancing act.

---

# Permutation Test

What if the sampling distribution under H₀ is unknown?

---

Use:

```text
Permutation Test
```

---

Idea:

Randomly shuffle labels many times.

---

Create empirical distribution.

---

Compare observed statistic against this distribution.

---

Advantages:

```text
No strong distribution assumptions.
```

---

Related Concept:

```text
Bootstrap
```

which resamples observations with replacement.

---

# Example Connecting Everything

Suppose:

100 tosses

Observed:

$$
65
\text{ heads}
$$

---

Hypotheses

$$
H_0:p=0.5
$$

$$
H_a:p>0.5
$$

---

Choose

$$
\alpha=0.05
$$

---

Compute

$$
p\text{-value}=0.002
$$

---

Decision

$$
0.002<0.05
$$

Reject H₀.

---

Interpretation

```text
There is strong evidence
that the coin favors heads.
```

---

NOT

```text
Coin is definitely biased.
```

---

NOT

```text
Probability H₀ is true is 0.2%.
```

---

Correct Interpretation

```text
If the coin were fair,
seeing data this extreme
would happen only about
0.2% of the time.
```

# Final Mental Map

```text
H₀ vs Hₐ
      ↓
Collect Data
      ↓
Compute Test Statistic
      ↓
Sampling Distribution Under H₀
      ↓
Compute P-value
      ↓
Compare With α
      ↓
Reject H₀ or Fail To Reject H₀
      ↓
Possible Errors
      ↓
Type-I Error = α
Type-II Error = β
Power = 1-β
Confidence = 1-α
```

---

# One-Line Summary

Hypothesis testing starts by assuming H₀ is true, measures how surprising the observed data would be under that assumption using a test statistic and p-value, then uses α to decide whether the evidence is strong enough to reject H₀ while controlling Type-I error, balancing Type-II error, and maximizing statistical power.
