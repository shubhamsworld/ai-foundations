# Why Do We Need Hypothesis Testing?

Suppose you observe something interesting.

Examples:

- Coin gave 65 heads in 100 tosses.
- New website version increased sales.
- New medicine seems better.
- Model A gives higher accuracy than Model B.
- Women appear to spend more than men.
- New recommendation system increases CTR.

Question:

```text
Is this a real effect?
OR
Did it happen by chance?
```

Hypothesis Testing provides a mathematical framework to answer this.

---

# Core Idea

Whenever we observe something unusual:

```text
Observation
     ↓
Could be Random Chance
     OR
Could be Real Effect
```

Hypothesis testing measures:

> "How surprising is the observation if nothing special is happening?"

# Final Intuition

Hypothesis Testing is nothing more than:

```text
1. Assume the boring explanation is true (H₀)

2. Measure how surprising your observation is.

3. If the observation is extremely unlikely under H₀,
   reject H₀.

4. Otherwise,
   fail to reject H₀.
```

For the coin example:

```text
Observed:
65 heads in 100 tosses

If coin were fair:
Probability = 0.17%

This is extremely rare.

Therefore:
Reject H₀

Evidence suggests the coin is biased toward heads.
```

Hypothesis Testing is a mathematical framework used to decide whether an observed result happened:

> "Given the data we observed, is there enough evidence against our default assumption?"

# Complete Mental Model

Hypothesis Testing is simply:

```text
1. Assume the boring explanation is true.

2. Measure how surprising the observation is.

3. Compute:
   P(Observed or More Extreme | H₀)

4. This probability is called the P-value.

5. Compare it with α.

6. If p-value is very small:
      Reject H₀

7. Otherwise:
      Fail to Reject H₀
```
