# Introduction to Probability and Statistics

## Why Do We Need Probability in Machine Learning?

Some datasets contain classes that are clearly separated.

In such cases, simple rules may be sufficient and complex Machine Learning models may not be necessary.

However, in many real-world datasets, classes overlap with each other.

Consider a query point:

$$X_q$$

that lies in a mixed region.

Instead of saying:

> This point definitely belongs to Class A.

we often assign probabilities:

$$P(\text{Versicolor}\mid X_q)=0.8$$

$$P(\text{Virginica}\mid X_q)=0.2$$

$$P(\text{Setosa}\mid X_q)=0$$

This means:

* 80% chance of being Versicolor
* 20% chance of being Virginica
* 0% chance of being Setosa

Probability allows Machine Learning models to express uncertainty rather than making absolute decisions.

---

## What Falls Under Probability and Statistics?

Many important concepts used in Machine Learning come from Probability and Statistics:

* Mean
* Median
* Variance
* Standard Deviation
* Histograms
* Percentiles
* Quartiles
* Probability Distributions
* PDF (Probability Density Function)
* CDF (Cumulative Distribution Function)

These tools help us understand data before building Machine Learning models.

# Random Variables

## Definition

A **Random Variable (RV)** is a variable whose value depends on the outcome of a random experiment.

A random variable can take different values depending on what outcome occurs.

---

## Types of Random Variables

There are two major types:

1. Discrete Random Variables
2. Continuous Random Variables

---

## Discrete Random Variable

A discrete random variable can take values from a finite or countable set.

### Example: Dice Roll

Let

$$X$$

represent the outcome of rolling a fair six-sided die.

Possible values:

$$X={1,2,3,4,5,6}$$

Since the variable can only take one value from this finite set, it is a **Discrete Random Variable**.

### Other Examples

* Number of students in a classroom
* Number of cars in a parking lot
* Number of heads obtained in coin tosses

These are countable quantities.

---

## Continuous Random Variable

A continuous random variable can take any real value within a range.

### Example: Height of a Student

Let

$$Y$$

represent the height of a randomly selected student.

Possible values:

$$162.45 \text{ cm}$$

$$171.23 \text{ cm}$$

$$183.789 \text{ cm}$$

Since infinitely many values are possible,

$$Y$$

is a **Continuous Random Variable**.

### Other Examples

* Height
* Weight
* Temperature
* Age
* Salary

These quantities are measured rather than counted.

---

# Probability Examples

## Example: Fair Dice

Consider a fair six-sided die.

Possible outcomes:

$$X={1,2,3,4,5,6}$$

Since all outcomes are equally likely:

$$P(X=1)=\frac{1}{6}$$

$$P(X=2)=\frac{1}{6}$$

$$P(X=3)=\frac{1}{6}$$

and so on.

---

## Probability of Getting an Even Number

Even numbers are:

$${2,4,6}$$

Therefore:

$$P(X\text{ is even})=P(X=2)+P(X=4)+P(X=6)$$

$$=\frac{1}{6}+\frac{1}{6}+\frac{1}{6}$$

$$=\frac{3}{6}$$

$$=\frac{1}{2}$$

Thus,

$$P(X\text{ is even})=0.5$$

---

## Probability of Getting an Odd Number

Odd numbers are:

$${1,3,5}$$

Therefore:

$$P(X\text{ is odd})=P(X=1)+P(X=3)+P(X=5)$$

$$=\frac{1}{6}+\frac{1}{6}+\frac{1}{6}$$

$$=\frac{3}{6}$$

$$=\frac{1}{2}$$

Thus,

$$P(X\text{ is odd})=0.5$$

# Population and Sample

## Population

A population is the complete set of observations we are interested in studying.

Examples:

* All humans in the world
* All customers of a company
* All flowers of a species

---

## Sample

A sample is a subset of the population.

Instead of studying the entire population, we collect a smaller sample and analyze it.

---

## Why Do We Need Sampling?

Suppose we want to estimate the average height of all humans on Earth.

Let

$$N \approx 7 \text{ billion}$$

It is practically impossible to measure every person's height.

Instead, we randomly select a sample.

Example:

```text
Sample Size = 1000
```

and estimate population characteristics from that sample.

---

## Population Mean

The mean of the entire population is denoted by:

$$\mu$$

If the population contains

$$N$$

observations:

$$\mu=\frac{1}{N}\sum_{i=1}^{N}x_i$$

where:

* (N) = population size
* (x_i) = ith observation

---

## Sample Mean

The mean of a sample is denoted by:

$$\bar{x}$$

If the sample contains

$$n$$

observations:

$$\bar{x}=\frac{1}{n}\sum_{i=1}^{n}x_i$$

where:

* (n) = sample size
* (x_i) = ith sample observation

---

## Important Idea

As the sample size increases:

$$\bar{x}\rightarrow\mu$$

This means:

> The sample mean becomes a better estimate of the population mean.

This idea is fundamental in Statistics and Machine Learning.

---

## Random Sampling

A sample collected randomly from the population is called a **Random Sample**.

Random sampling helps ensure that the sample represents the population fairly and reduces bias.

---

# Population vs Sample

# Gaussian (Normal) Distribution

## Why Do We Study Gaussian Distribution?

Many real-world quantities tend to follow a bell-shaped pattern:

- Human heights
- Human weights
- Exam scores
- Measurement errors
- Sensor noise

Because of this, Gaussian Distribution is one of the most important distributions in Statistics and Machine Learning.

It is also called the **Normal Distribution**.

---

## Bell-Shaped Curve

A Gaussian distribution looks like a symmetric bell-shaped curve.

Characteristics:

- Highest point occurs at the center.
- Values near the center occur most frequently.
- Values far from the center occur less frequently.
- Left and right sides are mirror images.

```text
                 ^
                 |
               ***
             **   **
            *       *
          **         **
---------*-----μ-------*-------->
```

---

# Parameters of a Gaussian Distribution

A Gaussian distribution is completely described by two parameters:

1. Mean ($\mu$)
2. Variance ($\sigma^2$)

Notation:

$$
X \sim N(\mu,\sigma^2)
$$

Read as:

> Random variable X follows a Normal Distribution with mean μ and variance σ².

---

## Mean ($\mu$)

The mean determines the center of the distribution.

Example:

$$
\mu = 50
$$

means the curve is centered at 50.

```text
            μ=50
              |
              v
            *****
          **     **
---------*---------*--------->
```

If μ changes:

- Curve shifts left or right.
- Shape remains the same.

---

## Variance ($\sigma^2$)

Variance controls how spread out the data is.

### Small Variance

Data points stay close to the mean.

```text
                *
              *   *
             *     *
------------*---μ---*------------>
```

### Large Variance

Data points spread farther away from the mean.

```text
          *             *
        *                 *
------*--------μ-----------*------->
```

Therefore:

- Small variance → narrow curve
- Large variance → wider curve

---

## Standard Deviation

Standard deviation is simply the square root of variance.

$$
\sigma = \sqrt{\sigma^2}
$$

Interpretation:

- Small σ → less spread
- Large σ → more spread

---

# Probability Density Function (PDF)

The PDF tells us:

> How likely values around x are.

For a Gaussian distribution:

$$
f(x)=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(
-\frac{(x-\mu)^2}{2\sigma^2}
\right)
$$

where:

- μ = mean
- σ = standard deviation
- exp() = exponential function

---

## Understanding the Formula Intuitively

Ignore constants for a moment:

$$
f(x) \propto
\exp(-x^2)
$$

Focus on:

$$
y = e^{-x^2}
$$

Let's evaluate a few values:

| x | y = e^(-x²) |
|---|---|
| 0 | 1 |
| 1 | 0.3679 |
| 2 | 0.0183 |
| 3 | 0.00012 |

Notice:

As x moves away from 0,

$$
x^2
$$

becomes larger,

therefore

$$
e^{-x^2}
$$

becomes very small.

This is why the Gaussian curve drops rapidly on both sides.

---

## Key Properties of PDF

### 1. Maximum at Mean

The highest probability density occurs at:

$$
x=\mu
$$

The center of the curve is the peak.

---

### 2. Symmetric

For a Gaussian distribution:

$$
f(\mu-a)=f(\mu+a)
$$

Both sides are mirror images.

---

### 3. Density Decreases Away from Mean

As distance from mean increases:

$$
|x-\mu|
$$

increases,

therefore PDF decreases.

---

### 4. Total Area Equals 1

The area under the entire PDF curve is:

$$
1
$$

This represents 100% probability.

```text
           PDF
            ^
            |
         ********
       ***      ***
     ***          ***
----*--------------*-------->
       Entire Area = 1
```

---

# Important Note

For continuous random variables:

$$
P(X=x)=0
$$

for any exact value x.

Instead we calculate probabilities over intervals:

$$
P(a \le X \le b)
$$

which corresponds to the area under the curve between a and b.

---

# Cumulative Distribution Function (CDF)

The CDF answers a different question.

Instead of asking:

> How dense is probability around x?

it asks:

> What is the probability of observing a value less than or equal to x?

Definition:

$$
F(x)=P(X\le x)
$$

---

## Relationship Between PDF and CDF

CDF is obtained by accumulating area under the PDF.

$$
F(x)
=
\int_{-\infty}^{x}
f(t)\,dt
$$

This means:

CDF = area under PDF from negative infinity up to x.

---

## Shape of a Gaussian CDF

The Gaussian PDF is bell-shaped.

The Gaussian CDF is S-shaped.

```text
CDF
 ^
1|                         ****
 |                     ****
 |                 ****
 |             ****
0.5----------***
 |        ***
 |     ***
 |  ***
0+---------------------------->
                 μ
```

---

## Important Property

At the mean:

$$
x=\mu
$$

the CDF value is:

$$
F(\mu)=0.5
$$

Why?

Because the Gaussian distribution is symmetric.

Half of the area lies to the left of μ.

Half lies to the right.

Therefore:

$$
P(X\le \mu)=0.5
$$

---

## Standard Normal Distribution

When:

$$
\mu=0
$$

and

$$
\sigma=1
$$

the distribution is called the Standard Normal Distribution.

Notation:

$$
Z \sim N(0,1)
$$

---

### Useful Values

$$
CDF(0)=0.5
$$

$$
CDF(1)\approx0.8413
$$

$$
CDF(-1)\approx0.1587
$$

---

# Effect of Variance on PDF

### Smaller Variance

```text
                *
              *   *
             *     *
------------*---μ---*------------>
```

- Narrow
- Taller
- Less spread

---

### Larger Variance

```text
          *             *
        *                 *
------*--------μ-----------*------->
```

- Wider
- Shorter
- More spread

---

# Effect of Variance on CDF

When variance is small:

- CDF rises very sharply near μ.

When variance is large:

- CDF rises slowly.

```text
Small Variance:

0 -----|████|----- 1

Large Variance:

0 --████████████-- 1
```

A Gaussian Distribution:

$$

$$

is controlled by:

# 68–95–99.7 Rule (Empirical Rule)

## Why Do We Need This Rule?

For a Gaussian (Normal) Distribution, we often want to know:

> How much of the data lies within 1, 2, or 3 standard deviations from the mean?

Instead of computing complicated integrals every time, we use the **68–95–99.7 Rule**.

This rule provides quick estimates of how data is distributed around the mean.

---

## Empirical Rule

For a Normal Distribution:

$$
X \sim N(\mu,\sigma^2)
$$

the following approximations hold:

| Interval | Range | Percentage of Data |
|-----------|-----------|-----------|
| 1σ | $\mu \pm \sigma$ | ≈ 68% |
| 2σ | $\mu \pm 2\sigma$ | ≈ 95% |
| 3σ | $\mu \pm 3\sigma$ | ≈ 99.7% |

---

## Visual Interpretation

```text
                Mean (μ)
                    |
                    ▼

         |----|----|----|----|----|----|
       -3σ  -2σ  -1σ   μ   +1σ +2σ +3σ

      2.1% 13.6% 34.1% 34.1% 13.6% 2.1%

```

Almost all observations are concentrated near the mean.

Very few observations occur far away from the mean.

---

# Understanding the Percentages

## Within One Standard Deviation

Range:

$$
\mu-\sigma
\quad \text{to} \quad
\mu+\sigma
$$

Contains approximately:

$$
68\%
$$

of all observations.

Since the distribution is symmetric:

- 34.1% lies between μ and +σ
- 34.1% lies between μ and −σ

Therefore:

$$
34.1\% + 34.1\% \approx 68\%
$$

---

## Within Two Standard Deviations

Range:

$$
\mu-2\sigma
\quad \text{to} \quad
\mu+2\sigma
$$

Contains approximately:

$$
95\%
$$

of all observations.

This means only about:

$$
5\%
$$

of observations lie outside this range.

---

## Within Three Standard Deviations

Range:

$$
\mu-3\sigma
\quad \text{to} \quad
\mu+3\sigma
$$

Contains approximately:

$$
99.7\%
$$

of all observations.

Only:

$$
0.3\%
$$

of observations lie outside this range.

---

# Area in Each Region

The Gaussian curve can be divided into regions:

| Region | Area |
|----------|----------|
| μ to +1σ | 34.1% |
| -1σ to μ | 34.1% |
| +1σ to +2σ | 13.6% |
| -2σ to -1σ | 13.6% |
| +2σ to +3σ | 2.1% |
| -3σ to -2σ | 2.1% |
| Beyond +3σ | 0.15% |
| Beyond -3σ | 0.15% |

---

## Complete Breakdown

```text
      0.15%   2.1%   13.6%   34.1%   34.1%   13.6%   2.1%   0.15%

---------|------|------|------|------|------|------|---------
       -3σ    -2σ    -1σ     μ     +1σ    +2σ    +3σ
```

---

# Example

Suppose:

$$
X \sim N(0,4)
$$

Then:

$$
\mu = 0
$$

and

$$
\sigma^2 = 4
$$

Therefore:

$$
\sigma = \sqrt{4}=2
$$

---

## One Standard Deviation Range

$$
\mu \pm \sigma
$$

$$
0 \pm 2
$$

Range:

$$
[-2,2]
$$

Approximately:

$$
68\%
$$

of observations lie between −2 and 2.

---

## Two Standard Deviation Range

$$
0 \pm 4
$$

Range:

$$
[-4,4]
$$

Approximately:

$$
95\%
$$

of observations lie between −4 and 4.

---

## Three Standard Deviation Range

$$
0 \pm 6
$$

Range:

$$
[-6,6]
$$

Approximately:

$$
99.7\%
$$

of observations lie between −6 and 6.

---

# Relation with PDF

The percentages above represent **areas under the PDF curve**.

For example:

$$
P(\mu-\sigma \le X \le \mu+\sigma)
\approx 0.68
$$

This means:

> The area under the PDF between μ−σ and μ+σ is approximately 68%.

Similarly:

$$
P(\mu-2\sigma \le X \le \mu+2\sigma)
\approx 0.95
$$

and

$$
P(\mu-3\sigma \le X \le \mu+3\sigma)
\approx 0.997
$$

---

# Relation with CDF

The Empirical Rule can also be understood using the CDF.

Since:

$$
CDF(\mu)=0.5
$$

exactly half of the area lies on each side of the mean.

For the Standard Normal Distribution:

$$
CDF(1)\approx0.8413
$$

and

$$
CDF(-1)\approx0.1587
$$

Therefore:

$$
P(-1 \le Z \le 1)
=
CDF(1)-CDF(-1)
$$

$$
=
0.8413-0.1587
$$

$$
\approx0.6826
$$

which is approximately:

$$
68\%
$$

---

# Why Is This Important in Machine Learning?

The Empirical Rule helps us:

- Detect outliers
- Understand data spread
- Standardize features
- Interpret z-scores
- Understand confidence intervals
- Analyze Gaussian assumptions in models

Many ML algorithms assume data is approximately normally distributed, so this rule is used frequently.

For a Gaussian Distribution:

$$
X \sim N(\mu,\sigma^2)
$$

- About 68% of data lies within 1 standard deviation of the mean.

- About 99.7% of data lies within 3 standard deviations of the mean.
- The percentages represent areas under the PDF curve.
- Values beyond 3σ are very rare and are often treated as potential outliers.

# 1. Symmetric Distribution

A distribution is called **symmetric** if its left and right sides are mirror images around the center.

The center is usually the mean ($\mu$).

---

## Visual Idea

```text
                ^
                |
              *****
            **     **
          **         **
--------**-----μ------**--------
          **         **
            **     **
              *****
```

Left side and right side look identical.

---

## Properties of a Symmetric Distribution

### Mean = Median = Mode

For a perfectly symmetric distribution:

$$
Mean = Median = Mode
$$

All measures of central tendency coincide.

---

### Equal Probability on Both Sides

Exactly half the observations lie on each side of the mean.

$$
P(X < \mu)=0.5
$$

$$
P(X > \mu)=0.5
$$

---

### CDF Property

At the mean:

$$
CDF(\mu)=0.5
$$

because exactly 50% of the area lies to the left.

---

## Example: Normal Distribution

The Gaussian distribution is symmetric.

$$
X \sim N(\mu,\sigma^2)
$$

Examples:

- Human height
- IQ scores
- Measurement errors
- Sensor noise

often approximately follow a symmetric distribution.

---

# 2. Skewness

## Motivation

Not all datasets are symmetric.

Consider income data.

Most people earn moderate salaries.

A few people earn extremely high salaries.

This creates a long tail on the right side.

Such distributions are called **skewed distributions**.

---

## Definition

Skewness measures:

> How asymmetric a distribution is around its mean.

It tells us:

1. Whether the distribution is symmetric
2. Which side has the longer tail
3. How far the distribution is from symmetry

---

## Skewness Formula

Population skewness:

$$
Skewness=
\frac{E[(X-\mu)^3]}
{\sigma^3}
$$

You don't need to memorize this formula now.

The intuition is much more important.

---

# Types of Skewness

---

## A. Symmetric Distribution

```text
                ^
                |
              *****
            **     **
          **         **
--------**-----μ------**--------
          **         **
            **     **
              *****
```

Skewness:

$$
0
$$

---

### Properties

- Mean = Median = Mode
- Left tail = Right tail
- Perfect balance

Example:

- Ideal Gaussian distribution

---

## B. Right Skewed Distribution (Positive Skew)

```text
      ^
      |
     ****
   ***   **
 ***       ***
*             *********
---------------------------->
```

Tail is longer on the right.

---

### Skewness

$$
Skewness > 0
$$

---

### Why?

A few very large observations pull the distribution to the right.

---

### Relationship

```text
Mode < Median < Mean
```

or

$$
Mean > Median > Mode
$$

---

### Real-Life Examples

#### Income

Most people earn moderate salaries.

A few people earn extremely high salaries.

```text
Many people ---> $50k

Few people ---> $5M
```

Long right tail.

---

#### House Prices

Most houses are affordable.

A few luxury properties are extremely expensive.

---

#### Social Media Followers

Most users have few followers.

Celebrities have millions.

---

# C. Left Skewed Distribution (Negative Skew)

```text
                   ^
                   |
               *****
             **    ***
          ***        **
**********            *
------------------------>
```

Tail is longer on the left.

---

### Skewness

$$
Skewness < 0
$$

---

### Relationship

```text
Mean < Median < Mode
```

---

### Examples

#### Retirement Age

Most people retire around 60.

Very few retire much earlier.

This creates a left tail.

---

#### Easy Examination Scores

Most students score very high.

Few students score extremely low.

---

# Summary of Skewness

| Distribution | Tail Direction | Skewness |
|-------------|--------------|----------|
| Symmetric | No tail dominance | 0 |
| Right Skewed | Right Tail Longer | > 0 |
| Left Skewed | Left Tail Longer | < 0 |

# 3. Kurtosis

## Motivation

Suppose two distributions have:

- Same mean
- Same variance

Can their shapes still be different?

Yes.

One may have:

- More extreme values
- More outliers
- Heavier tails

This behavior is measured using **Kurtosis**.

---

# Definition

Kurtosis measures:

> How heavy the tails are and how likely extreme observations are.

It is often described as:

- Tail heaviness
- Outlier-proneness
- Peak sharpness

Though modern statistics focuses more on tails than peak height.

---

# Kurtosis Formula

$$
Kurtosis=
\frac{E[(X-\mu)^4]}
{\sigma^4}
$$

Again, intuition matters more than the formula.

---

# Normal Distribution Kurtosis

A Gaussian distribution has:

$$
Kurtosis = 3
$$

This serves as the reference distribution.

---

# Excess Kurtosis

Most software libraries use:

$$
Excess\ Kurtosis
=
Kurtosis - 3
$$

Therefore:

For Gaussian:

$$
Excess\ Kurtosis=0
$$

---

# Standardization

Standardization transforms data into a common scale.

Formula:

$$
z=
\frac{x-\mu}{\sigma}
$$

---

## Interpretation

### Step 1

Subtract mean:

$$
x-\mu
$$

Measures distance from center.

---

### Step 2

Divide by standard deviation:

$$
\sigma
$$

Measures distance in units of standard deviations.

---

# Example

$$
x=80
$$

$$
\mu=70
$$

$$
\sigma=5
$$

Then:

$$
z=
\frac{80-70}{5}
=
2
$$

Interpretation:

> This observation is 2 standard deviations above the mean.

| Z-score | Meaning |
|----------|----------|
| 0 | Exactly at mean |
| +1 | One SD above mean |
| +2 | Two SD above mean |
| -1 | One SD below mean |
| -2 | Two SD below mean |

# Relationship with 68–95–99.7 Rule

For:

$$
Z \sim N(0,1)
$$

Approximately:

$$
P(-1<Z<1)\approx0.68
$$

$$
P(-2<Z<2)\approx0.95
$$

$$
P(-3<Z<3)\approx0.997
$$

This is exactly the empirical rule.

- Variance
- Scale

but does NOT change:

- Skewness
- Kurtosis

If data is right-skewed before standardization:

it remains right-skewed afterward.

If data is Gaussian:

it remains Gaussian.

Only the scale changes.

Salary dominates distance calculations.

```text
Age     -> mean 0 std 1
Salary  -> mean 0 std 1
```

Both features contribute fairly.

# Kernel Density Estimation (KDE)

## Motivation

So far, we have learned:

- Histograms
- Probability Density Function (PDF)
- Cumulative Distribution Function (CDF)
- Gaussian Distribution

A natural question arises:

> How do we obtain a smooth PDF from a finite set of observed data points?

This is exactly what **Kernel Density Estimation (KDE)** does.

KDE is one of the most important concepts in Statistics because it allows us to estimate the underlying probability density of data without assuming a specific distribution (such as Gaussian).

---

# The Problem with Histograms

Suppose we have the following observations:

```text
1, 2, 2.5, 3, 5, 6
```

A histogram groups these observations into bins.

Example:

```text
Bin 1 : [0,2]
Bin 2 : [2,4]
Bin 3 : [4,6]
Bin 4 : [6,8]
```

Histogram:

```text
Frequency
 ^
 |
 |      ███
 |      ███
 |      ███       ██
 |      ███       ██
 +------------------------>
      0   2   4   6   8
```

The histogram tells us:

> How many points fall inside each bin.

---

## Limitation of Histograms

Histograms are **discrete** and highly dependent on bin size.

Consider the same data:

### Small Bin Width

```text
|█|█|█|█|█|█|
```

Looks noisy.

---

### Large Bin Width

```text
|██████████|
```

Looks overly smooth.

---

Changing the bin width can completely change the shape of the histogram.

Therefore:

> Histograms are useful for visualization but not ideal for estimating the true underlying distribution.

---

# KDE: The Idea

Instead of creating rectangular bins, KDE creates a smooth curve.

The intuition is:

> Every data point contributes a small smooth bump (kernel).

Then we add all those bumps together.

The resulting smooth curve is the estimated PDF.

---

# What is a Kernel?

A kernel is simply:

> A smooth function centered at a data point.

Most commonly:

- Gaussian kernel (bell-shaped)
- Epanechnikov kernel
- Uniform kernel
- Triangular kernel

In practice:

> Gaussian kernels are used most of the time.

---

## Visual Intuition

Suppose data points are:

```text
1    2    3    5    6
```

Instead of drawing bars:

```text
      /\      /\      /\          /\      /\
     /  \    /  \    /  \        /  \    /  \
----/----\--/----\--/----\------/----\--/----\-----
    1      2      3          5      6
```

Each point gets its own Gaussian bump.

---

# Building the KDE

Step 1:

Place one kernel on every observation.

```text
x1 = 1
x2 = 2
x3 = 3
x4 = 5
x5 = 6
```

---

Step 2:

Add all kernel heights together.

```text
Total Density
=
Kernel1
+
Kernel2
+
Kernel3
+
Kernel4
+
Kernel5
```

---

Step 3:

The resulting smooth curve becomes:

```text
                /\
              /    \
            /        \__
          /              \_
--------/------------------\--------
```

This curve is the KDE estimate of the PDF.

---

# Important Insight

Think of every data point dropping a small hill on the ground.

```text
Point → Small Hill
```

Where hills overlap:

```text
Hill + Hill + Hill
```

the height becomes larger.

Thus:

> Regions with many nearby observations produce high density.

---

# KDE Formula

Mathematically:

$$
\hat f(x)
=
\frac{1}{nh}
\sum_{i=1}^{n}
K
\left(
\frac{x-x_i}{h}
\right)
$$

---

## Understanding Every Symbol

### Estimated Density

$$
\hat f(x)
$$

Estimated PDF at point x.

---

### Number of Observations

$$
n
$$

Total number of data points.

---

### Data Point

$$
x_i
$$

Individual observation.

---

### Kernel Function

$$
K(\cdot)
$$

Shape of the bump.

Usually Gaussian.

---

### Bandwidth

$$
h
$$

Controls width of each kernel.

This is the most important parameter in KDE.

---

# Bandwidth (h)

Bandwidth determines:

> How wide each kernel should be.

You can think of it as:

```text
Bandwidth = Smoothing Parameter
```

---

# Why Bandwidth Matters

The kernel shape is usually fixed.

The bandwidth controls:

- Smoothness
- Noise
- Detail

of the estimated PDF.

---

# Small Bandwidth

Small h:

```text
       /\    /\    /\
      /  \  /  \  /  \
-----/----\/----\/----\-------
```

Very wiggly curve.

Captures every tiny fluctuation.

---

## Problem

Overfitting.

The curve starts modeling noise.

---

### Interpretation

```text
Bias      ↓
Variance  ↑
```

---

# Large Bandwidth

Large h:

```text
           ______
         /        \
-------/------------\-------
```

Very smooth.

---

## Problem

Underfitting.

Important structure may disappear.

---

### Interpretation

```text
Bias      ↑
Variance  ↓
```

---

# Moderate Bandwidth

This is usually ideal.

```text
           /\
         /    \
       /        \
------/----------\------
```

Captures the true shape without excessive noise.

---

# Relationship with Bias-Variance Tradeoff

KDE is a beautiful example of the Bias-Variance Tradeoff.

| Bandwidth | Bias | Variance |
|------------|---------|-----------|
| Small | Low | High |
| Large | High | Low |
| Moderate | Balanced | Balanced |

---

# Understanding the Images

In the images:

### Red Curve

Small bandwidth.

```text
h ↓
```

Very jagged.

Overfits.

---

### Black Curve

Moderate bandwidth.

Reasonable estimate.

Usually preferred.

---

### Green Curve

Large bandwidth.

```text
h ↑
```

Too smooth.

Underfits.

---

# Density Interpretation

Suppose KDE value is high at:

```text
x = 3
```

This means:

> Many observations are concentrated around x = 3.

---

Suppose KDE value is low at:

```text
x = 10
```

This means:

> Very few observations exist around x = 10.

---

# KDE and PDF

One of the most important ideas:

> KDE is a method for estimating a PDF.

---

### PDF

Represents:

$$
f(x)
$$

the underlying probability density.

---

### KDE

Represents:

$$
\hat f(x)
$$

our estimate of the PDF from finite samples.

---

Therefore:

```text
True PDF
     ↓
Unknown
     ↓
Observed Data
     ↓
KDE
     ↓
Estimated PDF
```

---

# KDE vs Histogram

| Histogram | KDE |
|------------|------------|
| Discrete bars | Smooth curve |
| Depends on bin size | Depends on bandwidth |
| Abrupt changes | Smooth transitions |
| Easy to compute | More sophisticated |
| Rough estimate | Better density estimate |

# KDE vs Gaussian Distribution

Students often confuse these.

They are completely different.

---

## Gaussian Distribution

Assumes:

$$
X \sim N(\mu,\sigma^2)
$$

Specific distribution family.

Bell-shaped.

---

## KDE

Makes no distribution assumption.

Can estimate:

- Single peak
- Multiple peaks
- Skewed distributions
- Complex distributions

---

### Example

Data:

```text
2 2 2 3 3 3 8 8 8
```

KDE may reveal:

```text
Peak 1 near 2.5

Peak 2 near 8
```

This is called a:

**Bimodal Distribution**

A Gaussian would struggle to model this.

---

# KDE and Non-Parametric Statistics

KDE belongs to:

> Non-parametric statistics

because it does not assume a fixed distribution form.

---

## Parametric Methods

Assume:

$$
N(\mu,\sigma^2)
$$

Need only:

- Mean
- Variance

---

## Non-Parametric Methods

Do not assume any fixed shape.

The data determines the shape.

Examples:

- KDE
- KNN
- Decision Trees
