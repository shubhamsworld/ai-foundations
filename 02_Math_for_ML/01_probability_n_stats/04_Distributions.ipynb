# Introduction to Probability and Statistics

## Why Do We Need Probability in Machine Learning?

Some datasets contain classes that are clearly separated.

In such cases, simple rules may be sufficient and complex Machine Learning models may not be necessary.

However, in many real-world datasets, classes overlap with each other.

Consider a query point:

$$X_q$$

that lies in a mixed region.

Instead of saying:

> This point definitely belongs to Class A.

we often assign probabilities:

$$P(\text{Versicolor}\mid X_q)=0.8$$

$$P(\text{Virginica}\mid X_q)=0.2$$

$$P(\text{Setosa}\mid X_q)=0$$

This means:

* 80% chance of being Versicolor
* 20% chance of being Virginica
* 0% chance of being Setosa

Probability allows Machine Learning models to express uncertainty rather than making absolute decisions.

---

## What Falls Under Probability and Statistics?

Many important concepts used in Machine Learning come from Probability and Statistics:

* Mean
* Median
* Variance
* Standard Deviation
* Histograms
* Percentiles
* Quartiles
* Probability Distributions
* PDF (Probability Density Function)
* CDF (Cumulative Distribution Function)

These tools help us understand data before building Machine Learning models.

# Random Variables

## Definition

A **Random Variable (RV)** is a variable whose value depends on the outcome of a random experiment.

A random variable can take different values depending on what outcome occurs.

---

## Types of Random Variables

There are two major types:

1. Discrete Random Variables
2. Continuous Random Variables

---

## Discrete Random Variable

A discrete random variable can take values from a finite or countable set.

### Example: Dice Roll

Let

$$X$$

represent the outcome of rolling a fair six-sided die.

Possible values:

$$X={1,2,3,4,5,6}$$

Since the variable can only take one value from this finite set, it is a **Discrete Random Variable**.

### Other Examples

* Number of students in a classroom
* Number of cars in a parking lot
* Number of heads obtained in coin tosses

These are countable quantities.

---

## Continuous Random Variable

A continuous random variable can take any real value within a range.

### Example: Height of a Student

Let

$$Y$$

represent the height of a randomly selected student.

Possible values:

$$162.45 \text{ cm}$$

$$171.23 \text{ cm}$$

$$183.789 \text{ cm}$$

Since infinitely many values are possible,

$$Y$$

is a **Continuous Random Variable**.

### Other Examples

* Height
* Weight
* Temperature
* Age
* Salary

These quantities are measured rather than counted.

---

# Probability Examples

## Example: Fair Dice

Consider a fair six-sided die.

Possible outcomes:

$$X={1,2,3,4,5,6}$$

Since all outcomes are equally likely:

$$P(X=1)=\frac{1}{6}$$

$$P(X=2)=\frac{1}{6}$$

$$P(X=3)=\frac{1}{6}$$

and so on.

---

## Probability of Getting an Even Number

Even numbers are:

$${2,4,6}$$

Therefore:

$$P(X\text{ is even})=P(X=2)+P(X=4)+P(X=6)$$

$$=\frac{1}{6}+\frac{1}{6}+\frac{1}{6}$$

$$=\frac{3}{6}$$

$$=\frac{1}{2}$$

Thus,

$$P(X\text{ is even})=0.5$$

---

## Probability of Getting an Odd Number

Odd numbers are:

$${1,3,5}$$

Therefore:

$$P(X\text{ is odd})=P(X=1)+P(X=3)+P(X=5)$$

$$=\frac{1}{6}+\frac{1}{6}+\frac{1}{6}$$

$$=\frac{3}{6}$$

$$=\frac{1}{2}$$

Thus,

$$P(X\text{ is odd})=0.5$$

Example:

```text
150
155
160
158
162
157
500
```

The value:

```text
500
```

Therefore it is an outlier.

---

## Why Do Outliers Occur?

Outliers may occur because of:

* Human error
* Data entry error
* Sensor error
* Measurement error

## Why Are Outliers Important?

Outliers can significantly affect:

* Mean

# Population and Sample

## Population

A population is the complete set of observations we are interested in studying.

Examples:

* All humans in the world
* All customers of a company
* All flowers of a species

---

## Sample

A sample is a subset of the population.

Instead of studying the entire population, we collect a smaller sample and analyze it.

---

## Why Do We Need Sampling?

Suppose we want to estimate the average height of all humans on Earth.

Let

$$N \approx 7 \text{ billion}$$

It is practically impossible to measure every person's height.

Instead, we randomly select a sample.

Example:

```text
Sample Size = 1000
```

and estimate population characteristics from that sample.

---

## Population Mean

The mean of the entire population is denoted by:

$$\mu$$

If the population contains

$$N$$

observations:

$$\mu=\frac{1}{N}\sum_{i=1}^{N}x_i$$

where:

* (N) = population size
* (x_i) = ith observation

---

## Sample Mean

The mean of a sample is denoted by:

$$\bar{x}$$

If the sample contains

$$n$$

observations:

$$\bar{x}=\frac{1}{n}\sum_{i=1}^{n}x_i$$

where:

* (n) = sample size
* (x_i) = ith sample observation

---

## Important Idea

As the sample size increases:

$$\bar{x}\rightarrow\mu$$

This means:

> The sample mean becomes a better estimate of the population mean.

This idea is fundamental in Statistics and Machine Learning.

---

## Random Sampling

A sample collected randomly from the population is called a **Random Sample**.

Random sampling helps ensure that the sample represents the population fairly and reduces bias.

---

# Gaussian (Normal) Distribution

## Why Do We Study Gaussian Distribution?

Many real-world quantities tend to follow a bell-shaped pattern:

- Human heights
- Human weights
- Exam scores
- Measurement errors
- Sensor noise

Because of this, Gaussian Distribution is one of the most important distributions in Statistics and Machine Learning.

It is also called the **Normal Distribution**.

---

## Bell-Shaped Curve

A Gaussian distribution looks like a symmetric bell-shaped curve.

Characteristics:

- Highest point occurs at the center.
- Values near the center occur most frequently.
- Values far from the center occur less frequently.
- Left and right sides are mirror images.

```text
                 ^
                 |
               ***
             **   **
            *       *
          **         **
---------*-----μ-------*-------->
```

---

# Parameters of a Gaussian Distribution

A Gaussian distribution is completely described by two parameters:

1. Mean ($\mu$)
2. Variance ($\sigma^2$)

Notation:

$$
X \sim N(\mu,\sigma^2)
$$

Read as:

> Random variable X follows a Normal Distribution with mean μ and variance σ².

---

## Mean ($\mu$)

The mean determines the center of the distribution.

Example:

$$
\mu = 50
$$

means the curve is centered at 50.

```text
            μ=50
              |
              v
            *****
          **     **
---------*---------*--------->
```

If μ changes:

- Curve shifts left or right.
- Shape remains the same.

---

## Variance ($\sigma^2$)

Variance controls how spread out the data is.

### Small Variance

Data points stay close to the mean.

```text
                *
              *   *
             *     *
------------*---μ---*------------>
```

### Large Variance

Data points spread farther away from the mean.

```text
          *             *
        *                 *
------*--------μ-----------*------->
```

Therefore:

- Small variance → narrow curve
- Large variance → wider curve

---

## Standard Deviation

Standard deviation is simply the square root of variance.

$$
\sigma = \sqrt{\sigma^2}
$$

Interpretation:

- Small σ → less spread
- Large σ → more spread

---

# Probability Density Function (PDF)

The PDF tells us:

> How likely values around x are.

For a Gaussian distribution:

$$
f(x)=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(
-\frac{(x-\mu)^2}{2\sigma^2}
\right)
$$

where:

- μ = mean
- σ = standard deviation
- exp() = exponential function

---

## Understanding the Formula Intuitively

Ignore constants for a moment:

$$
f(x) \propto
\exp(-x^2)
$$

Focus on:

$$
y = e^{-x^2}
$$

Let's evaluate a few values:

| x | y = e^(-x²) |
|---|---|
| 0 | 1 |
| 1 | 0.3679 |
| 2 | 0.0183 |
| 3 | 0.00012 |

Notice:

As x moves away from 0,

$$
x^2
$$

becomes larger,

therefore

$$
e^{-x^2}
$$

becomes very small.

This is why the Gaussian curve drops rapidly on both sides.

---

## Key Properties of PDF

### 1. Maximum at Mean

The highest probability density occurs at:

$$
x=\mu
$$

The center of the curve is the peak.

---

### 2. Symmetric

For a Gaussian distribution:

$$
f(\mu-a)=f(\mu+a)
$$

Both sides are mirror images.

---

### 3. Density Decreases Away from Mean

As distance from mean increases:

$$
|x-\mu|
$$

increases,

therefore PDF decreases.

---

### 4. Total Area Equals 1

The area under the entire PDF curve is:

$$
1
$$

This represents 100% probability.

```text
           PDF
            ^
            |
         ********
       ***      ***
     ***          ***
----*--------------*-------->
       Entire Area = 1
```

---

# Important Note

For continuous random variables:

$$
P(X=x)=0
$$

for any exact value x.

Instead we calculate probabilities over intervals:

$$
P(a \le X \le b)
$$

which corresponds to the area under the curve between a and b.

---

# Cumulative Distribution Function (CDF)

The CDF answers a different question.

Instead of asking:

> How dense is probability around x?

it asks:

> What is the probability of observing a value less than or equal to x?

Definition:

$$
F(x)=P(X\le x)
$$

---

## Relationship Between PDF and CDF

CDF is obtained by accumulating area under the PDF.

$$
F(x)
=
\int_{-\infty}^{x}
f(t)\,dt
$$

This means:

CDF = area under PDF from negative infinity up to x.

---

## Shape of a Gaussian CDF

The Gaussian PDF is bell-shaped.

The Gaussian CDF is S-shaped.

```text
CDF
 ^
1|                         ****
 |                     ****
 |                 ****
 |             ****
0.5----------***
 |        ***
 |     ***
 |  ***
0+---------------------------->
                 μ
```

---

## Important Property

At the mean:

$$
x=\mu
$$

the CDF value is:

$$
F(\mu)=0.5
$$

Why?

Because the Gaussian distribution is symmetric.

Half of the area lies to the left of μ.

Half lies to the right.

Therefore:

$$
P(X\le \mu)=0.5
$$

---

## Standard Normal Distribution

When:

$$
\mu=0
$$

and

$$
\sigma=1
$$

the distribution is called the Standard Normal Distribution.

Notation:

$$
Z \sim N(0,1)
$$

---

### Useful Values

$$
CDF(0)=0.5
$$

$$
CDF(1)\approx0.8413
$$

$$
CDF(-1)\approx0.1587
$$

---

# Effect of Variance on PDF

### Smaller Variance

```text
                *
              *   *
             *     *
------------*---μ---*------------>
```

- Narrow
- Taller
- Less spread

---

### Larger Variance

```text
          *             *
        *                 *
------*--------μ-----------*------->
```

- Wider
- Shorter
- More spread

---

# Effect of Variance on CDF

When variance is small:

- CDF rises very sharply near μ.

When variance is large:

- CDF rises slowly.

```text
Small Variance:

0 -----|████|----- 1

Large Variance:

0 --████████████-- 1
```
