# Probability Distributions -

# Why Do We Need Probability Distributions?

Suppose we collect baby weights:

| Baby | Weight (kg) |
|--------|--------|
| 1 | 2.5 |
| 2 | 3.1 |
| 3 | 2.9 |
| 4 | 3.3 |
| 5 | 2.8 |

We have raw numbers.

But raw numbers do not answer questions like:

- What weight is considered normal?
- What is the average baby weight?
- How likely is a baby to weigh more than 4kg?
- What range contains most babies?

To answer such questions we need a **probability distribution**.

A distribution summarizes how data behaves.

---

# What Does a Distribution Do?

A distribution tells us:

1. What values are likely?
2. What values are unlikely?
3. Where the center lies?
4. How much variability exists?

Instead of storing millions of observations, we store a few parameters.

Example:

Baby weights:

$$
X \sim N(\mu=3.2,\sigma=0.5)
$$

This single statement summarizes thousands of babies.

---

Examples:

- Number of heads in 10 coin tosses

- Waiting time for bus
- Height of a student
- Weight of a baby

Notation:

$$
X
$$

$$

Examples:

---

### Continuous RV

Possible values:

$$

- Height

- Salary

# PMF vs PDF

---

## PMF (Probability Mass Function)

Used for discrete random variables.

Direct probability exists.

Example:

Coin toss:

$$
P(X=1)=0.5
$$

$$
P(X=0)=0.5
$$

---

## PDF (Probability Density Function)

Used for continuous variables.

Exact probability is always:

$$
P(X=x)=0
$$

Instead:

$$
P(a<X<b)
$$

Area under curve.

Example:

$$
P(2kg < Weight < 3kg)
$$

---

# Conditions for a Probability Distribution

Probabilities must satisfy:

## Rule 1

Probability can never be negative.

$$
P(X=x)\ge0
$$

---

## Rule 2

Total probability equals 1.

$$
\sum P(X=x)=1
$$

Discrete case.

or

$$
\int f(x)dx=1
$$

Continuous case.

# Poisson Distribution 
---

# Why Do We Need Poisson Distribution?

Many real-world problems involve **counting events**.

Examples:

- Number of visitors to a website in the next hour
- Number of calls received in a call center per minute
- Number of goals scored in a football match
- Number of accidents at a traffic signal per day
- Number of defects in a manufactured sheet
- Number of emails received in an hour

In all these problems:

We are counting:

$$
0,1,2,3,4,...
$$

events occurring within a fixed interval.

This is where Poisson Distribution comes into the picture.

---

# Core Idea

Poisson Distribution models:

> Number of events occurring in a fixed interval of time, area, distance, volume, or space.

Examples:

| Interval | Event |
|-----------|-----------|
| 1 hour | Website visitors |
| 1 day | Accidents |
| 90 minutes | Goals in football |
| 1 meter | Defects in steel |
| 1 minute | Incoming calls |

---

# Random Variable

Let

$$
X
$$

denote:

Number of events in a fixed interval.

Examples:

### Website

$$
X = \text{Number of visitors in next hour}
$$

---

### Football

$$
X = \text{Number of goals in a match}
$$

---

### Call Center

$$
X = \text{Number of calls in next minute}
$$

Since X counts events:

$$
X \in \{0,1,2,3,\dots\}
$$

Therefore Poisson is a:

**Discrete Random Variable**

---

# When Can We Use Poisson?

Poisson can only be used when certain assumptions hold.

---

## Condition 1: Constant Average Rate

Events occur at an average rate:

$$
\lambda
$$

which remains approximately constant.

Example:

Website receives:

$$
20
$$

visitors/hour on average.

Then

$$
\lambda = 20
$$

---

### Good Example

Visitors from:

7 PM to 8 PM

Average:

$$
20/hour
$$

---

### Bad Example

Website traffic:

3 AM = 2 visitors/hour

8 PM = 500 visitors/hour

Rate is not constant.

Poisson assumption becomes weak.

---

## Condition 2: Events Are Independent

One event should not affect another.

Example:

One visitor visiting website should not change probability of another visitor.

---

### Good Example

Visitors from different cities.

Independent.

---

### Bad Example

A viral Instagram reel.

One person shares it.

Thousands immediately arrive.

Events become dependent.

Poisson assumption breaks.

---

## Condition 3: No Simultaneous Events

Two events should not occur exactly at same instant.

Mathematically:

Probability of multiple events in an infinitesimally small interval should be negligible.

---

### Example

Website visitors:

In one microsecond:

Either

- no visitor
- one visitor

should happen.

Not hundreds simultaneously.

---

# Poisson Process

When all three assumptions hold:

1. Constant rate
2. Independent events
3. No simultaneous events

The underlying mechanism is called:

**Poisson Process**

---

Poisson Process generates:

$$
X \sim Poisson(\lambda)
$$

# Why Mean = Variance?

Poisson is one of the few distributions where:

$$

$$

This is a signature property.

Whenever data has:

$$
Mean \approx Variance
$$

Poisson becomes a strong candidate.

---

# MGF of Poisson

Moment Generating Function:

$$
M_X(t)
=
E[e^{tX}]
$$

For Poisson:

$$
M_X(t)
=
\exp
\left(
\lambda(e^t-1)
\right)
$$

Used to derive:

- Mean
- Variance
- Higher moments

---

# Deriving Variance

Recall:

$$
Var(X)
=
E[X^2]
-
(E[X])^2
$$

Using MGF:

$$
E[X]=\lambda
$$

and

$$
E[X^2]
=
\lambda+\lambda^2
$$

Therefore:

$$
Var(X)
=
\lambda
$$

---

## Medium λ

$$

$$

---

$$

Probability distributions are used to model uncertainty.

A very important chain of distributions is:

```text
Bernoulli  →  Binomial  →  Poisson Process
                                |
                                ↓
                        Poisson Distribution
                                |
                                ↓
                    Exponential Distribution
```

Think of them as answering different questions:

| Distribution | Answers |
|------------|------------|

| Poisson | How many events occur in a fixed interval? |
| Exponential | How long until the next event occurs? |

# Poisson Process

A Poisson Process is the mechanism that generates events randomly over time.

Imagine a timeline.

```text
Time ---------------------------------------------------->

     |     |  |       ||    |     |||      |      |
   Event Event Event Event Event Event Event Event
```

Events occur randomly.

The gaps between events are:

$$
\Delta t_1,\Delta t_2,\Delta t_3,...
$$

Those gaps follow:

$$
\Delta t_i \sim Exponential(\lambda)
$$

---

# Conditions for a Poisson Process

For a process to be modeled as a Poisson Process, three conditions should hold.

## 1. Constant Rate

Events occur at a constant average rate.

$$
\lambda = \text{constant}
$$

Example:

```text
Average 5 calls/hour
Average 2.5 goals/match
Average 10 customers/minute
```

Good:

```text
Meteor strikes
Radioactive decay
Website requests
```

Bad:

```text
Traffic during day/night
Store customers during festival season
```

Rate changes with time.

---

## 2. Independent Events

One event should not affect another.

Good:

```text
Meteor strikes
Machine failures
Incoming emails
```

Bad:

```text
Earthquakes and aftershocks
Disease outbreaks
Social media viral posts
```

One event increases chance of another.

---

## 3. No Simultaneous Events

Two events cannot happen at exactly the same instant.

Mathematically:

$$
P(\text{2 or more events in tiny interval } dt)
\approx 0
$$

or more precisely:

$$
P(N(dt)\ge2)=o(dt)
$$

---

$$
\lambda

> Average number of events per unit time.

Examples:

| Scenario | λ |
|-----------|-----------|
| Goals per match | 2.5 |
| Calls per hour | 20 |
| Customers per minute | 4 |
| Meteor strikes per century | 1 |

# Geometric Distribution — Complete Intuition, Mathematics, Memorylessness, and Connections

---

# Big Picture

The Geometric Distribution is the **discrete version of the Exponential Distribution**.

Think of the relationship:

```text
Continuous Time World
---------------------
Poisson Process
        ↓
Exponential Distribution
(waiting time)

Discrete Trial World
---------------------
Bernoulli Trials
        ↓
Geometric Distribution
(waiting trials)
```

---

# The Main Question

Geometric Distribution answers:

> How many independent Bernoulli trials do I need to perform before getting the first success?

or

> How many failures occur before the first success?

Depending on convention.

---

# Building Block: Bernoulli Trial

A Bernoulli Trial is the simplest random experiment.

Only two outcomes are possible:

```text
Success (S)
Failure (F)
```

Examples:

| Experiment | Success |
|------------|----------|
| Coin toss | Head |
| Exam question | Correct answer |
| Manufacturing | Non-defective item |
| Customer call | Sale happens |
| Email campaign | Customer clicks |

---

# Bernoulli Distribution

If:

$$
X \sim Bernoulli(p)
$$

then

$$
P(X=1)=p
$$

$$
P(X=0)=1-p
$$

where

```text
1 = Success
0 = Failure
```

---

# Example

Suppose:

$$
p = 0.3
$$

Then:

```text
Success = 30%
Failure = 70%
```

Every trial looks like:

```text
Trial 1 → S or F
Trial 2 → S or F
Trial 3 → S or F
...
```

---

# Conditions Required for Geometric Distribution

A Geometric Distribution exists only when:

## Condition 1

Trials are independent.

Meaning:

```text
Result of Trial 1
does NOT affect
Result of Trial 2
```

Example:

Coin tosses.

Good.

---

Bad example:

```text
Drawing cards WITHOUT replacement
```

Because probabilities change.

---

## Condition 2

Probability of success remains constant.

Same:

$$
p
$$

for every trial.

Example:

```text
P(Head)=0.5
P(Head)=0.5
P(Head)=0.5
...
```

Constant.

---

Bad example:

```text
Weather prediction
Stock market
Changing business environment
```

Probability changes.

---

## Condition 3

Stop immediately after first success.

This is extremely important.

Geometric Distribution is about:

> Waiting for FIRST success.

---

# Visual Intuition

Suppose:

$$
p=0.3
$$

Sequence:

```text
Trial 1 → F
Trial 2 → F
Trial 3 → F
Trial 4 → F
Trial 5 → S
```

We stop.

---

# Two Common Definitions

Different books define Geometric Distribution differently.

---

## Definition A (Most Common in ML)

Random variable:

$$
X
$$

=

Number of trials until first success.

Example:

```text
F F F F S
```

Total trials:

$$
X=5
$$

because success occurred on trial 5.

Possible values:

$$
1,2,3,4,...
$$

---

## Definition B

Random variable:

$$
X
$$

=

Number of failures before first success.

Example:

```text
F F F F S
```

Failures:

$$
X=4
$$

Possible values:

$$
0,1,2,3,...
$$

---

# Which Definition Is Used Most?

Most ML, Data Science, Statistics books use:

> Number of trials until first success

Therefore:

$$
X=1,2,3,...
$$

We will use this definition.

---

# PMF Derivation

Suppose:

$$
X=k
$$

Meaning:

```text
F F F ... F S
```

Exactly:

```text
(k−1) failures
1 success
```

---

Example:

$$
X=5
$$

Means:

```text
F F F F S
```

---

Because trials are independent:

$$
P(X=5)
=
P(F)P(F)P(F)P(F)P(S)
$$

---

Substitute:

$$
P(F)=1-p
$$

$$
P(S)=p
$$

Therefore:

$$
P(X=5)
=
(1-p)^4p
$$

---

Generalizing:

$$
P(X=k)
=
(1-p)^{k-1}p
$$

---

# PMF

The Probability Mass Function is:



for

$$
k=1,2,3,...
$$

---

# Example

Suppose:

$$
p=0.3
$$

Find:

$$
P(X=5)
$$

---

Solution:

$$
P(X=5)
=
(1-0.3)^4(0.3)
$$

$$
=
(0.7)^4(0.3)
$$

$$
=
0.07203
$$

Approximately:

$$
7.2\%
$$

---

# Interpretation

Probability that:

```text
Fail
Fail
Fail
Fail
Success
```

occurs exactly.

---

# PMF Shape

The largest probability is always at:

$$
X=1
$$

because:

```text
Success immediately
```

is most likely.

As:

$$
k
$$

increases,

probability decreases exponentially.

---

Example:

For:

$$
p=0.3
$$

```text
P(X=1)=0.30
P(X=2)=0.21
P(X=3)=0.147
P(X=4)=0.1029
...
```

---

# CDF

CDF answers:

> Probability first success occurs on or before trial k.

Definition:

$$
F(k)
=
P(X\le k)
$$

---

Using geometric series:

$$
F(k)
=
1-(1-p)^k
$$

---

# Example

Suppose:

$$
p=0.3
\qquad
k=5
$$

Then:

$$
P(X\le5)
=
1-(0.7)^5
$$

$$
=
0.83193
$$

Therefore:

$$
83.2\%
$$

chance of obtaining a success within first 5 trials.

---

# Survival Function

Probability that success has NOT occurred yet.

$$
P(X>k)
$$

Meaning:

```text
Still no success after k trials
```

Formula:

$$
P(X>k)
=
(1-p)^k
$$

---

# Example

Suppose:

$$
p=0.3
$$

Probability no success in first 5 trials:

$$
P(X>5)
=
(0.7)^5
$$

$$
=
0.16807
$$

---

# Mean (Expected Value)

One of the most important formulas.

For geometric distribution:

$$
E[X]
=
\frac1p
$$

---

# Interpretation

Average number of trials needed to obtain first success.

---

Examples

| p | Mean |
|----|----|
| 0.5 | 2 |
| 0.25 | 4 |
| 0.1 | 10 |
| 0.01 | 100 |

---

# Real Meaning

Suppose:

$$
p=0.1
$$

A customer purchases with probability 10%.

Then:

$$
E[X]=10
$$

On average:

```text
Need 10 customers
for first purchase
```

---

# Variance

$$
Var(X)
=
\frac{1-p}{p^2}
$$

---

# Standard Deviation

$$
\sigma
=
\sqrt{\frac{1-p}{p^2}}
$$

---

# Memoryless Property

This is the MOST IMPORTANT property.

Only two distributions possess memorylessness:

```text
Geometric (Discrete)
Exponential (Continuous)
```

---

# Intuition

Suppose:

```text
20 trials already failed
```

Question:

> Does this make success more likely now?

Answer:

NO.

The process forgets history.

---

Mathematically:

$$
P(X>s+t\mid X>s)
=
P(X>t)
$$

---

# Example

Suppose:

$$
p=0.3
$$

You've already failed:

```text
10 times
```

Probability of waiting 5 more failures:

Same as if you started today.

Past does not matter.

---

# Why Memorylessness Happens

Every trial is independent.

```text
F F F F F F F F F F
```

does not change future probabilities.

Trial 11 still has:

$$
P(S)=0.3
$$
