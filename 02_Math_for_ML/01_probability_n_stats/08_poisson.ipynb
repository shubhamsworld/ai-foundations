# Probability Distributions - 
---

# Why Do We Need Probability Distributions?

Suppose we collect baby weights:

| Baby | Weight (kg) |
|--------|--------|
| 1 | 2.5 |
| 2 | 3.1 |
| 3 | 2.9 |
| 4 | 3.3 |
| 5 | 2.8 |

We have raw numbers.

But raw numbers do not answer questions like:

- What weight is considered normal?
- What is the average baby weight?
- How likely is a baby to weigh more than 4kg?
- What range contains most babies?

To answer such questions we need a **probability distribution**.

A distribution summarizes how data behaves.

---

# What Does a Distribution Do?

A distribution tells us:

1. What values are likely?
2. What values are unlikely?
3. Where the center lies?
4. How much variability exists?

Instead of storing millions of observations, we store a few parameters.

Example:

Baby weights:

$$
X \sim N(\mu=3.2,\sigma=0.5)
$$

This single statement summarizes thousands of babies.

---

# Random Variable (RV)

A random variable is a variable whose value depends on randomness.

Examples:

- Number of heads in 10 coin tosses
- Number of website visitors in next hour
- Waiting time for bus
- Height of a student
- Weight of a baby

Notation:

$$
X
$$

Examples:

### Discrete RV

Possible values:

$$
X \in \{0,1,2,3,4,\dots\}
$$

Examples:

- Number of heads
- Number of customers
- Number of defects

---

### Continuous RV

Possible values:

$$
X \in [0,\infty)
$$

Examples:

- Height
- Weight
- Temperature
- Salary

---

# Discrete vs Continuous

| Discrete | Continuous |
|-----------|-----------|
| Counting | Measuring |
| 0,1,2,3... | Infinite values |
| PMF | PDF |
| Binomial | Gaussian |

Examples:

### Discrete

Number of heads:

$$
0,1,2,\dots,10
$$

### Continuous

Weight:

$$
2.43,2.431,2.4317,...
$$

Infinite possibilities.

# PMF vs PDF

---

## PMF (Probability Mass Function)

Used for discrete random variables.

Direct probability exists.

Example:

Coin toss:

$$
P(X=1)=0.5
$$

$$
P(X=0)=0.5
$$

---

## PDF (Probability Density Function)

Used for continuous variables.

Exact probability is always:

$$
P(X=x)=0
$$

Instead:

$$
P(a<X<b)
$$

Area under curve.

Example:

$$
P(2kg < Weight < 3kg)
$$

---

# Conditions for a Probability Distribution

Probabilities must satisfy:

## Rule 1

Probability can never be negative.

$$
P(X=x)\ge0
$$

---

## Rule 2

Total probability equals 1.

$$
\sum P(X=x)=1
$$

Discrete case.

or

$$
\int f(x)dx=1
$$

Continuous case.

# Bernoulli Distribution

Smallest possible distribution.

Only two outcomes.

$$
X=
\begin{cases}
1 & \text{Success}\\
0 & \text{Failure}
\end{cases}
$$

---

## Examples

### Student attends class

$$
1 = Present
$$

$$
0 = Absent
$$

---

### Email opened?

$$
1 = Yes
$$

$$
0 = No
$$

---

### Loan default?

$$
1 = Default
$$

$$
0 = No Default
$$

---

## Parameter

Probability of success:

$$
p
$$

Probability of failure:

$$
1-p
$$

---

## Mean

$$
E[X]=p
$$

---

## Variance

$$
Var(X)=p(1-p)
$$

---

# Binomial Distribution

A collection of Bernoulli trials.

Question:

How many successes occur in n trials?

---

## Example

10 coin tosses.

How many heads?

Each toss:

Bernoulli.

Entire experiment:

Binomial.

---

Notation:

$$
X\sim Bin(n,p)
$$

where

- n = number of trials
- p = probability of success

---

## Example

10 tosses

$$
n=10
$$

Probability of head:

$$
p=0.5
$$

Question:

How many heads?

$$
X\sim Bin(10,0.5)
$$

---

## Mean

$$
E[X]=np
$$

Example:

$$
10\times0.5=5
$$

Expected heads = 5.

---

## Variance

$$
Var(X)=np(1-p)
$$

---

## Standard Deviation

$$
\sigma=\sqrt{np(1-p)}
$$

---

# Geometric Distribution

Question:

How many failures before first success?

---

Example

Interview applications.

Success = offer letter.

Suppose:

$$
p=0.1
$$

Question:

How many rejections before first offer?

Geometric Distribution.

# Poisson Distribution 
---

# Why Do We Need Poisson Distribution?

Many real-world problems involve **counting events**.

Examples:

- Number of visitors to a website in the next hour
- Number of calls received in a call center per minute
- Number of goals scored in a football match
- Number of accidents at a traffic signal per day
- Number of defects in a manufactured sheet
- Number of emails received in an hour

In all these problems:

We are counting:

$$
0,1,2,3,4,...
$$

events occurring within a fixed interval.

This is where Poisson Distribution comes into the picture.

---

# Core Idea

Poisson Distribution models:

> Number of events occurring in a fixed interval of time, area, distance, volume, or space.

Examples:

| Interval | Event |
|-----------|-----------|
| 1 hour | Website visitors |
| 1 day | Accidents |
| 90 minutes | Goals in football |
| 1 meter | Defects in steel |
| 1 minute | Incoming calls |

---

# Random Variable

Let

$$
X
$$

denote:

Number of events in a fixed interval.

Examples:

### Website

$$
X = \text{Number of visitors in next hour}
$$

---

### Football

$$
X = \text{Number of goals in a match}
$$

---

### Call Center

$$
X = \text{Number of calls in next minute}
$$

Since X counts events:

$$
X \in \{0,1,2,3,\dots\}
$$

Therefore Poisson is a:

**Discrete Random Variable**

---

# When Can We Use Poisson?

Poisson can only be used when certain assumptions hold.

---

## Condition 1: Constant Average Rate

Events occur at an average rate:

$$
\lambda
$$

which remains approximately constant.

Example:

Website receives:

$$
20
$$

visitors/hour on average.

Then

$$
\lambda = 20
$$

---

### Good Example

Visitors from:

7 PM to 8 PM

Average:

$$
20/hour
$$

---

### Bad Example

Website traffic:

3 AM = 2 visitors/hour

8 PM = 500 visitors/hour

Rate is not constant.

Poisson assumption becomes weak.

---

## Condition 2: Events Are Independent

One event should not affect another.

Example:

One visitor visiting website should not change probability of another visitor.

---

### Good Example

Visitors from different cities.

Independent.

---

### Bad Example

A viral Instagram reel.

One person shares it.

Thousands immediately arrive.

Events become dependent.

Poisson assumption breaks.

---

## Condition 3: No Simultaneous Events

Two events should not occur exactly at same instant.

Mathematically:

Probability of multiple events in an infinitesimally small interval should be negligible.

---

### Example

Website visitors:

In one microsecond:

Either

- no visitor
- one visitor

should happen.

Not hundreds simultaneously.

---

# Poisson Process

When all three assumptions hold:

1. Constant rate
2. Independent events
3. No simultaneous events

The underlying mechanism is called:

**Poisson Process**

---

Poisson Process generates:

$$
X \sim Poisson(\lambda)
$$

---

# Understanding λ (Lambda)

Lambda is the most important quantity.

---

## Meaning of λ

Average number of events per interval.

Examples:

### Website

Average visitors/hour:

$$
\lambda = 20
$$

---

### Football

Average goals/match:

$$
\lambda = 2.5
$$

---

### Call Center

Average calls/minute:

$$
\lambda = 8
$$

---

# Poisson Notation

$$
X \sim Poisson(\lambda)
$$

Read as:

"X follows a Poisson Distribution with parameter lambda."

---

# PMF of Poisson Distribution

The probability of observing exactly k events is:



where:

- k = 0,1,2,3,...
- λ = average rate
- e ≈ 2.718

---

# Understanding PMF Intuitively

Suppose:

Average goals/match:

$$
\lambda = 2.5
$$

Question:

Probability of exactly:

$$
2
$$

goals?

Use:

$$
P(X=2)
=
\frac{2.5^2 e^{-2.5}}{2!}
$$

Result:

$$
0.257
$$

Meaning:

About

$$
25.7\%
$$

of matches have exactly 2 goals.

---

# Football Example

Average goals:

$$
\lambda = 2.5
$$

---

Probability of:

### 0 Goals

$$
P(X=0)
=
e^{-2.5}
\approx 0.082
$$

8.2%

---

### 1 Goal

$$
P(X=1)
=
2.5e^{-2.5}
\approx 0.205
$$

20.5%

---

### 2 Goals

$$
P(X=2)
=
\frac{2.5^2 e^{-2.5}}{2}
\approx0.257
$$

25.7%

---

Notice:

The probabilities add up to 1.

---

# PMF Interpretation

PMF answers:

$$
P(X=k)
$$

Meaning:

Probability of observing exactly k events.

Examples:

$$
P(X=5)
$$

Exactly 5 visitors.

---

$$
P(X=2)
$$

Exactly 2 goals.

---

$$
P(X=0)
$$

Exactly 0 accidents.

---

# CDF of Poisson

CDF answers:

$$
P(X\le k)
$$

Meaning:

Probability of observing at most k events.

---

Example:

$$
P(X\le3)
$$

means

$$
P(X=0)+P(X=1)+P(X=2)+P(X=3)
$$

---

General Formula

$$
P(X\le k)
=
\sum_{i=0}^{k}
P(X=i)
$$

---

# Mean of Poisson

One beautiful property:

$$
E[X]
=
\lambda
$$

---

Example

Average visitors:

$$
\lambda = 20
$$

Then

$$
E[X]=20
$$

---

# Variance of Poisson

Another unique property:

$$
Var(X)
=
\lambda
$$

---

Example

$$
\lambda=20
$$

Then

$$
Var(X)=20
$$

---

# Standard Deviation

$$
\sigma
=
\sqrt{\lambda}
$$

---

Example

$$
\lambda=25
$$

Then

$$
\sigma=5
$$

---

# Why Mean = Variance?

Poisson is one of the few distributions where:

$$
Mean = Variance
$$

This is a signature property.

Whenever data has:

$$
Mean \approx Variance
$$

Poisson becomes a strong candidate.

---

# MGF of Poisson

Moment Generating Function:

$$
M_X(t)
=
E[e^{tX}]
$$

For Poisson:

$$
M_X(t)
=
\exp
\left(
\lambda(e^t-1)
\right)
$$

Used to derive:

- Mean
- Variance
- Higher moments

---

# Deriving Variance

Recall:

$$
Var(X)
=
E[X^2]
-
(E[X])^2
$$

Using MGF:

$$
E[X]=\lambda
$$

and

$$
E[X^2]
=
\lambda+\lambda^2
$$

Therefore:

$$
Var(X)
=
\lambda
$$

---

# Shape of Poisson Distribution

Depends entirely on λ.

---

## Small λ

Example:

$$
\lambda=1
$$

Most probability near:

$$
0,1,2
$$

Right-skewed.

---

## Medium λ

Example:

$$
\lambda=4
$$

More balanced.

---

## Large λ

Example:

$$
\lambda=20
$$

Looks bell-shaped.

Very close to Normal Distribution.

---

# Poisson → Normal Approximation

When λ becomes large:

$$
X \sim Poisson(\lambda)
$$

can be approximated by:

$$
X
\approx
N(\lambda,\lambda)
$$

because:

Mean:

$$
\lambda
$$

Variance:

$$
\lambda
$$

---

Example

$$
Poisson(100)
$$

approximately behaves like:

$$
N(100,100)
$$

---

# Relationship With Binomial Distribution

Poisson is a limiting case of Binomial.

---

Binomial:

$$
X\sim Bin(n,p)
$$

When:

$$
n \rightarrow \infty
$$

and

$$
p \rightarrow 0
$$

while

$$
np=\lambda
$$

remains constant,

then:

$$
Bin(n,p)
\rightarrow
Poisson(\lambda)
$$

---

# Example

Website has:

1,000,000 users.

Each user clicks with probability:

$$
0.00002
$$

Expected clicks:

$$
np
=
20
$$

Instead of Binomial:

$$
Bin(1000000,0.00002)
$$

we use:

$$
Poisson(20)
$$

Much easier.

---

# Relationship With Exponential Distribution

This is extremely important.

---

Poisson answers:

> How many events occur?

Exponential answers:

> How long until next event?

---

Example

Website:

Poisson:

$$
X=\text{Visitors next hour}
$$

---

Exponential:

$$
T=\text{Time until next visitor}
$$

---

Poisson → Counts

Exponential → Waiting Time

Both arise from the same Poisson Process.
