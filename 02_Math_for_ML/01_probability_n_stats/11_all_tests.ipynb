# Big Picture of Hypothesis Testing

Hypothesis testing is simply answering:

> Is the observed difference large enough that it is unlikely to have occurred by random chance?

or

> If the Null Hypothesis were true, how surprising is our data?

---

# General Framework

```text
1. Define H0 and Ha
2. Collect Sample
3. Compute Test Statistic
4. Find Distribution under H0
5. Compute p-value
6. Compare with α
7. Make Decision
```

---

# Visual Intuition

```text
Assume H0 is true
        ↓
What values of T are expected?
        ↓
Observe Tobs
        ↓
Is Tobs too far from expected values?
        ↓
Yes  → Reject H0
No   → Fail to Reject H0
```

---

# Step 1: Define Hypotheses

## Null Hypothesis

The default assumption.

Denoted by:

$$
H_0
$$

Examples:

$$
H_0:\mu=100
$$

$$
H_0:p=0.5
$$

$$
H_0:\mu_1=\mu_2
$$

---

## Alternative Hypothesis

The claim we are trying to find evidence for.

Denoted by:

$$
H_a
$$

Examples:

$$
H_a:\mu\neq100
$$

$$
H_a:p>0.5
$$

$$
H_a:\mu_1<\mu_2
$$

---

# Significance Level

Denoted by:

$$
\alpha
$$

Represents maximum probability of making a Type-I Error.

Common choices:

| α | Meaning |
|---|---|
| 0.10 | 10% |
| 0.05 | 5% |
| 0.01 | 1% |
| 0.001 | 0.1% |

---

# Type-I Error

Rejecting a true null hypothesis.

```text
Reality: H0 is True
Decision: Reject H0
```

Probability:

$$
P(\text{Type-I Error})
=
\alpha
$$

---

# Type-II Error

Failing to reject a false null hypothesis.

```text
Reality: H0 is False
Decision: Fail to Reject H0
```

Probability:

$$
\beta
$$

---

# Power of Test

Probability of detecting a true effect.

$$
Power
=
1-\beta
$$

Higher power is better.

---

# Test Statistic

A number computed from sample data.

General form:

$$
T
=
\frac{\text{Observed}-\text{Expected}}
{\text{Standard Error}}
$$

---

# Why Standardize?

Suppose:

```text
Difference = 10
```

Is 10 large?

Depends.

```text
If std = 1      → huge
If std = 100    → tiny
```

Therefore:

$$
T
=
\frac{\text{Difference}}
{\text{Variability}}
$$

---

# Sampling Distribution

A distribution of the statistic under:

$$
H_0
$$

All hypothesis tests are based on this idea.

## Hypotheses

$$
H_0:\mu=\mu_0
$$

$$
H_a:\mu<\mu_0
$$

---

## Visual

```text
██████|-----------|-----------
     Tcr         μ0
```

---

## p-value

$$
P(T\le T_{obs}|H_0)
$$

---

## Examples

H_a:\mu_{after}<\mu_{before}

# 3. Two-Tailed Test

Question:

```text
Is the value different?
```

---

## Hypotheses

$$
H_0:\mu=\mu_0
$$

$$
H_a:\mu\neq\mu_0
$$

---

## Visual

```text
█████------μ0------█████
```

---

## p-value

Both tails contribute.

$$
p
=
P(T\le -|T_{obs}|)
+
P(T\ge |T_{obs}|)
$$

or

$$
p
=
2P(T\ge |T_{obs}|)
$$

for symmetric distributions.

---

# Why Split α?

Suppose:

$$
\alpha=0.05
$$

Then:

$$
0.025
$$

on each side.

---

# Critical Region

The set of values for which we reject:

$$
H_0
$$

Also called:

```text
Rejection Region
```

---

# Acceptance Region

More accurately:

```text
Fail to Reject H0 Region
```

We never prove:

```text
H0 is true
```

We only say:

```text
Insufficient evidence against H0.
```

---

# Critical Value

Boundary between rejection and non-rejection region.

---

# Relationship Between α and Critical Value

Smaller α:

```text
More strict
Critical value moves away from center.
```

Larger α:

```text
Less strict
Critical value moves toward center.
```

---

# Common Z Critical Values

| α | Right Tail | Two Tail |
|---|---|---|
| 0.10 | 1.282 | ±1.645 |
| 0.05 | 1.645 | ±1.96 |
| 0.01 | 2.326 | ±2.576 |

---

# Method 1: Critical Value Approach

Right-tailed:

$$
T_{obs}>T_{cr}
$$

Left-tailed:

$$
T_{obs}<T_{cr}
$$

Two-tailed:

$$
|T_{obs}|>T_{cr}
$$

---

# Method 2: P-value Approach

Reject if:

$$
p<\alpha
$$

Fail to reject if:

$$
p\ge \alpha
$$

---

# Why Both Methods Agree

Critical value approach:

```text
Compare statistic with threshold.
```

p-value approach:

```text
Compare probability with threshold.
```

Both are mathematically identical.

# Z-Test

Used when:

```text
Population is Normal
Population standard deviation is known
OR
Sample size is large (n ≥ 30)
```

---

# One-Sample Z-Test

Question:

```text
Is sample mean different from population mean?
```

---

## Hypotheses

$$
H_0:\mu=\mu_0
$$

$$
H_a:\mu\neq\mu_0
$$

---

## Test Statistic

$$
Z=
\frac{\bar X-\mu_0}
{\sigma/\sqrt n}
$$

Under:

$$
H_0
$$

$$
Z\sim N(0,1)
$$

---

# Why Does Z Follow N(0,1)?

By Central Limit Theorem:

$$
\bar X
\sim
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

Standardizing:

$$
Z=
\frac{\bar X-\mu}
{\sigma/\sqrt n}
\sim N(0,1)
$$

---

# Example

Population:

$$
\mu_0=100
$$

Known:

$$
\sigma=20
$$

Sample:

$$
n=64,\quad \bar X=105
$$

Then:

$$
Z=
\frac{105-100}
{20/8}
=
2
$$

At:

$$
\alpha=0.05
$$

Critical value:

$$
1.96
$$

Since:

$$
2>1.96
$$

Reject:

$$
H_0
$$

---

# Two-Sample Z-Test

Question:

> Are two population means same?

---

## Input

```text
Sample1 → n1
Sample2 → n2
```

Sample means:

$$
\bar X_1,\bar X_2
$$

Population variances known:

$$
\sigma_1^2,\sigma_2^2
$$

---

## Hypotheses

$$
H_0:\mu_1=\mu_2
$$

$$
H_a:\mu_1\neq\mu_2
$$

---

## Test Statistic

$$
Z=
\frac{
\bar X_1-\bar X_2
}
{
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
}
$$

Under:

$$
H_0
$$

$$
Z\sim N(0,1)
$$

---

# Running Example (Medicine)

Medicine-1:

```text
n1=50
mean=72
σ1=10
```

Medicine-2:

```text
n2=40
mean=68
σ2=12
```

Compute:

$$
Z=
\frac{72-68}
{
\sqrt{
\frac{10^2}{50}
+
\frac{12^2}{40}
}
}
=
1.72
$$

At:

$$
\alpha=0.05
$$

Critical value:

$$
1.96
$$

Since:

$$
1.72<1.96
$$

Fail to reject:

$$
H_0
$$

$$

# One-Sample Proportion Test

Hypotheses:

$$
H_0:p=p_0
$$

---

Statistic:

$$
Z=
\frac{
\hat p-p_0
}
{
\sqrt{
\frac{p_0(1-p_0)}{n}
}
}
$$

---

# Example

100 tosses:

Heads:

$$
65
$$

Then:

$$
\hat p
=
0.65
$$

Test:

$$
H_0:p=0.5
$$

Statistic:

$$
Z=
\frac{0.65-0.5}
{\sqrt{\frac{0.5(0.5)}{100}}}
=
3
$$

Since:

$$
3>1.96
$$

Reject:

$$
H_0
$$

# Two-Proportion Z-Test

Question:

```text
Are two conversion rates equal?
```

---

Hypotheses:

$$
H_0:p_1=p_2
$$

---

Pooled proportion:

$$
\hat p
=
\frac{x_1+x_2}
{n_1+n_2}
$$

---

Statistic:

$$
Z=
\frac{
\hat p_1-\hat p_2
}
{
\sqrt{
\hat p(1-\hat p)
\left(
\frac1{n_1}
+
\frac1{n_2}
\right)
}
}
$$

# Final One-Line Summary

Hypothesis testing measures how surprising the observed data would be if the null hypothesis were true, and uses either a p-value or a critical value to decide whether that surprise is large enough to reject the null hypothesis.

# Two-Sample Z-Test: Complete Theory, Intuition, CLT Connection, and Workflow

---

# Big Picture

Suppose we have **two populations**:

Population 1:

$$
\mu_1,\sigma_1^2
$$

Population 2:

$$
\mu_2,\sigma_2^2
$$

We want to answer:

> Are the two population means equal?

or

> Is one population mean larger than the other?

Examples:

- Is the average salary of Software Engineers in Company A different from Company B?
- Is the average response time of API version 1 different from API version 2?
- Is a new medicine more effective than the old one?

---

# Assumptions of Two-Sample Z-Test

The Z-test for two means can be used when:

### Assumption 1

The two samples are independent.

```text
Sample 1 does not influence Sample 2.
```

---

### Assumption 2

Population variances are known:

$$
\sigma_1,\sigma_2
$$

In practice, this is rare.

If variances are unknown, we usually use a **t-test**.

---

### Assumption 3

At least one of the following is true:

1. Both populations are Normal

OR

2. Sample sizes are sufficiently large:

$$
n_1 \ge 30,\qquad n_2 \ge 30
$$

because of the Central Limit Theorem (CLT).

---

# Why CLT Makes Z-Test Possible

The CLT says:

If

$$
X_1,X_2,\dots,X_n
$$

are i.i.d. random variables with

$$
E[X]=\mu
$$

and

$$
Var(X)=\sigma^2
$$

then

$$
\bar X
=
\frac{X_1+\cdots+X_n}{n}
$$

has

$$
E[\bar X]=\mu
$$

and

$$
Var(\bar X)
=
\frac{\sigma^2}{n}
$$

Thus,

$$
\bar X
\approx
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

for large samples.

---

# Sampling Distribution

The distribution of:

$$
\bar X
$$

is called the

# Sampling Distribution of the Mean.

It has:

Mean:

$$
\mu
$$

Variance:

$$
\frac{\sigma^2}{n}
$$

Standard Error:

$$
SE
=
\frac{\sigma}{\sqrt n}
$$

---

# Why Does Variance Shrink?

Averaging reduces randomness.

```text
Large n
   ↓
Less variability
   ↓
More accurate estimate of μ
```

Since:

$$
Var(\bar X)
=
\frac{\sigma^2}{n}
$$

doubling the sample size cuts variability.

---

# Intuition

Individual observations may vary a lot.

But averages vary much less.

For example:

```text
Individual heights:
150, 160, 170, 180, 190

Sample means:
169, 171, 170, 168, 172
```

The sample means are much more stable.

---

# Distribution of Sample Means

Population:

```text
Can be skewed.
```

Repeated sampling:

```text
Sample means become increasingly Normal.
```

As

$$
n\to\infty
$$

the sampling distribution becomes:

$$
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

---

# Why n ≥ 30?

There is nothing magical about 30.

It is only a rule of thumb.

Some distributions need:

```text
n > 100
```

Some distributions work well with:

```text
n = 10
```

---

# Standardizing

Since

$$
\bar X
\sim
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

we standardize:

$$
Z
=
\frac{\bar X-\mu}
{\sigma/\sqrt n}
$$

Then:

$$
Z\sim N(0,1)
$$

This is the foundation of the **One-Sample Z-Test**.

---

# Moving to Two Samples

Suppose:

Sample 1:

$$
X_1,\dots,X_{n_1}
$$

Sample 2:

$$
Y_1,\dots,Y_{n_2}
$$

Sample means:

$$
\bar X_1,\qquad \bar X_2
$$

---

# By CLT

$$
\bar X_1
\sim
N
\left(
\mu_1,
\frac{\sigma_1^2}{n_1}
\right)
$$

and

$$
\bar X_2
\sim
N
\left(
\mu_2,
\frac{\sigma_2^2}{n_2}
\right)
$$

---

# What Do We Care About?

Not:

$$
\bar X_1
$$

Not:

$$
\bar X_2
$$

We care about:

$$
\bar X_1-\bar X_2
$$

because our question is:

```text
Are the two population means equal?
```

---

# Distribution of Difference of Means

Mean:

$$
E[\bar X_1-\bar X_2]
=
\mu_1-\mu_2
$$

---

Because the samples are independent:

$$
Var(X-Y)
=
Var(X)+Var(Y)
$$

Therefore,

$$
Var(\bar X_1-\bar X_2)
=
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
$$

---

# Standard Error of Difference

$$
SE
=
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
$$

---

# Sampling Distribution of Difference

Therefore,

$$
\bar X_1-\bar X_2
\sim
N
\left(
\mu_1-\mu_2,
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
\right)
$$

---

# Step 1: State the Hypotheses

## Two-Tailed Test

$$
H_0:\mu_1=\mu_2
$$

$$
H_a:\mu_1\neq\mu_2
$$

---

## Left-Tailed Test

$$
H_0:\mu_1=\mu_2
$$

$$
H_a:\mu_1<\mu_2
$$

---

## Right-Tailed Test

$$
H_0:\mu_1=\mu_2
$$

$$
H_a:\mu_1>\mu_2
$$

---

# Under the Null Hypothesis

Under

$$
H_0
$$

we assume:

$$
\mu_1-\mu_2=0
$$

Hence,

$$
\bar X_1-\bar X_2
\sim
N
\left(
0,
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
\right)
$$

---

# Step 2: Compute Test Statistic

Therefore,

$$
\boxed{
Z_{obs}
=
\frac{
\bar X_1-\bar X_2
}
{
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
}
}
$$

Under

$$
H_0
$$

$$
Z_{obs}\sim N(0,1)
$$

---

# Why Is Numerator NOT Divided by n?

Many students ask:

```text
Why not

(\bar X_1-\bar X_2)/n ?
```

Because:

```text
The quantity we are testing IS the difference itself.
```

The effect of sample size is already included in:

$$
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
$$

---

# Step 3: Compute P-value

The p-value is:

```text
Probability of obtaining a result at least as extreme
as the observed statistic assuming H0 is true.
```

---

## Left Tail

$$
p
=
P(Z\le Z_{obs})
$$

---

## Right Tail

$$
p
=
P(Z\ge Z_{obs})
$$

---

## Two Tail

$$
p
=
2P(Z\ge |Z_{obs}|)
$$

---

# Visual Interpretation

```text
Left Tail

|*****------------------|
```

```text
Right Tail

|------------------*****|
```

```text
Two Tail

|****------------****|
```

---

# Step 4: Compare with Significance Level

Choose:

$$
\alpha=0.05
$$

or

$$
\alpha=0.01
$$

Decision rule:

If

$$
p < \alpha
$$

Reject

$$
H_0
$$

Else:

Fail to Reject

$$
H_0
$$

---

# Equivalent Critical Value Method

For

$$
\alpha=0.05
$$

Two-tailed:

$$
z_{critical}
=
\pm1.96
$$

---

For

$$
\alpha=0.01
$$

Two-tailed:

$$
z_{critical}
=
\pm2.576
$$

---

For

$$
\alpha=0.001
$$

Two-tailed:

$$
z_{critical}
=
\pm3.29
$$

---

# Decision Rule

If

$$
|Z_{obs}|
>
z_{critical}
$$

Reject

$$
H_0
$$

Otherwise,

Fail to Reject

$$
H_0
$$

---

# Example

Suppose:

$$
\bar X_1=100
$$

$$
\bar X_2=90
$$

$$
n_1=40
$$

$$
n_2=50
$$

$$
\sigma_1=15
$$

$$
\sigma_2=20
$$

---

Compute:

$$
SE
=
\sqrt{
\frac{15^2}{40}
+
\frac{20^2}{50}
}
=
\sqrt{13.625}
=
3.69
$$

---

Thus,

$$
Z_{obs}
=
\frac{100-90}{3.69}
=
2.71
$$

---

At

$$
\alpha=0.05
$$

Critical value:

$$
1.96
$$

Since:

$$
2.71>1.96
$$

Reject

$$
H_0
$$

There is significant evidence that the means differ.

| Sample Mean | $\bar X$ |
| Sampling Variance | $\sigma^2/n$ |
| Standard Error | $\sigma/\sqrt n$ |
| Test Statistic | $(Statistic-Parameter)/(Std\ Error)$ |
| Significance Level | $\alpha$ |
| P-value | Probability under $H_0$ |
| Critical Value | Boundary for rejection |

---

# Complete Mental Model

```text
Population 1 -----> Sample -----> Mean X̄1
Population 2 -----> Sample -----> Mean X̄2
                             ↓
                      Difference X̄1−X̄2
                             ↓
                  Sampling Distribution
                             ↓
                     Approximately Normal
                             ↓
                      Standardize (Z)
                             ↓
                      Compute P-value
                             ↓
                  Compare with α
                             ↓
                 Reject or Fail to Reject H0
```

---

# One-Line Summary

The Two-Sample Z-Test works because the Central Limit Theorem makes the difference of two independent sample means approximately Normal, allowing us to convert the observed difference into a standard Normal statistic and compute probabilities under the null hypothesis.

# Sampling Distribution

```text
Sample 1 → Mean = 9.2
Sample 2 → Mean = 9.4
Sample 3 → Mean = 9.1
Sample 4 → Mean = 9.3
```

The distribution of these sample means is called:

# Sampling Distribution of Mean

# Central Limit Theorem (CLT)

If:

$$
n
$$

is sufficiently large,

then:

$$
\bar X
$$

becomes approximately Normal.

Even if the original data is skewed.

---

# CLT Assumptions

1. Samples are independent.

2. Sample size is sufficiently large.

3. Population has finite variance.

---

# Result

For large n:

$$
\bar X
\sim
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

---

# For Difference of Means

If:

$$
\bar X_1
\sim
N
\left(
\mu_1,
\frac{\sigma_1^2}{n_1}
\right)
$$

and

$$
\bar X_2
\sim
N
\left(
\mu_2,
\frac{\sigma_2^2}{n_2}
\right)
$$

then:

$$
\bar X_1-\bar X_2
\sim
N
\left(
\mu_1-\mu_2,
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
\right)
$$

---

# Standard Error (SE)

The variability of sample means is called:

# Standard Error

For two means:

$$
SE
=
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
$$

---

# Why Standardize?

Suppose:

Difference:

```text
0.2 hours
```

Is it large?

We don't know.

We need to compare it relative to its variability.

Therefore:

$$
Z
=
\frac{
Difference
}{
Standard Error
}
$$

---

# Z-Test Statistic

$$
Z=
\frac{
(\bar X_1-\bar X_2)-(\mu_1-\mu_2)
}{
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
}
$$

Under:

$$
H_0:\mu_1-\mu_2=0
$$

it becomes:

$$
Z=
\frac{
\bar X_1-\bar X_2
}{
\sqrt{
\frac{\sigma_1^2}{n_1}
+
\frac{\sigma_2^2}{n_2}
}
}
$$

---

# Z-Test Assumptions

1. Samples are independent.

2. Population variance known.

3. Population normal OR sample size large.

4. Data is quantitative.

Because:

$$

and

$$
S_2 \approx \sigma_2
$$

# Practical Rule

If:

$$

$$

$$

# Two Sample Z-Test
