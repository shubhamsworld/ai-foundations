# Mode

## Definition

The **Mode** is the value that occurs most frequently in a dataset.

In simple words:

> Mode = Most frequently occurring value

---

# What Does Mode Tell Us?

Mode tells us:

> Which value or category appears most often.

---


# Real-World Use Case: Popular Shoe Size

A shoe company wants to know:

> Which shoe size should be manufactured the most?

Suppose customer shoe sizes are:

$$
[7,\ 8,\ 8,\ 9,\ 8,\ 10,\ 7,\ 8]
$$

Frequency Table:

| Shoe Size | Count |
|------------|--------|
| 7 | 2 |
| 8 | 4 |
| 9 | 1 |
| 10 | 1 |

The most frequently purchased size is:

$$
\text{Mode} = 8
$$

Therefore, the company may decide to:

- Manufacture more size 8 shoes
- Maintain larger inventory for size 8
- Reduce the chance of stock shortages

This is a classic example where **Mode helps identify the most popular or most demanded category/value**.

# Weighted Mean (Weighted Average)

## Why Do We Need Weighted Mean?

The normal mean assumes that every observation is equally important.

$$
\bar{x}=\frac{x_1+x_2+\cdots+x_n}{n}
$$

However, in many real-world situations, some observations are more important than others.

In such cases, we use a **Weighted Mean**.

---

# Definition

A Weighted Mean assigns a weight (importance) to each observation.

$$
\bar{x}_w
=
\frac{\sum_{i=1}^{n} w_i x_i}
{\sum_{i=1}^{n} w_i}
$$

where:

- $x_i$ = observation/value
- $w_i$ = weight assigned to that value

---

# Intuition

Think of weights as:

> "How much importance should be given to each value?"

Larger weight ⇒ greater influence on the final average.

Smaller weight ⇒ less influence on the final average.

---
# Real-World Use Case: Weighted Mean in Business

Suppose a company wants to estimate the average monthly spending on wine.

However, recent purchases are more important than older purchases because customer behavior changes over time.

Therefore, we assign higher weights to recent months.

| Month | Wine Sales ($) | Weight |
|---------|---------|---------|
| Jan | 100 | 1 |
| Feb | 120 | 2 |
| Mar | 150 | 3 |
| Apr | 180 | 4 |

Notice that the latest month (Apr) receives the highest weight.

The weighted mean is:

$$
\bar{x}_w
=
\frac{\sum_{i=1}^{n} w_i x_i}
{\sum_{i=1}^{n} w_i}
$$

Substituting values:

$$
\bar{x}_w
=
\frac{(1 \times 100)+(2 \times 120)+(3 \times 150)+(4 \times 180)}
{1+2+3+4}
$$

$$
=
\frac{1510}{10}
=
151
$$

---

## Why Not Use Simple Mean?

Simple Mean:

$$
\frac{100+120+150+180}{4}
=
137.5
$$

This treats all months equally.

Weighted Mean:

$$
151
$$

This gives more importance to recent sales data.

Therefore, weighted mean often provides a better estimate when recent observations are more relevant than older observations.

---

## Common Applications of Weighted Mean

- Recent product sales are given more importance than old sales.
- Recent stock prices are given more importance in financial analysis.
- GPA calculation where courses have different credit hours.
- Customer ratings where verified purchases may receive higher importance.

---

## One-Line Summary

> Weighted Mean is used when some observations are more important than others and should contribute more to the final average.
---

# Variance and Standard Deviation

## Why Do We Need Them?

The mean tells us the center of the data.

However, two datasets can have the same mean but very different spreads.

### Dataset A

$$
[8,\ 10,\ 12]
$$

### Dataset B

$$
[0,\ 10,\ 20]
$$

Both datasets have:

$$
\bar{x} = 10
$$

Yet Dataset B is much more spread out than Dataset A.

Therefore, besides measuring the center of the data, we also need a way to measure how dispersed or spread out the observations are.

This is where **Variance** and **Standard Deviation** are used.

---

# Variance

## Definition

Variance measures the average squared distance of observations from the mean.

### Population Variance

$$
\sigma^2
=
\frac{1}{n}
\sum_{i=1}^{n}
(x_i-\bar{x})^2
$$

where:

- $x_i$ = observation
- $\bar{x}$ = mean
- $n$ = total number of observations

---

## Intuition

Variance answers the question:

> How far, on average, are the data points from the mean?

- Small variance → observations are clustered near the mean.
- Large variance → observations are spread far from the mean.

---

## Why Do We Square the Differences?

Suppose we calculate:

$$
x_i - \bar{x}
$$

Positive and negative deviations cancel each other.

Example:

$$
[-2,\ 0,\ 2]
$$

Sum:

$$
-2 + 0 + 2 = 0
$$

Even though the data is spread out, the total becomes zero.

To prevent this cancellation, we square each deviation:

$$
(x_i-\bar{x})^2
$$

This ensures:

- All values are positive
- Larger deviations receive a larger penalty

---

## Example

Dataset:

$$
[8,\ 10,\ 12]
$$

Mean:

$$
\bar{x}=10
$$

Deviations:

$$
[-2,\ 0,\ 2]
$$

Squared Deviations:

$$
[4,\ 0,\ 4]
$$

Variance:

$$
\sigma^2
=
\frac{4+0+4}{3}
=
\frac{8}{3}
=
2.67
$$

---

# Standard Deviation

## Definition

Standard Deviation is the square root of Variance.

$$
\sigma
=
\sqrt{\sigma^2}
$$

---

## Why Do We Need Standard Deviation?

Variance is measured in squared units.

For example:

- Height → cm
- Variance → cm²

Squared units are difficult to interpret.

Taking the square root gives us a measure in the original unit.

Therefore Standard Deviation is usually easier to understand.

---

## Example

Dataset:

$$
[0,\ 10,\ 20]
$$

Mean:

$$
\bar{x}=10
$$

Deviations:

$$
[-10,\ 0,\ 10]
$$

Squared Deviations:

$$
[100,\ 0,\ 100]
$$

Variance:

$$
\sigma^2
=
\frac{100+0+100}{3}
=
66.67
$$

Standard Deviation:

$$
\sigma
=
\sqrt{66.67}
\approx 8.16
$$

---

# Mean-Based Measure of Spread

Notice that Variance and Standard Deviation use:

$$
x_i-\bar{x}
$$

Therefore, the mean acts as the reference point.

Both measures describe how observations vary around the mean.

---

# Mean-Based vs Median-Based Spread

There are two major approaches for measuring spread.

## 1. Mean-Based Measures

Examples:

- Variance
- Standard Deviation

Uses:

$$
x_i-\bar{x}
$$

Advantages:

- Smooth mathematical formulation
- Differentiable
- Easy to optimize
- Widely used in Machine Learning

Disadvantage:

- Sensitive to outliers

---

## 2. Median-Based Measures

Example:

### Median Absolute Deviation (MAD)

$$
MAD
=
\text{Median}
\left(
|x_i-\text{Median}|
\right)
$$

Advantages:

- Robust to outliers
- Works well for skewed distributions

Disadvantage:

- Harder to optimize mathematically

---

# Why Machine Learning Often Uses Mean-Based Measures

Most Machine Learning algorithms use optimization methods such as:

- Gradient Descent
- Stochastic Gradient Descent (SGD)

These methods rely on derivatives.

Mean-based measures lead to smooth and differentiable objective functions.

Because of this, many ML loss functions are built around:

- Mean
- Variance
- Squared Error (MSE)

Examples:

$$
MSE
=
\frac{1}{n}
\sum_{i=1}^{n}
(y_i-\hat{y}_i)^2
$$

---

# Important Observation

Mean-based methods:

- Sensitive to outliers
- Easy to optimize
- Common in ML

Median-based methods:

- Robust to outliers
- Less sensitive to extreme values
- Common in Robust Statistics

---

# Key Takeaways

- Mean measures the center of data.
- Variance measures spread around the mean.
- Standard Deviation is the square root of Variance.
- Variance uses squared deviations from the mean.
- Standard Deviation is easier to interpret because it has the same unit as the data.
- Mean-based measures are heavily used in Machine Learning because they are differentiable.
- Median-based measures are preferred when strong outliers are present.

````python
#std-dev
print("Gold:", df['MntGoldProds'].std())
print("Fruits:", df['MntFruits'].std())
print("Sweets:", df['MntSweetProducts'].std())
print("Wine:", df['MntWines'].std())
print("Meat:", df['MntMeatProducts'].std())
print("Fish:", df['MntFishProducts'].std())
````

````python
raw_mad = stats.median_abs_deviation(df["MntFruits"])

scaled_mad = stats.median_abs_deviation(
    df["MntFruits"],
    scale="normal"
)

print(raw_mad)
print(scaled_mad)
````

Why Scaled MAD?

Raw MAD measures dispersion around the median.

Scaled MAD provides an estimate of spread that is comparable to standard deviation when the data follows a normal distribution.

Comparison
Measure	Value
Raw MAD	8.0
Scaled MAD	11.8608
Which One Should We Use?
Raw MAD → when you want the true median-based spread.
Scaled MAD → when you want a robust alternative to standard deviation.
A useful memory trick:

```text
Raw MAD     -> Robust spread around Median
Scaled MAD  -> Robust approximation of Standard Deviation

# Percentiles

## What is a Percentile?

A percentile tells us the value below which a given percentage of observations lie.

In simple words:

> The \(k^{th}\) percentile is the value below which atleast \(k\%\) of the data falls.

---

## Example

Consider the dataset:

$$
[3,\ 2,\ 1,\ 5,\ 6]
$$

First, sort the data:

$$
[1,\ 2,\ 3,\ 5,\ 6]
$$

Number of observations:

$$
n = 5
$$

---

## Finding the 10th Percentile

We want to find:

$$
P_{10}
$$

### Step 1: Compute the Position

A simple percentile position formula is:

$$
\text{Position}
=
\frac{k}{100}\times n
$$

Substituting \(k=10\) and \(n=5\):

$$
\text{Position}
=
\frac{10}{100}\times 5
=
0.5
$$

The position lies near the first value in the sorted dataset.

Therefore:

$$
P_{10} \approx 1
$$

---

## Interpretation

$$
P_{10} \approx 1
$$

means:

> Approximately 10% of observations are less than or equal to 1.

---

## Another Example

Suppose exam marks are sorted as:

$$
[20,\ 35,\ 40,\ 50,\ 60,\ 75,\ 80,\ 90,\ 95,\ 100]
$$

Then:

$$
P_{50}
$$

is the median.

This means:

> 50% of students scored below the median and 50% scored above it.

---

# Median as a Percentile

The median is simply the 50th percentile.

$$
\text{Median} = P_{50}
$$

Visual view:

```

# Interquartile Range (IQR) and Outlier Detection

## Why Do We Need IQR?

Mean and standard deviation are sensitive to extreme values (outliers).

To measure spread more robustly, we often use quartiles and the Interquartile Range (IQR).

IQR focuses only on the middle 50% of the data and ignores extreme values.

---

# Quartiles Recap

Quartiles divide the sorted dataset into four equal parts.

| Quartile | Percentile | Meaning |
|-----------|------------|----------|
| Q1 | P25 | 25% of observations lie below it |
| Q2 | P50 | Median |
| Q3 | P75 | 75% of observations lie below it |
| Q4 | P100 | Maximum value |

Visual representation:

```text
Q0       Q1       Q2       Q3       Q4
|--------|--------|--------|--------|
Min      25%     50%      75%      Max
                  ^
               Median
```

# Example

                     ^
                    Q2

Assume:

Q_1 = 40,000
$$

$$
Q_2 = P_{50} = \text{Median}
$$

$$
Q_3 = 150,000

# Finding the Position of a Percentile

# What is IQR?

Interquartile Range (IQR) measures the spread of the middle 50% of the data.

Definition:

$$
IQR = Q_3 - Q_1
$$

For the example above:

$$
IQR = 150,000 - 40,000
$$

$$
IQR = 110,000
$$

---

# Interpretation

The middle 50% of incomes are spread across:

$$
110,000
$$

units.

Since IQR ignores the lowest 25% and highest 25% of observations, it is much less affected by outliers than variance or standard deviation.

---

# Using IQR to Detect Outliers

A common rule for detecting outliers is the Tukey Rule (also called the 1.5 × IQR Rule).

---

## Lower Bound

$$
\text{Lower Bound}
=
Q_1 - 1.5 \times IQR
$$

Substituting values:

$$
40,000 - 1.5 \times 110,000
$$

$$
40,000 - 165,000
$$

$$
-125,000
$$

Since income cannot be negative, there are effectively no lower outliers in this example.

---

## Upper Bound

$$
\text{Upper Bound}
=
Q_3 + 1.5 \times IQR
$$

Substituting values:

$$
150,000 + 1.5 \times 110,000
$$

$$
150,000 + 165,000
$$

$$
315,000
$$

---

# Outlier Rule

Any observation satisfying:

$$
x < Q_1 - 1.5(IQR)
$$

or

$$
x > Q_3 + 1.5(IQR)
$$

is considered a potential outlier.

For our example:

$$
x > 315,000
$$

would be flagged as an outlier.

---

# Why Does This Work?

Most normal observations lie inside the interval:

$$
[Q_1 - 1.5(IQR),\; Q_3 + 1.5(IQR)]
$$

Values far outside this range are unusually distant from the majority of the data.

---

# Important Note

The 1.5 × IQR rule is a guideline, not a law.

Whether something is truly an outlier depends on:

- Domain knowledge
- Business context
- Data collection process

Example:

- A salary of ₹5,00,00,000 may be an outlier in a general employee dataset.
- The same salary may be completely normal in a CEO dataset.

Therefore:

> Outlier detection is partly statistical and partly domain-specific.

---

# Why IQR is Popular

Advantages:

- Robust to extreme values
- Easy to compute
- Works well on skewed data
- Commonly used in Exploratory Data Analysis (EDA)

---

# IQR vs Standard Deviation

| IQR | Standard Deviation |
|-------|------------------|
| Based on quartiles | Based on mean |
| Robust to outliers | Sensitive to outliers |
| Uses middle 50% of data | Uses all observations |
| Good for skewed distributions | Good for approximately normal distributions |

---

# One-Line Summary

- Quartiles divide data into four equal parts.
- \(Q_1=P_{25}\), \(Q_2=P_{50}\), \(Q_3=P_{75}\).
- Interquartile Range measures the spread of the middle 50% of data.

$$
IQR = Q_3 - Q_1
$$

- A common outlier rule is:

$$
Q_1 - 1.5(IQR)
\leq x \leq
Q_3 + 1.5(IQR)
$$

- Values outside this range are potential outliers.

````python
#outlier using IQR Range
r = 1.5 * stats.iqr(df['MntWines'])
lb = np.percentile(df['MntWines'], 25)-r
ub = np.percentile(df['MntWines'], 75)+r
print(lb)
print(ub)
````

````python
print(df['MntWines'].max())
````

````python
sum(df['MntWines']>ub)
````

# Histograms

## What is a Histogram?

A histogram is a graphical representation of the distribution of numerical data.

It shows:

- How data is distributed
- Where most observations lie
- Whether the data is symmetric or skewed
- Presence of outliers

---

## Histogram Axes

- X-axis → Values of the variable
- Y-axis → Frequency (count of observations)

Example:

- Wine spending
- Salary
- Age
- Height

The height of a bar indicates how many observations fall within a given range.

---

# Bins

A histogram divides the data range into intervals called **bins**.

Example:

Suppose:

$$
\text{Minimum Value} = 0
$$

$$
\text{Maximum Value} = 1500
$$

and we choose:

$$
30 \text{ bins}
$$

Then:

$$
\text{Bin Width}
=
\frac{\text{Max} - \text{Min}}{\text{Number of Bins}}
$$

$$
=
\frac{1500-0}{30}
$$

$$
= 50
$$

Therefore:

- Number of bins = 30
- Width of each bin = 50

---

# Bin Width Formula

Given:

- Minimum value = min
- Maximum value = max
- Number of bins = b

Bin width:

$$
\text{Bin Width}
=
\frac{\text{max}-\text{min}}{b}
$$

---

# Effect of Number of Bins

### Few Bins

```python
df["MntWines"].plot.hist(bins=10)
```

- Wider bars
- Less detail
- Smoother appearance

---

### Many Bins

```python
df["MntWines"].plot.hist(bins=100)
```

- Narrow bars
- More detail
- Reveals local patterns

---

## Important Observation

As the number of bins increases:

- Bin width decreases
- The histogram becomes more detailed
- The plot becomes finer

However:

Too many bins can introduce noise.

Too few bins can hide important patterns.

---

# Smoothed Histogram

A histogram is made of rectangular bars.

We can create a smoother representation by:

1. Taking the center of each bar
2. Connecting neighboring centers
3. Drawing a smooth curve

This produces a **smoothed histogram**.

The goal is to visualize the underlying distribution more clearly.

# Kernel Density Estimation (KDE)

Kernel Density Estimation (KDE) is a smooth approximation of the data distribution.

It is often viewed as a continuous version of a histogram.

Instead of counting observations inside bins:

- A smooth kernel is placed around each observation.
- All kernels are added together.
- The resulting curve estimates the probability density.

---

## Why Use KDE?

Advantages:

- Smooth visualization
- No abrupt bar boundaries
- Easier to understand overall distribution shape

---

## Gaussian Kernel

The most commonly used kernel is the Gaussian kernel.

It creates smooth bell-shaped bumps around observations.

Later, while studying Probability Density Functions (PDFs), we will understand KDE more formally.

For now:

> KDE is simply a smooth estimate of the underlying distribution.

---

# Density Plot

A density plot displays a probability density curve.

Unlike histograms:

- Y-axis is density
- Not raw frequency counts

---

## Important Property

The total area under a density curve is always:

$$
1
$$

or

$$
100\%
$$

This represents the entire probability mass.

---

## Area Interpretation

Suppose the shaded region between:

$$
1000 \le x \le 1200
$$

has area:

$$
0.2
$$

This means:

Approximately 20% of observations lie between 1000 and 1200.

---

## General Interpretation

If a region under the density curve has area:

$$
0.7
$$

then:

Approximately 70% of observations lie in that interval.

---

# Area Under the Curve (AUC)

For any density plot:

$$
\text{Total Area Under Curve}
=
1
$$

This is a fundamental property of probability density functions.

---

# Shape of Distributions

Density plots help us identify the shape of the data.

Common shapes include:

- Symmetric
- Right-skewed
- Left-skewed

---

# Right-Skewed Distribution

Characteristics:

- Long tail on the right
- Most observations concentrated on the left
- A few unusually large values

Visual intuition:

```text
Peak
 ^
 |\
 | \
 |  \
 |   \______
 +------------>
```

Examples:

- Income
- House prices
- Customer spending

---

# Left-Skewed Distribution

Characteristics:

- Long tail on the left
- Most observations concentrated on the right
- A few unusually small values

Visual intuition:

```text
        Peak
          /|
         / |
 _______/  |
           |
 +------------>
```

Examples:

- Easy exam scores
- Product ratings

---

# Symmetric Distribution

Characteristics:

- Left and right sides are approximately mirror images
- Centered around the mean

Visual intuition:

```text
      /\
     /  \
    /    \
   /      \
  /        \
 +------------>
```

A normal distribution is a common example.
