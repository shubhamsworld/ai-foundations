Python's:

```text
Continuous Uniform Distribution
between 0 and 1
```

# Uniform Sampling Idea

Suppose we have:

```text
150 rows
```

and want:

```text
30 rows
```

Let:

$$
n=150
$$

$$
m=30
$$

Desired sampling probability:

$$
p=\frac{m}{n}
$$

Therefore:

$$
p=\frac{30}{150}
$$

$$
p=0.2
$$

---

# Interpretation of p = 0.2

Each observation should have:

```text
20% chance
```

of being selected.

This is equivalent to saying:

```text
Select each point independently
with probability 0.2
```

---

# Example Dataset

Suppose:

```python
from sklearn import datasets

iris = datasets.load_iris()
d = iris.data

d.shape
```

Output:

```python
(150,4)
```

Meaning:

```text
150 rows
4 features
```

---

# Sampling Algorithm

We generate a random number:

$$
r \sim U(0,1)
$$

for every data point.

If:

$$
r < p
$$

keep the point.

Otherwise:

$$
r \ge p
$$

discard the point.

---

# Code

```python
import random

n = 150
m = 30

p = m/n

sampled_data = []

for i in range(n):

    if random.random() < p:
        sampled_data.append(d[i,:])
```

For every observation:

```python
random.random()
```

returns:

```text
0.13
0.91
0.45
0.02
0.73
...
```

Each value comes from:

$$
U(0,1)
$$

Because:

$$

for

$$

$$

The area from:

```text
0 to 0.2
```

is:

$$
0.2
$$

Thus every row has:

```text
20% probability
```

of entering the sample.

# Important Concept: Bernoulli Trial

Each row undergoes:

```text
Select?
Yes or No
```

This is a Bernoulli Random Variable.

Let:

$$
X_i=
\begin{cases}
1 & \text{selected}\\
0 & \text{not selected}
\end{cases}
$$

Then:

$$
X_i \sim Bernoulli(p)
$$

---

# Probability of Selection

$$
P(X_i=1)=p
$$

$$
P(X_i=0)=1-p
$$

---

# Total Number Selected

Total selected:

$$
S
=
\sum_{i=1}^{n} X_i
$$

Since each:

$$
X_i \sim Bernoulli(p)
$$

Then:

$$
S \sim Binomial(n,p)
$$

---

# Example

$$
n=150
$$

$$
p=0.2
$$

Then:

$$
S\sim Binomial(150,0.2)
$$

Expected value:

$$
E[S]=30
$$

Variance:

$$
Var(S)=np(1-p)
$$

$$
=
150(0.2)(0.8)
=
24
$$

---

# Connection to Binomial Distribution

Sampling each row independently:

```text
Select / Not Select
```

creates:

```text
Bernoulli trials
```

Many Bernoulli trials together form:

```text
Binomial Distribution
```

---

# Uniform Distribution → Bernoulli Distribution

This is a very important connection.

We start with:

$$
U\sim U(0,1)
$$

Apply threshold:

$$
U<p
$$

Then:

$$
X=
\begin{cases}
1 & U<p\\
0 & U\ge p
\end{cases}
$$

Now:

$$
X\sim Bernoulli(p)
$$

---

# Uniform Distribution → Sampling

Pipeline:

```text
Uniform(0,1)
       ↓
Threshold p
       ↓
Bernoulli(p)
       ↓
Sampling Decision
       ↓
Random Sample
```

This is one of the most important ideas in machine learning.

# Final Mental Model

Imagine every row in a dataset receives a random score:

```text
0.13
0.82
0.45
0.02
0.91
...
```

generated from:

$$
U(0,1)
$$

If the score is below some threshold:

```text
p = desired probability
```

we keep the row.

Thus:

```text
Uniform Distribution
      ↓
Random Numbers
      ↓
Bernoulli Decisions
      ↓
Random Sampling
      ↓
Modern Statistics & Machine Learning
```

**One-line summary:**

Almost all random sampling algorithms begin by generating values from the Continuous Uniform Distribution \(U(0,1)\). These values are converted into selection decisions, which ultimately power train/test splits, simulations, randomized algorithms, and most of modern machine learning.
