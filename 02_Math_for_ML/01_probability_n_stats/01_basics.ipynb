# Probability & Statistics 

## What is Probability?
Probability is the likelihood of an event occurring.

- Probability measures uncertainty.
- Higher probability ⇒ more likely.
- Lower probability ⇒ less likely.


## Experiment

An **experiment** is a process that produces one outcome from a well-defined set of possible outcomes.

### Example: Tossing a Coin
- Experiment: Toss a coin
- Outcomes: {H, T}

### Example: Administering Medicine
- Experiment: Give medicine to a patient
- Possible outcomes:
  - Recovery in 1–7 days
  - Recovery in 8–14 days
  - Recovery after 15+ days
  - Death


## Sample Space (S)

A **Sample Space** is the set of all possible outcomes of an experiment.

### Coin Toss
S = {H, T}

### Medicine Example
S = {Recovery 1–7 days, Recovery 8–14 days, Recovery 15+ days, Death}

### Important Properties of Sets
- Sets are unordered.
- {a, b} = {b, a}
- Order does not matter inside a set.

## Event (E)

An **event** is a subset of the sample space.

E ⊆ S

### Medicine Example
S = {R1-7, R8-14, R15+, D}

Event E₁ = Patient Recovered

E₁ = {R1-7, R8-14, R15+}

Event E₂ = Recovered within 14 days

E₂ = {R1-7, R8-14}


## Fair Coin

A coin is called **fair** if heads and tails are equally likely.

P(H) = 0.5
P(T) = 0.5


## Empirical Interpretation of Probability

If a coin is tossed many times:

| Number of Tosses (n) | Number of Heads | Fraction |
|----------------------|-----------------|----------|
| 10 | 6 | 6/10 = 0.6 |
| 20 | 11 | 11/20 = 0.55 |
| 100 | 51 | 51/100 = 0.51 |

As the number of tosses increases, the observed fraction approaches the true probability.

## Probability of Heads

For a large number of trials:

P(H) = (Number of times Head occurs) / n

as n → ∞

For a fair coin:
P(H) → 0.5

## Range of Probability

For any event E:

0 ≤ P(E) ≤ 1

- P(E) = 0 ⇒ Impossible Event
- P(E) = 1 ⇒ Certain Event
- 0 < P(E) < 1 ⇒ Event may or may not occur

## Key Takeaways

1. Probability measures the likelihood of an event.
2. Every experiment has a sample space.
3. Events are subsets of the sample space.
4. Sets are unordered.
5. A fair coin has equal chance of heads and tails.
6. Relative frequency approaches probability as trials increase.
7. Probability always lies between 0 and 1.







# Empirical Probability using Simulation

## Important Observation

No sample size is ever sufficient to prove with 100% certainty that a coin or dice is fair.

We can only estimate probabilities by conducting more and more trials.

As the number of trials increases, the observed probability approaches the true probability.

This idea is called the **Frequentist Interpretation of Probability**.

# Probability Theory - Fair Dice Example

## Fair Dice

A **fair dice** is a dice in which all outcomes are equally likely.

### Experiment
Throw a fair dice once.

### Sample Space

The set of all possible outcomes is:

S = {1, 2, 3, 4, 5, 6}

### Event

Let,

E = {6}

Event E represents the outcome of observing a 6.

Since every event must belong to the sample space:

E ⊆ S

### Probability of Observing a 6

Since a fair dice has 6 equally likely outcomes:

P(Observing 6) = 1/6

Similarly,

P(Observing 1) = 1/6

P(Observing 2) = 1/6

P(Observing 3) = 1/6

P(Observing 4) = 1/6

P(Observing 5) = 1/6

P(Observing 6) = 1/6

### Key Idea

For a fair dice:

Probability of any specific face

= Favorable Outcomes / Total Outcomes

= 1 / 6

````python
#Sample size is never sufficient and we can not be ever 100% sure/certain that coin/dice  is fair

# Fair-Dice throws
import random

# throw dice i times
for i in [5, 10, 100, 10000, 1000000]:
    wins = 0
    throws = int(i) #n

    for j in range(throws):
        #Random number Generator
        wins+= int(random.randint(1,6)==6)
        #random.randint(a,b): Returns a random integer N Such that a<= N <= b
    

    print("Number of Trails: {}, P(win): {}" .format(throws, wins/throws))
````

# Interpretation

True Probability:

P(6) = 1/6

≈ 0.166667

Notice:

- With 5 trials, estimate is very inaccurate.
- With 10 trials, estimate is still unstable.
- With 100 trials, estimate becomes better.
- With 10,000 trials, estimate gets very close.
- With 1,000,000 trials, estimate is extremely close to 1/6.

As:

n → ∞

Observed Frequency → True Probability

This is known as the **Law of Large Numbers**.

# Probability Example: Drawing Marbles from a Bag

## Problem Statement

A bag contains:

- 3 Red Marbles (R)
- 2 Blue Marbles (B)

Total Marbles = 5

The experiment is:

Draw one marble randomly and then put it back into the bag (**with replacement**).

This process is repeated many times.

---

## Sample Space

S = {R, R, R, B, B}

or simply

S = {Red, Blue}

---

## Probability of Drawing a Blue Marble

Number of Blue Marbles = 2

Total Marbles = 5

Therefore,

P(Blue) = Number of Blue Marbles / Total Number of Marbles

= 2/5

= 0.4

---

## Probability of Drawing a Red Marble

Number of Red Marbles = 3

Total Marbles = 5

P(Red) = 3/5

= 0.6

---

## Important Observation

Since the marble is placed back into the bag after every draw:

The bag composition never changes.

Every draw is independent.

Therefore:

P(Blue) always remains 2/5.

P(Red) always remains 3/5.

This is called **sampling with replacement**.

# Conditional Probability

## Intuition

Conditional Probability answers:

> What is the probability of Event A happening if we already know Event B has happened?

Instead of looking at the entire sample space, we restrict ourselves only to the outcomes where B has occurred.

---

## Notation

P(A | B)

Read as:

"Probability of A given B"

or

"Probability of A conditioned on B"

---

## Medicine Example

Let:

M₁ = Patient received Medicine M₁

S = Patient survived

Universe = 1000 patients

---

### Given Data

Patients receiving M₁ = 100

Patients surviving = 700

Patients receiving M₁ and surviving = 95

---

## Probability of Survival Given M₁

Question:

What is the probability that a patient survives given that the patient received M₁?

Notation:

P(S | M₁)

Since we already know the patient belongs to M₁,

our new sample space becomes only those 100 patients.

Out of those 100 patients:

95 survived.

Therefore,

P(S | M₁)

= 95 / 100

= 0.95

= 95%

---

## Interpretation

P(S | M₁) = 0.95

means

"If a patient has already received Medicine M₁, then the probability of survival is 95%."

Notice:

We are no longer dividing by 1000.

We divide only by the conditioned event.


# Definition of Conditional Probability

For any two events A and B,

provided:

P(B) ≠ 0

Conditional Probability is defined as:

P(A|B)
=
P(A ∩ B) / P(B)

---

## Important Condition

The denominator cannot be zero.

Therefore:

P(B) > 0

must hold.

Otherwise conditional probability is undefined.

---

## Meaning of Formula

Numerator:

P(A ∩ B)

Probability that both A and B occur.

Denominator:

P(B)

Probability of the condition event.

---

## Easy Way to Remember

Conditioned Event

↓

Goes in the denominator.

Example:

P(S | M₁)

=

P(S ∩ M₁)/
P(M₁)

# Reverse Conditional Probability

Conditional probability works in both directions.

P(S | M₁)

and

P(M₁ | S)

are NOT the same.

---

## Example

P(S | M₁)

=

Probability patient survives
given medicine M₁ was administered.

=

95 / 100

=

0.95

---

P(M₁ | S)

=

Probability patient received M₁
given patient survived.

Here we condition on survivors.

Total survivors = 700

Among survivors who received M₁ = 95

Therefore:

P(M₁ | S)

=

95 / 700

≈ 0.136

≈ 13.6%

---

## Key Lesson

Changing the condition changes the denominator.

P(A|B)

≠

P(B|A)

in general.

# Conditional Probability

## Definition

Conditional Probability answers:

> What is the probability of event A occurring given that event B has already occurred?

Instead of considering the entire sample space, we restrict ourselves to the outcomes where event B has occurred.

---

## Notation

$$
P(A \mid B)
$$

Read as:

**"Probability of A given B"**

---

## Conditional Probability Formula

For any two events A and B,

$$
P(A \mid B)
=
\frac{P(A \cap B)}{P(B)}
$$

provided that

$$
P(B) \neq 0
$$

---

## Similarly

$$
P(B \mid A)
=
\frac{P(A \cap B)}{P(A)}
$$

provided that

$$
P(A) \neq 0
$$

---

## Important Observation

In general,

$$
P(A \mid B)
\neq
P(B \mid A)
$$

Changing the condition changes the denominator.

# Example: Two Fair Dice

Consider two fair dice:

- $D_1$
- $D_2$

rolled simultaneously.

---

## Define Events

Let:

$$
E = \{\text{Sum of the two dice is } 7\}
$$

Let:

$$
F = \{D_1 = 3\}
$$

---

## Problem

Find:

$$
P(E \mid F)
$$

That is,

> What is the probability that the sum of the two dice equals 7, given that the first die has already shown 3?

---

# Solving the Dice Example

## Step 1: Find $P(F)$

Event $F$ means:

$$
D_1 = 3
$$

Possible outcomes satisfying $F$:

$$
(3,1), (3,2), (3,3), (3,4), (3,5), (3,6)
$$

Total favorable outcomes:

$$
6
$$

Since there are:

$$
6 \times 6 = 36
$$

total equally likely outcomes,

$$
P(F)=\frac{6}{36}
$$

Therefore,

$$
P(F)=\frac{1}{6}
$$

---

## Step 2: Find $P(E \cap F)$

Event $E$:

$$
D_1 + D_2 = 7
$$

Event $F$:

$$
D_1 = 3
$$

Both events occur together only when:

$$
(3,4)
$$

Thus,

$$
|E \cap F| = 1
$$

Therefore,

$$
P(E \cap F)=\frac{1}{36}
$$

---

# Final Calculation

Using the conditional probability formula:

$$
P(E \mid F)
=
\frac{P(E \cap F)}{P(F)}
$$

Substituting the values:

$$
P(E \mid F)
=
\frac{\frac{1}{36}}{\frac{1}{6}}
$$

$$
=
\frac{1}{36}\times 6
$$

$$
=
\frac{1}{6}
$$

---

## Final Answer

$$
\boxed{
P(E \mid F)=\frac{1}{6}
}
$$

---

# Intuition

We already know:

$$
D_1 = 3
$$

Therefore, the second die can be:

$$
1,2,3,4,5,6
$$

Only one value makes the total sum equal to 7:

$$
D_2 = 4
$$

Hence,

$$
P(E \mid F)=\frac{1}{6}
$$

which matches the result obtained using the formula.

# Independent Events

## Idea

Two events are called **independent** if the occurrence of one event does **not affect** the probability of the other.

In simple words:

> Knowing that one event happened gives no extra information about the other event.

---

## Example

Let:

- $S$ = Patient survived
- $M_1$ = Patient received Medicine 1

Suppose:

$$
P(S \mid M_1) = 0.70
$$

and

$$
P(S) = 0.70
$$

Notice:

$$
P(S \mid M_1) = P(S)
$$

This means that knowing whether a patient received Medicine 1 does not change the probability of survival.

Therefore,

$$
S \text{ and } M_1
$$

are **independent events**.

---

## Definition

If

$$
P(A \mid B) = P(A)
$$

then events $A$ and $B$ are independent.

provided

$$
P(B) > 0
$$

---

## Equivalent Condition

For independent events:

$$
P(A \cap B)
=
P(A)\,P(B)
$$

This is the most commonly used formula for checking independence.

---

## Intuition

For independent events:

$$
\text{Before knowing } B
\quad \text{and} \quad
\text{After knowing } B
$$

the probability of $A$ remains the same.

That is,

$$
P(A \mid B) = P(A)
$$

---

## Quick Summary

### Independent Events

$$
P(A \mid B)=P(A)
$$

or equivalently,

$$
P(A \cap B)=P(A)P(B)
$$

---

### Meaning

One event does **not influence** the probability of the other event.

# Mutually Exclusive Events

## Definition

Two events are called **mutually exclusive** if they **cannot occur at the same time**.

Mathematically,

$$
A \cap B = \varnothing
$$

This means there are no common outcomes between the two events.

---

## Example

Let:

- $S$ = Patient survived
- $D$ = Patient died

A patient cannot both survive and die simultaneously.

Therefore,

$$
S \cap D = \varnothing
$$

and

$$
P(S \cap D)=0
$$

---

## Visual Intuition

Mutually exclusive events have no overlap.

```
(A)      (B)
```

---

## Key Property

If events are mutually exclusive,

$$
A \cap B = \varnothing
$$

then

$$
P(A \cap B)=0
$$

---

# Independent Events

## Definition

Two events are called **independent** if the occurrence of one event does not affect the probability of the other.

Mathematically,

$$
P(A|B)=P(A)
$$

provided

$$
P(B)>0
$$

---

## Equivalent Formula

Starting from

$$
P(A|B)=P(A)
$$

and using

$$
P(A|B)=\frac{P(A\cap B)}{P(B)}
$$

we get

$$
P(A\cap B)=P(A)P(B)
$$

This is the most commonly used formula for independent events.

---

## Visual Intuition

Independent events can occur together.

```
      _______
    /         \
   /  A   B    \
   \ overlap   /
    \_________/
```

The overlap is allowed.

---

# Independent vs Mutually Exclusive

This is one of the most common interview questions.

| Independent Events | Mutually Exclusive Events |
|------------|--------------------|
| One event does not affect the other | Both events cannot occur together |
| $$P(A|B)=P(A)$$ | $$A\cap B=\varnothing$$ |
| Overlap usually exists | No overlap exists |
| $$P(A\cap B)=P(A)P(B)$$ | $$P(A\cap B)=0$$ |

---

# Important Observation

For non-zero probability events:

$$
\text{Mutually Exclusive}
\neq
\text{Independent}
$$

Why?

For mutually exclusive events,

$$
P(A\cap B)=0
$$

For independent events,

$$
P(A\cap B)=P(A)P(B)
$$

If

$$
P(A)>0
\quad\text{and}\quad
P(B)>0
$$

then

$$
P(A)P(B)>0
$$

which cannot equal zero.

Therefore:

> Two non-zero probability events cannot be both mutually exclusive and independent.

# Mutually Exclusive and Exhaustive Events

This is an extension of mutually exclusive events.

---

## Mutually Exclusive Events

Events are mutually exclusive if they cannot occur together.

Mathematically,

$$
A \cap B = \varnothing
$$

Example:

- $S$ = Survived
- $D$ = Died

A patient cannot both survive and die.

Therefore,

$$
S \cap D = \varnothing
$$

and

$$
P(S \cap D)=0
$$

---

## Exhaustive Events

A collection of events is called **exhaustive** if together they cover the entire sample space.

Mathematically,

$$
A_1 \cup A_2 \cup \cdots \cup A_n = U
$$

where $U$ is the sample space.

This means one of the events must occur.

---

## Example

Let:

- $S$ = Survived
- $D$ = Died

Every patient either survives or dies.

Therefore,

$$
S \cup D = U
$$

So $S$ and $D$ are exhaustive.

---

# Mutually Exclusive and Exhaustive Together

A set of events is **mutually exclusive and exhaustive** if:

### 1. They do not overlap

$$
A_i \cap A_j = \varnothing
\qquad (i \ne j)
$$

### 2. Together they cover the entire sample space

$$
A_1 \cup A_2 \cup \cdots \cup A_n = U
$$

---

## Example

Suppose:

- $S$ = Survived
- $D$ = Died
- $C$ = Medical Case Category 3

If these three events represent all possible outcomes, then:

### Mutually Exclusive

$$
S \cap D = \varnothing
$$

$$
S \cap C = \varnothing
$$

$$
D \cap C = \varnothing
$$

No two events can occur simultaneously.

### Exhaustive

$$
S \cup D \cup C = U
$$

Together they cover the entire sample space.

---

## Visual Intuition

```
+----------------------+
|          U           |
|                      |
|   (S)   (D)   (C)    |
|                      |
+----------------------+
```

- No circles overlap → Mutually Exclusive
- Entire sample space is covered → Exhaustive

---

## Important Property

For mutually exclusive and exhaustive events:

$$
P(A_1)+P(A_2)+\cdots+P(A_n)=1
$$

because they partition the entire sample space.

---

## Examples to Remember

### Example 1

Coin Toss

- H = Head
- T = Tail

$$
H \cap T = \varnothing
$$

$$
H \cup T = U
$$

Therefore:

 Mutually Exclusive

 Exhaustive

### Example 2

Dice Roll

# Problem 1: Coin Toss and Dice Roll

## Question

A coin is tossed and a die is rolled.

Let:

* $E$ = Coin shows Head
* $F$ = Die shows 3

Are $E$ and $F$ independent?

---

## Solution

The result of the coin toss does not affect the result of the die roll.

Similarly, the result of the die roll does not affect the coin toss.

Therefore,

$$
P(E|F)=P(E)
$$

and

$$
P(F|E)=P(F)
$$

Hence,

$$
E \text{ and } F
$$

are **independent events**.

---

## Intuition

Knowing that the die showed 3 gives us no information about whether the coin landed Head.

The two experiments are completely unrelated.

# Problem 2: Medicine Example

## Given

Let:

* $S$ = Patient survives
* $M_1$ = Medicine 1 is given

Suppose

$$
P(S|M_1)=0.7
$$

and

$$
P(S)=0.7
$$

---

## Observation

The probability of survival remains unchanged even after knowing that Medicine 1 was given.

Therefore,

$$
P(S|M_1)=P(S)
$$

which implies

$$
S \text{ and } M_1
$$

are independent.

---

## Extra Result

If $S$ and $M_1$ are independent, then

$$
S
$$

is also independent of

$$
M_1^c
$$

(where $M_1^c$ means Medicine 1 was **not** given).

Thus,

$$
P(S|M_1^c)=P(S)
$$

---

## Key Learning

If an event is independent of another event, it is also independent of its complement.

---

## Relationship Between Independence and Joint Probability

### Definition of Independence

Events $A$ and $B$ are independent if

$$
P(A|B)=P(A)
$$

Using conditional probability,

$$
P(A|B)=\frac{P(A\cap B)}{P(B)}
$$

Substituting independence,

$$
\frac{P(A\cap B)}{P(B)}
=
P(A)
$$

Multiplying both sides by $P(B)$,

$$
P(A\cap B)
=
P(A)P(B)
$$

---

### Most Important Formula

For independent events,

$$
P(A\cap B)
=
P(A)P(B)
$$

---

### Memory Trick

#### Conditional Probability

"Given" relationship

$$
P(A|B)
$$

#### Joint Probability

"AND" relationship

$$
P(A\cap B)
$$

#### Independence

Knowing one event does not change the probability of the other.

$$
P(A|B)=P(A)
$$

which leads to

$$
P(A\cap B)=P(A)P(B)
$$


# Problem 3: Tossing n Coins

## Question

$n$ fair coins are tossed.

Find the probability that all coins show Heads.

---

## Define Events

Let

- $E_1$ = First coin is Head
- $E_2$ = Second coin is Head
- ...
- $E_n$ = $n^{th}$ coin is Head

Then

$$
P(E_i)=\frac{1}{2}
$$

for every coin.

---

## Independence

Coin tosses are independent.

Therefore,

$$
P(E_1 \cap E_2 \cap \cdots \cap E_n)
=
P(E_1)P(E_2)\cdots P(E_n)
$$

---

## Calculation

Since

$$
P(E_i)=\frac{1}{2}
$$

for every coin,

$$
P(E_1 \cap E_2 \cap \cdots \cap E_n)
=
\left(\frac{1}{2}\right)^n
$$

---

## Final Answer

$$
P(\text{All Heads})
=
\left(\frac{1}{2}\right)^n
=
\frac{1}{2^n}
$$

---

## General Formula

If

$$
E_1,E_2,\ldots,E_n
$$

are independent events, then

$$
P(E_1 \cap E_2 \cap \cdots \cap E_n)
=
P(E_1)P(E_2)\cdots P(E_n)
$$

---

## Memory Trick

For independent events:

$$
P(\text{AND})
=
P(\text{1st}) \times P(\text{2nd}) \times \cdots
$$

Simply multiply the probabilities.
