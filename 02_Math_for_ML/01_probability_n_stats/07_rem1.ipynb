# The Big Picture

Many students learn these topics separately:

- Standardization (Z-score)
- Central Limit Theorem (CLT)
- Sampling Distribution

Think of the flow:

```text
Raw Data
   ↓
Population
   ↓
Take Samples
   ↓
Sample Means
   ↓
Distribution of Sample Means
   ↓
CLT
   ↓
Standard Error (SEM)
   ↓
Standardization (Z-score)
   ↓
Confidence Intervals
   ↓
Statistical Inference
```

```text
How far is a point
from the mean
in standard deviation units?
```

# Part 3: Central Limit Theorem (CLT)

---

# The Question CLT Answers

Suppose population distribution is:

```text
Unknown
Skewed
Messy
Not Normal
```

What happens if we repeatedly take samples?

CLT answers this.

---

# CLT Statement

If:

- observations are independent
- sample size sufficiently large

then:

```text
Distribution of sample means
approaches Normal Distribution
```

even when original population is not normal.

---

# Rule of Thumb

Usually:

$$
n \ge 30
$$

is considered sufficient.

---

# Sampling Process

Population:

```text
All heights
```

Take sample:

```text
Sample 1
```

Compute:

$$
\bar X_1
$$

Take another sample:

```text
Sample 2
```

Compute:

$$
\bar X_2
$$

Continue:

```text
Sample 3
Sample 4
...
Sample k
```

Now collect:

$$
\bar X_1,\bar X_2,\bar X_3,...
$$

---

# What Do We Get?

A new distribution:

```text
Distribution of Sample Means
```

called:

```text
Sampling Distribution
```

---

# CLT Result

The sample means follow:



More precisely:

Mean:

$$
\mu_{\bar X}
=
\mu
$$

Variance:

$$
Var(\bar X)
=
\frac{\sigma^2}{n}
$$

Standard Deviation:

$$
\frac{\sigma}{\sqrt n}
$$

---

# Most Important CLT Result

Population:

```text
Mean = μ
SD = σ
```

Sample Means:

```text
Mean = μ
SD = σ/√n
```

Notice:

```text
Mean stays same
Spread shrinks
```

Because averaging removes randomness.

Formula:

$$

$$

As:

$$

$$

$$

$$

therefore:

$$

$$

=
4
$$

## n = 100

$$

=

=

$$

```text
Smaller uncertainty
```

# Step 4: Central Limit Theorem (CLT)

---

## CLT Statement

Regardless of the original population distribution,

if:

```text
Samples are random
n is sufficiently large
```

then:

```text
Distribution of sample means
approaches a Normal Distribution
```

---

# Mathematical Form

If population has:

Mean:

$$
\mu
$$

Variance:

$$
\sigma^2
$$

Then:

$$
\bar X
\sim
N\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

or equivalently:

$$
\bar X
\sim
N\left(
\mu,
\frac{\sigma}{\sqrt n}
\right)
$$

---

# Most Important Result

Mean of sample means:

$$
\mu_{\bar X}
=
\mu
$$

The sample means center exactly around population mean.

---

Variance of sample means:

$$
Var(\bar X)
=
\frac{\sigma^2}{n}
$$

---

Standard deviation of sample means:

$$
SD(\bar X)
=
\frac{\sigma}{\sqrt n}
$$

This quantity is called:

```text
Standard Error of Mean (SEM)
```

---

# Why Does $\frac{1}{\sqrt n}$ Appear?

This is one of the most important questions.

---

## Variance of Average

Suppose:

$$
\bar X
=
\frac{X_1+X_2+\cdots+X_n}{n}
$$

For independent observations:

$$
Var(X_1+X_2+\cdots+X_n)
=
n\sigma^2
$$

Therefore:

$$
Var(\bar X)
=
Var\left(
\frac{X_1+\cdots+X_n}{n}
\right)
$$

$$
=
\frac{1}{n^2}
(n\sigma^2)
$$

$$
=
\frac{\sigma^2}{n}
$$

Taking square root:

$$
SD(\bar X)
=
\frac{\sigma}{\sqrt n}
$$

---

# Why Not $\frac{\sigma}{n}$ ?

Because:

```text
Variance scales with n
Standard deviation is square root of variance
```

Thus:

$$
\frac{\sigma}{\sqrt n}
$$

not

$$
\frac{\sigma}{n}
$$

# One-Line Summary

**The Central Limit Theorem says that if we repeatedly take random samples and compute their means, those sample means form an approximately normal distribution centered at the true population mean $\mu$, with spread equal to the Standard Error of Mean (SEM), $\sigma/\sqrt{n}$, which shrinks as sample size increases and forms the foundation of confidence intervals and statistical inference.**

# Bootstrap Confidence Intervals, Sampling Distribution, and Why We Need Bootstrap

---

# Big Picture

Previously, Central Limit Theorem (CLT) helped us build confidence intervals for the mean.

We learned:

$$
\bar X \sim N\left(\mu,\frac{\sigma}{\sqrt n}\right)
$$

which leads to:

$$
95\% \text{ CI}
=
\bar X \pm 1.96 \cdot \frac{\sigma}{\sqrt n}
$$

But this creates a question:

```text
What if we are not interested in the mean?
```

What if we want:

- Median
- 90th percentile
- 99th percentile
- Maximum
- Correlation
- Model accuracy
- F1 Score
- Regression coefficient
- Revenue per customer

Can CLT always help?

```text
Not easily.
```

This is where Bootstrap comes in.

---

# Two Ways To Build Confidence Intervals

## Method 1: CLT-Based

Uses mathematical theory.

```text
Population
      ↓
Random Samples
      ↓
Sample Means
      ↓
CLT
      ↓
Normal Distribution
      ↓
Confidence Interval
```

Works beautifully for:

- Means
- Proportions
- Many common estimators

---

## Method 2: Bootstrap-Based

Uses computation instead of mathematics.

```text
Observed Data
      ↓
Resample Many Times
      ↓
Compute Statistic
      ↓
Distribution of Statistic
      ↓
Confidence Interval
```

Works for almost anything.

---

# Why Bootstrap Was Invented

Suppose we have:

```text
100 observations
```

```text
Median
```

Question:

```text
What is the standard error of median?
```

Not obvious.

Also difficult.

Question:

```text
Can CLT directly help?
```

Not easily.

Bootstrap solves this.

# Core Bootstrap Idea

Treat observed data as a mini-population.

Suppose collected data:

$$
D=\{x_1,x_2,\ldots,x_n\}
$$

Example:

```text
[10,12,15,18,20]
```

We assume:

```text
This dataset approximately represents the population.
```

Original:

```text
[10,12,15,18,20]
```

```text
[10,10,15,20,20]
```

```text
[12,18,18,20,20]
```

```text
[15,15,15,18,20]
```

Notice:

```text
Same observation can appear multiple times.
```

This is called:

```text
Sampling with replacement
```

# Why Replacement Is Necessary

Without replacement:

```text
Every bootstrap sample
would be identical to original data.
```

No variability.

We need variability.

CLT says:

```text
If we repeatedly sample
from population
and compute means,
those means form a distribution.
```

Bootstrap mimics exactly this process.

```text
Observed Data → Bootstrap Samples
```

```text
Simulation of Sampling Distribution
```

# CLT + Bootstrap Together

Think of Bootstrap as:

```text
Computer-generated CLT
```

CLT says:

```text
Imagine infinitely many samples
from the population.
```

Bootstrap says:

```text
We don't have the population.
Let's create artificial samples
from observed data.
```

Both are trying to estimate:

```text
Sampling Distribution
```

which is the heart of statistical inference.

# Ultimate Mental Model

There are three different distributions:

### 1. Population Distribution

```text
Actual data in nature
```

Examples:

- Heights
- Salaries
- Waiting times

---

### 2. Sampling Distribution

Distribution of:

$$
\bar X
$$

or

$$
\tilde X
$$

across repeated samples.

Created theoretically by CLT.

---

### 3. Bootstrap Distribution

Distribution of statistics generated by:

```text
Resampling observed data
```

Approximates the sampling distribution.

# One-Line Summary

**Bootstrap creates thousands of artificial samples from the observed dataset (using sampling with replacement), computes a statistic on each sample, builds an empirical sampling distribution, and uses that distribution to estimate standard errors and confidence intervals for means, medians, percentiles, model metrics, and almost any statistic—even when analytical formulas are unavailable.**

CLT  → naturally gives CI for means

Bootstrap → gives CI for almost ANY statistic
            (mean, median, percentile, max, accuracy, F1, etc.)
