Many Machine Learning algorithms assume that data is approximately Gaussian (Normal).

- Logistic Regression (some statistical assumptions)
- PCA
- LDA
- Statistical Hypothesis Tests
- ANOVA
- Confidence Intervals
- Many Parametric Models

Therefore before applying such techniques, we often ask:

> Is my data approximately Normal (Gaussian)?

This is where:

1. QQ Plot
2. KS Test
3. Shapiro-Wilk Test
4. Transformations (Box-Cox, Log, Yeo-Johnson)
5. Skewness
6. Kurtosis

$$
X \sim N(\mu,\sigma^2)

where

$$

$$

$$
\sigma = Standard Deviation
$$

# QQ Plot

(Quality Check of Normality)

Actually:

```text
QQ = Quantile-Quantile
```

---

# What Does QQ Plot Compare?

It compares:

```text
Observed Quantiles
vs
Theoretical Quantiles
```

---

# Core Idea

If data is truly Normal:

Then

```text
Observed Quantiles
≈
Theoretical Normal Quantiles
```

Therefore points should fall on a straight line.

---

# What is a Quantile?

Quantiles divide data into equal probability regions.

Examples:

```text
25th percentile
50th percentile
75th percentile
90th percentile
95th percentile
99th percentile
```

---

# Median

Median is simply:

```text
50th percentile
```

---

# Example Dataset

Suppose

```python
X=[10,12,14,16,18]
```

Sorted already.

---

Percentiles:

```text
10 → lowest
14 → median
18 → highest
```

Each point corresponds to a percentile.

---

# Theoretical Quantiles

Now assume:

$$
X \sim N(\mu,\sigma)
$$

with same mean and standard deviation.

For every percentile we compute:

```text
Expected value under Normal Distribution
```

These become:

```text
Theoretical Quantiles
```

---

# QQ Plot Construction

Step 1:

Sort data.

$$
x_{(1)},x_{(2)},...,x_{(n)}
$$

---

Step 2:

Compute empirical probabilities

$$
p_i=\frac{i-0.5}{n}
$$

---

Step 3:

Convert probabilities into Normal quantiles

$$
z_i=\Phi^{-1}(p_i)
$$

where

$$
\Phi^{-1}
$$

is inverse normal CDF.

---

Step 4:

Plot

```text
Theoretical Quantiles → X-axis
Observed Quantiles → Y-axis
```

---

# Interpretation

---

## Perfect Normal Data

Points fall on line.

```text
•
  •
    •
      •
        •
```

Good.

---

## Right Skewed Data

Upper points bend upward.

```text
•
  •
    •
      ••
          •••
```

Right Tail Heavy.

---

## Left Skewed Data

Lower points bend downward.

```text
•••
  ••
    •
      •
        •
```

Left Tail Heavy.

---

## Heavy Tails

Both ends deviate.

```text
••
  •
    •
      •
        ••
```

More outliers than Normal.

---

## Light Tails

Middle stretches.

```text
     •
   •
 •
   •
     •
```

Fewer outliers.

---

If no:

# Standardization Before QQ Plot

Given:

$$
X=[x_1,x_2,\ldots,x_n]
$$

Mean:

$$
\mu
$$

Std:

$$
\sigma
$$

Transform:

$$
z_i=\frac{x_i-\mu}{\sigma}
$$

---

This produces:

$$
Z \sim N(0,1)
$$

if original data was Normal.

---

# Important

Standardization DOES NOT make data normal.

It only changes:

```text
Mean → 0
Std → 1
```

Shape remains same.

---

# Common Misconception

Wrong:

```text
StandardScaler makes data Gaussian.
```

Correct:

```text
StandardScaler only rescales data.
```

---

# KS Test (Kolmogorov-Smirnov Test)

Visual inspection is subjective.

Need numerical proof.

---

# Goal

Compare:

```text
Empirical Distribution
vs
Theoretical Distribution
```

---

# Empirical CDF

Given data:

$$
x_1,x_2,...,x_n
$$

Empirical CDF:

$$
F_n(x)
=
\frac{\#(X_i\le x)}{n}
$$

---

Interpretation:

```text
Fraction of observations
less than x
```

---

# Example

Data:

```python
[10,20,30,40,50]
```

For:

```text
x=30
```

Three values are ≤ 30.

Therefore

$$
F_n(30)=\frac35=0.6
$$

---

# Theoretical CDF

Suppose

$$
X \sim N(\mu,\sigma)
$$

Then

$$
F(x)=P(X\le x)
$$

---

# KS Statistic

KS computes maximum difference:

$$
D
=
\max_x |F_n(x)-F(x)|
$$

---

Interpretation

```text
Largest vertical gap
between two CDFs
```

---

# Decision Rule

---

## Null Hypothesis

$$
H_0:
$$

Data follows specified distribution.

---

## Alternative

$$
H_1:
$$

Data does NOT follow distribution.

---

# P-value

If:

$$
p<0.05
$$

Reject

$$
H_0
$$

Data is likely not Normal.

---

If:

$$
p>0.05
$$

Fail to reject

$$
H_0
$$

Data can be considered Normal.

---

# QQ Plot vs KS Test

| QQ Plot | KS Test |
|----------|----------|
| Visual | Statistical |
| Intuitive | Objective |
| Shows type of deviation | Gives p-value |
| Human judgment | Mathematical decision |

Best practice:

```text
Use both together
```

# Why Transform?

Goals:

1. Reduce skewness
2. Stabilize variance
3. Improve normality
4. Improve model performance

---

# Common Transformations

---

## Log Transform

$$
Y=\log(X)
$$

Good for:

```text
Income
Sales
Population
House Prices
```

---

Example

Before:

```text
10
20
50
500
5000
```

After:

```text
1
1.3
1.7
2.7
3.7
```

Much less skewed.

---

# Square Root Transform

$$
Y=\sqrt{X}
$$

Useful for count data.

Example:

```text
Number of defects
Number of calls
Number of accidents
```

---

# Cube Root Transform

$$
Y=X^{1/3}
$$

Handles stronger skewness.

---

# Reciprocal Transform

$$
Y=\frac1X
$$

Very aggressive.

Used rarely.

---

# Box-Cox Transformation

Most famous normalizing transformation.

---

# Goal

Find best exponent:

$$
\lambda
$$

that makes data closest to Normal.

---

Formula

For:

$$
\lambda\neq0
$$

$$
Y=
\frac{X^\lambda-1}{\lambda}
$$

---

For:

$$
\lambda=0
$$

$$
Y=\log(X)
$$

---

# Important Restriction

Box-Cox requires:

$$
X>0
$$

No zero.

No negative values.

---

# Examples

---

## λ = 1

No change.

$$
Y=X
$$

---

## λ = 0

Log transform.

$$
Y=\log(X)
$$

---

## λ = 0.5

Square Root.

$$
Y=\sqrt X
$$

---

## λ = -1

Reciprocal.

$$
Y=\frac1X
$$

---

# Yeo-Johnson Transformation

Improved version.

Works with:

```text
Positive values
Zero values
Negative values
```

Unlike Box-Cox.

---

# Mode

Mode means:

> Most frequently occurring value.

---

# Example

```python
[1,2,2,3,4]
```

Mode:

```text
2
```

---

# Why Mode Matters in Distributions

Mode identifies peaks.

---

# Unimodal Distribution

One peak.

```text
     /\
    /  \
   /    \
```

One mode.

---

# Bimodal Distribution

Two peaks.

```text
 /\    /\
/  \__/  \
```

Two modes.

---

# Example of Bimodal Data

Suppose employee salaries:

```text
Junior Employees
Senior Employees
```

Two groups create two peaks.

---

# Multimodal Distribution

More than two peaks.

---

Example:

```text
School students
College students
Working professionals
Retired people
```

One dataset.

Multiple peaks.

---

# Why Bimodal Distributions Matter

Mean can become misleading.

Example:

```text
20,22,24,80,82,84
```

Mean:

$$
52
$$

But nobody earns 52.

There are actually two groups.
