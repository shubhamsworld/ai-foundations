# Theoretical Quantiles

Now assume:

$$



# QQ Plot Construction

Step 1:

Sort data.

$$
x_{(1)},x_{(2)},...,x_{(n)}
$$

---

Step 2:

Compute empirical probabilities

$$
p_i=\frac{i-0.5}{n}
$$

---

Step 3:

Convert probabilities into Normal quantiles

$$
z_i=\Phi^{-1}(p_i)
$$

where

$$
\Phi^{-1}
$$

is inverse normal CDF.

---

Step 4:

Plot

```text
Theoretical Quantiles → X-axis
Observed Quantiles → Y-axis
```

# Standardization Before QQ Plot

Given:

$$
X=[x_1,x_2,\ldots,x_n]
$$

Mean:

$$
\mu
$$

Std:

$$
\sigma
$$

Transform:

$$
z_i=\frac{x_i-\mu}{\sigma}
$$

---

This produces:

$$
Z \sim N(0,1)
$$

if original data was Normal.

---

# Important

Standardization DOES NOT make data normal.

It only changes:

```text
Mean → 0
Std → 1
```

Shape remains same.

---

# Common Misconception

Wrong:

```text
StandardScaler makes data Gaussian.
```

Correct:

```text
StandardScaler only rescales data.
```

# KS Test (Kolmogorov-Smirnov Test)

Visual inspection is subjective.

Need numerical proof.

---

# Goal

Compare:

```text
Empirical Distribution
vs
Theoretical Distribution
```

---

# Empirical CDF

Given data:

$$
x_1,x_2,...,x_n
$$

Empirical CDF:

$$
F_n(x)
=
\frac{\#(X_i\le x)}{n}
$$

---

Interpretation:

```text
Fraction of observations
less than x
```

---

# Example

Data:

```python
[10,20,30,40,50]
```

For:

```text
x=30
```

Three values are ≤ 30.

Therefore

$$
F_n(30)=\frac35=0.6
$$

---

# Theoretical CDF

Suppose

$$
X \sim N(\mu,\sigma)
$$

Then

$$
F(x)=P(X\le x)
$$

---

# KS Statistic

KS computes maximum difference:

$$
D
=
\max_x |F_n(x)-F(x)|
$$

---

Interpretation

```text
Largest vertical gap
between two CDFs
```

---

# Decision Rule

---

## Null Hypothesis

$$
H_0:
$$

Data follows specified distribution.

---

## Alternative

$$
H_1:
$$

Data does NOT follow distribution.

---

# P-value

If:

$$
p<0.05
$$

Reject

$$
H_0
$$

Data is likely not Normal.

---

If:

$$
p>0.05
$$

Fail to reject

$$
H_0
$$

Data can be considered Normal.

---

# QQ Plot vs KS Test

| QQ Plot | KS Test |
|----------|----------|
| Visual | Statistical |
| Intuitive | Objective |
| Shows type of deviation | Gives p-value |
| Human judgment | Mathematical decision |

Best practice:

```text
Use both together
```

# Common Transformations

---

## Log Transform

$$
Y=\log(X)
$$

Good for:

```text
Income
Sales
Population
House Prices
```

---

Example

Before:

```text
10
20
50
500
5000
```

After:

```text
1
1.3
1.7
2.7
3.7
```

Much less skewed.

---

# Square Root Transform

$$
Y=\sqrt{X}
$$

Useful for count data.

Example:

```text
Number of defects
Number of calls
Number of accidents
```

---

# Cube Root Transform

$$
Y=X^{1/3}
$$

Handles stronger skewness.

---

# Reciprocal Transform

$$
Y=\frac1X
$$

Very aggressive.

Used rarely.

---

# Box-Cox Transformation

Most famous normalizing transformation.

---

# Goal

Find best exponent:

$$
\lambda
$$

that makes data closest to Normal.

---

Formula

For:

$$
\lambda\neq0
$$

$$
Y=
\frac{X^\lambda-1}{\lambda}
$$

---

For:

$$
\lambda=0
$$

$$
Y=\log(X)
$$

---

# Important Restriction

Box-Cox requires:

$$
X>0
$$

No zero.

No negative values.

---

# Examples

---

## λ = 1

No change.

$$
Y=X
$$

---

## λ = 0

Log transform.

$$
Y=\log(X)
$$

---

## λ = 0.5

Square Root.

$$
Y=\sqrt X
$$

---

## λ = -1

Reciprocal.

$$
Y=\frac1X
$$

---

# Yeo-Johnson Transformation

Improved version.

Works with:

```text
Positive values
Zero values
Negative values
```

Unlike Box-Cox.

# Gaussian / Normal Distribution

A Normal Distribution is written as:

$$
X \sim N(\mu,\sigma^2)
$$

where

$$
\mu = Mean
$$

$$
\sigma = Standard\ Deviation
$$

---

## Important Properties

### Symmetric

Left side mirrors right side.

---

### Bell Shaped

```text
           /\
         /    \
       /        \
-----/------------\-----
```

---

### Mean = Median = Mode

All occur at the center.

```text
Mean = Median = Mode
```

# Standard Normal Distribution

Special Normal Distribution:

$$
Z \sim N(0,1)

$$

# Standardization (Z-Score)

Convert any normal distribution into standard normal.

Formula:



---

## Example

Suppose

$$
\mu=170
$$

$$
\sigma=10
$$

Person height:

$$
x=190
$$

Then:

$$
z=\frac{190-170}{10}
=2
$$

Meaning:

```text
2 standard deviations above mean
```

---

# CDF (Cumulative Distribution Function)

CDF answers:

> What is the probability that X is less than or equal to a particular value?

Definition:

$$
F(x)=P(X\le x)
$$

---

## Example

Suppose

$$
F(170)=0.75
$$

Meaning:

```text
75% observations
are below 170
```

---

# Relationship Between PDF and CDF

PDF:

```text
Density at a point
```

CDF:

```text
Accumulated probability till that point
```

Mathematically:

$$
CDF(x)=\int_{-\infty}^{x} PDF(t)\,dt
$$

# Box-Cox Transformation

One of the most important transformations.

Works only when:

$$
x>0
$$

---

Formula:

$$
y=
\frac{x^\lambda-1}
{\lambda}
\qquad
(\lambda \neq 0)
$$

---

Special case:

$$
\lambda=0
$$

becomes:

$$
\log(x)
$$

---

## Why Box-Cox?

Automatically finds:

```text
Best λ
```

that makes data closest to normal.

---

## Scikit-Learn Example

```python

X_transformed = pt.fit_transform(X)
```

```text
       Box-Cox /
       Yeo-Johnson
             ↓
       QQ Plot Again
             ↓
      Better Gaussian
```

### One-Line Summary

> A QQ Plot compares the quantiles of your observed data with the quantiles of a theoretical distribution (usually Normal). If points follow a straight line, the data is approximately Gaussian; if not, transformations such as Log, Box-Cox, or Yeo-Johnson can often make the data more suitable for statistical modeling and machine learning.
