# Sampling Distribution and Central Limit Theorem (CLT)

---

# Why Do We Need Sampling?

In real life, we usually cannot collect data from the entire population.

Examples:

- All people in India
- All customers of Amazon
- All students in the world
- All houses in a city

The population may be:

- Too large
- Too expensive to measure
- Constantly changing

Therefore, we collect a **sample** and use it to learn about the population.

```text
Population
     ↓
Random Sample
     ↓
Statistics
     ↓
Inference about Population
```

This is the foundation of statistical inference.

---

# Population vs Sample

## Population

The complete collection of observations we want to study.

Examples:

- Heights of all adults in India
- Income of all citizens
- Lifetime of all manufactured bulbs

Population size is usually denoted by:

$$
N
$$

---

## Sample

A subset selected from the population.

Examples:

- 1000 randomly selected adults
- 500 customers surveyed
- 200 bulbs tested

Sample size is denoted by:

$$
n
$$

where

$$
n < N
$$

---

# Population Parameters

Parameters describe the population.

## Population Mean

$$
\mu
$$

Average value of the population.

---

## Population Variance

$$
\sigma^2
$$

Measures spread.

---

## Population Standard Deviation

$$
\sigma
$$

Square root of variance.

$$
\sigma=\sqrt{\sigma^2}
$$

---

# Sample Statistics

Statistics are estimates computed from a sample.

---

## Sample Mean

$$
\bar X
=
\frac{1}{n}
\sum_{i=1}^{n}X_i
$$

---

## Sample Variance

$$
s^2
=
\frac{1}{n-1}
\sum_{i=1}^{n}(X_i-\bar X)^2
$$

---

## Sample Standard Deviation

$$
s=\sqrt{s^2}
$$

---

# Important Idea

Population quantities are fixed but usually unknown.

```text
Population Mean      → μ
Population Variance  → σ²
```

Sample quantities vary from sample to sample.

```text
Sample Mean          → X̄
Sample Variance      → s²
```

---

# Random Sampling

Suppose we repeatedly draw samples.

Sample 1:

```text
20, 30, 40
```

---

Sample 2:

```text
10, 20, 50
```

---

Sample 3:

```text
30, 40, 50
```

Each sample is:

- Random
- Independent

# Computing Sample Means

For each sample:

## Sample 1

$$
\bar X_1
=
\frac{20+30+40}{3}
=
30
$$

---

## Sample 2

$$
\bar X_2
=
\frac{10+20+50}{3}
=
26.67
$$

---

## Sample 3

$$
\bar X_3
=
40
$$

---

Now we have:

```text
30
26.67
40
...
```

These are sample means.

# What Happens Next?

Instead of studying the original observations, we study:

```text
Distribution of Sample Means
```

This new distribution is called the:

# Sampling Distribution

# Sampling Distribution

A sampling distribution is:

> The probability distribution of a statistic calculated from many samples.

Examples:

- Distribution of sample means
- Distribution of sample variances
- Distribution of sample proportions

# Sampling Distribution of the Mean

Most important sampling distribution:

$$
\bar X
$$

(the sample mean)

---

Process:

```text
Population
      ↓
Many Samples
      ↓
Compute Mean of Each Sample
      ↓
Collect All Means
      ↓
Sampling Distribution of Mean
```

```text
X̄1
X̄2
X̄3
...
X̄m
```

# Central Limit Theorem (CLT)

One of the most important results in all statistics.

---

## Formal Statement

If:

1. We draw random samples of size n
2. Samples are independent
3. Population has finite mean and finite variance

then

the sampling distribution of the sample mean approaches a Normal distribution as n becomes large.

---

Mathematically:

$$
\bar X
\overset{approx}{\sim}
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
\quad
\text{for large }n
$$

---

# Meaning of the Formula

CLT says:

$$
\bar X
\sim
N
\left(
\mu,
\frac{\sigma^2}{n}
\right)
$$

---

## Mean

Sampling distribution has mean:

$$
E[\bar X]=\mu
$$

Same as population mean.

---

### Interpretation

Average of sample means equals population mean.

```text
Population Mean
       =
Mean of Sample Means
```

This property is called:

# Unbiasedness

---

# Variance of Sample Means

CLT also says:

$$
Var(\bar X)
=
\frac{\sigma^2}{n}
$$

---

Standard deviation becomes:

$$
\sigma_{\bar X}
=
\frac{\sigma}{\sqrt n}
$$

This quantity is extremely important.

It is called:

# Standard Error (SE)

---

# Standard Error

Definition:

$$
SE
=
\frac{\sigma}{\sqrt n}
$$

or

$$
SE
=
\frac{s}{\sqrt n}
$$

when population σ is unknown.

---

# Why Standard Error Matters

It tells us:

> How much sample means fluctuate from sample to sample.

Small SE:

```text
More stable estimates
```

Large SE:

```text
More uncertainty
```

---

# Effect of Increasing Sample Size

Suppose:

$$
\sigma=20
$$

---

For

$$
n=25
$$

$$
SE=\frac{20}{5}=4
$$

---

For

$$
n=100
$$

$$
SE=\frac{20}{10}=2
$$

---

For

$$
n=400
$$

$$
SE=\frac{20}{20}=1
$$

---

Notice:

```text
n ↑
SE ↓
```

Larger samples produce more stable means.

---

# Why Does CLT Work?

Think about averaging.

Suppose a sample contains:

```text
Very large value
Very small value
Average value
```

When averaged together:

```text
Extremes cancel out
```

The mean becomes more stable.

Repeated averaging creates a smooth bell-shaped distribution.

---

# Rule of Thumb

In practice:

$$
n \ge 30
$$

is often sufficient.

---

Important:

This is NOT a strict requirement.

---

If population is already normal:

Even small n works.

---

If population is heavily skewed:

May require much larger n.

Examples:

```text
n = 50
n = 100
n = 500
```

depending on skewness.

---

# Conditions for CLT

## 1. Random Sampling

Sample should be randomly selected.

---

## 2. Independence

Observations should be independent.

---

## 3. Finite Mean

Population mean exists.

$$
\mu < \infty
$$

---

## 4. Finite Variance

Population variance exists.

$$
\sigma^2 < \infty
$$

---

# When CLT May Fail

Some distributions have:

```text
Infinite Variance
```

Examples:

- Cauchy Distribution
- Certain Pareto Distributions

In such cases:

CLT may not hold.

# Sampling Distribution of Mean vs Population Distribution

| Property | Population Distribution | Sampling Distribution |
|-----------|-----------|-----------|
| Represents | Individual observations | Sample means |
| Shape | Any shape | Approximately Normal |
| Mean | μ | μ |
| Variance | σ² | σ²/n |
| Spread | Larger | Smaller |
| Depends on n | No | Yes |

---

# Why Sampling Distribution Is Narrower

Because:

$$
Var(\bar X)
=
\frac{\sigma^2}{n}
$$

Since:

$$
n>1
$$

then

$$
\frac{\sigma^2}{n}
<
\sigma^2
$$

Therefore:

```text
Sampling Distribution
=
Less Spread
=
More Concentrated
```

---

# Connection to Z-Scores

Earlier we learned:

$$
Z
=
\frac{X-\mu}{\sigma}
$$

---

For sample means:

$$
Z
=
\frac{\bar X-\mu}
{\sigma/\sqrt n}
$$

Notice:

We use Standard Error instead of Standard Deviation.

---

# Why CLT Is So Important

Without CLT:

- Confidence Intervals would not work
- Hypothesis Testing would not work
- Many ML evaluation methods would not work
- Statistical inference would be nearly impossible

CLT is the reason we can use normal-distribution mathematics even when the original data is not normal.

# One-Line Summary

The Central Limit Theorem states that if we repeatedly draw independent random samples of size n from any population with finite mean and variance, the distribution of the sample means becomes approximately Gaussian with mean μ and variance σ²/n, regardless of the shape of the original population distribution.

# Q-Q Plot (Quantile–Quantile Plot)

---

# Why Do We Need a Q-Q Plot?

In Machine Learning and Statistics, many algorithms assume that data follows a **Normal (Gaussian) Distribution**.

Examples:

- Linear Regression (for residual analysis)
- Statistical Hypothesis Testing
- Confidence Intervals
- ANOVA
- Many classical statistical models

Before using these methods, we often ask:

> Does my data actually follow a Normal Distribution?

A Q-Q Plot is one of the most commonly used graphical tools to answer this question.

---

# What is a Q-Q Plot?

Q-Q stands for:

```text
Quantile vs Quantile
```

A Q-Q plot compares:

```text
Quantiles of Observed Data
               VS
Quantiles of Reference Distribution
```

Usually the reference distribution is:

$$
N(0,1)
$$

(Standard Normal Distribution)

---

# Intuition

Suppose we have a dataset:

$$
X = \{x_1,x_2,x_3,\ldots,x_n\}
$$

and we want to know:

> Does X follow a Gaussian Distribution?

Instead of comparing every data point individually, we compare:

```text
1st percentile
2nd percentile
3rd percentile
...
99th percentile
```

of the data

with

```text
1st percentile
2nd percentile
3rd percentile
...
99th percentile
```

of a Normal Distribution.

If both sets of percentiles match closely:

```text
Data ≈ Normal Distribution
```

---

# What is a Quantile?

A quantile divides data into ordered portions.

Examples:

| Quantile | Meaning |
|-----------|----------|
| 25th percentile | 25% of observations lie below it |
| 50th percentile | Median |
| 75th percentile | 75% of observations lie below it |

---

Example:

Sorted data:

```text
2, 4, 5, 7, 8, 10, 12, 15, 18
```

Median:

```text
8
```

because 50% of values lie below it.

Therefore:

```text
8 = 50th percentile
```

---

# Goal of Q-Q Plot

Given:

$$
X=\{x_1,x_2,\ldots,x_n\}
$$

Unknown distribution.

Question:

```text
Is X Normally Distributed?
```

Possible approaches:

### Graphical

- Histogram
- KDE
- Q-Q Plot

### Statistical Tests

- Kolmogorov-Smirnov Test (KS Test)
- Anderson-Darling Test (AD Test)
- Shapiro-Wilk Test
- Jarque-Bera Test

Among graphical methods:

```text
Q-Q Plot is the most informative.
```

---

# Core Idea Behind Q-Q Plot

We compare:

```text
Observed Quantiles
```

with

```text
Theoretical Quantiles
```

from a known distribution.

Usually:

$$
N(0,1)
$$

---

# Step-by-Step Construction

Suppose:

```text
X = observed data
```

containing:

```text
500 observations
```

---

# Step 1: Sort the Data

Original:

```text
x₁, x₂, x₃, ...
```

After sorting:

```text
x(1), x(2), x(3), ... , x(n)
```

where:

```text
x(1) = smallest value
x(n) = largest value
```

---

Example:

Original:

```text
10, 2, 18, 5, 7
```

Sorted:

```text
2, 5, 7, 10, 18
```

---

# Step 2: Compute Observed Quantiles

Each sorted observation corresponds to a percentile.

Example:

```text
x(1)   → 1st percentile
x(25)  → 25th percentile
x(50)  → 50th percentile
...
```

These are called:

```text
Observed Quantiles
```

---

# Step 3: Create Reference Distribution

Generate values from:

$$
Y \sim N(0,1)
$$

Example:

```python
np.random.normal(0,1,size=1000)
```

---

These values represent a Standard Normal Distribution.

Sort them:

```text
y(1), y(2), y(3), ..., y(n)
```

---

Compute their percentiles.

These become:

```text
Theoretical Quantiles
```

---

# Step 4: Create Quantile Pairs

Pair corresponding quantiles:

```text
(y(1), x(1))
(y(2), x(2))
(y(3), x(3))
...
(y(n), x(n))
```

---

Meaning:

```text
1st percentile vs 1st percentile
2nd percentile vs 2nd percentile
...
```

---

# Step 5: Plot Points

X-axis:

```text
Theoretical Quantiles
```

Y-axis:

```text
Observed Quantiles
```

Each point:

```text
(theoretical quantile,
 observed quantile)
```

---

# Perfect Normal Case

Suppose data is actually:

$$
X \sim N(20,5)
$$

Even though:

```text
Mean = 20
Std = 5
```

instead of

$$
N(0,1)
$$

the Q-Q plot points still lie approximately on a straight line.

Example:

```text
            •
         •
      •
   •
•
-------------------->
```

Reason:

Both distributions belong to the same Normal family.

Only location and scale differ.

---

---

## Case 1: Straight Line

```text
•
  •
    •
      •
        •
```

Interpretation:

```text
Data is approximately normal.
```

Good sign.

---

## Case 2: S-Shaped Curve

```text
•
   •
      •
         •
      •
   •
•
```

Interpretation:

```text
Data has heavier tails than Normal.
```

Possible:

- Outliers
- Heavy-tailed distributions

---

## Case 3: Inverted S

```text
      •
    •
  •
•
  •
    •
      •
```

Interpretation:

```text
Lighter tails than Normal.
```

---

## Case 4: Right Tail Deviates Upward

```text
•
 •
  •
   •
      •
           •
```

Interpretation:

```text
Positive Skewness
```

Right-skewed distribution.

Examples:

- Income
- House prices
- Web traffic

---

## Case 5: Left Tail Deviates Downward

```text
          •
       •
    •
  •
 •
•
```

Interpretation:

```text
Negative Skewness
```

# Why Straight Line Means Normal?

Suppose:

Observed:

```text
10th percentile = 15
50th percentile = 20
90th percentile = 26
```

Normal distribution:

```text
10th percentile = -1.28
50th percentile = 0
90th percentile = 1.28
```

Notice:

Percentiles correspond consistently.

Therefore:

```text
Observed Quantiles
      ∝
Theoretical Quantiles
```

which produces a straight line.

---

# Mathematical Interpretation

If:

$$
X
=
\mu+\sigma Z
$$

where

$$
Z\sim N(0,1)
$$

then:

```text
All quantiles scale linearly.
```

Therefore:

```text
Q-Q plot becomes a line.
```

---

# Relationship with Standardization

Recall:

$$
Z
=
\frac{X-\mu}{\sigma}
$$

Standardization converts:

$$
X\sim N(\mu,\sigma^2)
$$

into

$$
Z\sim N(0,1)
$$

Q-Q plot is essentially checking:

> Can the observed data be transformed into a standard normal through a linear scaling?

If yes:

```text
Points lie on a straight line.
```

# Q-Q Plot vs Statistical Tests

---

## Q-Q Plot

Advantages:

- Visual
- Shows type of deviation
- Detects skewness
- Detects heavy tails

Disadvantages:

- Subjective

---

## KS Test

Compares:

```text
Empirical CDF
VS
Theoretical CDF
```

Returns p-value.

---

## Anderson-Darling Test

Similar to KS but gives more weight to tails.

Usually stronger for detecting non-normality.

---

## Shapiro-Wilk Test

One of the most popular normality tests.

Good for smaller datasets.

---

# Limitations of Q-Q Plot

## Small Sample Size

Example:

```text
n = 10
```

Random variation may look non-normal.

Hard to judge.

---

## Very Large Datasets

Example:

```text
n = 1,000,000
```

Tiny deviations become visible.

Need practical judgment.

---

## Subjective Interpretation

Two people may interpret the same plot differently.

Therefore:

```text
Use Q-Q Plot
+
Statistical Test
```

together.

# Example 1: T-Shirt Inventory Planning

---

## Problem

A company has:

```text
100,000 employees
```

and needs to order T-shirts.

Available sizes:

```text
S, M, L, XL
```

Question:

> How many XL T-shirts should be ordered?

---

## Naive Approach

Measure height of all:

```text
100,000 employees
```

and determine size.

Problem:

```text
Expensive
Time consuming
Impractical
```

---

## Statistical Approach

Take a random sample:

```text
n = 500 employees
```

Measure heights.

---

Suppose after collecting data:

```text
Mean Height = 170 cm
Standard Deviation = 5 cm
```

We model:

$$
H \sim N(170,5^2)
$$

where

```text
H = employee height
```

---

# Why Gaussian?

Human physical measurements often approximately follow:

```text
Normal Distribution
```

Examples:

- Height
- Weight
- Blood pressure
- IQ

---

# Defining XL Size

Suppose company policy says:

```text
Height > 180 cm
→ XL
```

We need:

$$
P(H > 180)
$$

---

# Converting to Z-score

Recall:

$$
Z=\frac{X-\mu}{\sigma}
$$

Therefore:

$$
Z=\frac{180-170}{5}
$$

$$
Z=2
$$

---

Using Z-table:

$$
P(Z<2)=0.9772
$$

Thus:

$$
P(H>180)
=
1-0.9772
\approx0.0228
$$

---

Meaning:

```text
2.28% employees need XL shirts
```

---

For:

```text
100,000 employees
```

Expected XL shirts:

$$
100000 \times 0.0228
=
2280
$$

---

# Business Insight

Without measuring:

```text
100,000 employees
```

we estimated:

```text
≈ 2280 XL shirts
```

using:

- Sampling
- Distribution fitting
- Probability

---

# Why PDF Helps Here

Recall:

PDF tells:

```text
Density of observations
```

Higher PDF:

```text
More people likely around that height
```

Lower PDF:

```text
Fewer people around that height
```

---

# Why CDF Helps Here

CDF tells:

$$
P(X \le x)
$$

Example:

$$
P(H \le 180)
=
0.9772
$$

Meaning:

```text
97.72% employees are
180 cm or shorter
```

Therefore:

$$
P(H>180)
=
1-CDF(180)
$$

---

# Visualization

PDF:

```text
Area under curve
to the right of 180 cm
```

gives:

$$
P(H>180)
$$

---

CDF:

```text
Height = 180 cm
```

might correspond to:

```text
97.72%
```

Thus:

```text
Remaining 2.28%
```

need XL shirts.

# Chebyshev's Inequality

---

# Why Do We Need Chebyshev's Inequality?

Previously, for a **Normal (Gaussian) Distribution**, we used the **68-95-99.7 Rule**:

| Interval | Probability |
|-----------|-------------|
| μ ± 1σ | ≈ 68% |
| μ ± 2σ | ≈ 95% |
| μ ± 3σ | ≈ 99.7% |

For example, if student heights follow:

$$
X \sim N(\mu,\sigma^2)
$$

then:

$$
P(\mu-2\sigma \le X \le \mu+2\sigma)\approx95\%
$$

meaning roughly 95% of students lie within two standard deviations of the mean.

---

## The Problem

The Empirical Rule only works when:

```text
Distribution is Normal (Gaussian)
```

But in practice we often know:

- Mean (μ)
- Standard deviation (σ)

and **do not know the underlying distribution**.

Examples:

- Salaries
- Product demand
- Customer spending
- Website traffic
- Stock returns
- Insurance claims

Question:

> Can we still say something useful about how much data lies near the mean?

The answer is:

Yes.

This is exactly what **Chebyshev's Inequality** provides.

---

# Core Idea

Chebyshev's Inequality works for:

```text
ANY distribution
```

provided:

1. Mean exists and is finite
2. Variance exists and is finite
3. Standard deviation is non-zero

The distribution may be:

- Normal
- Skewed
- Bimodal
- Uniform
- Unknown

It doesn't matter.

---

# Formal Statement

For any random variable \(X\) with mean \(μ\) and standard deviation \(σ\):



for:

$$
k > 1
$$

---

## Interpretation

The probability that a value lies

```text
k standard deviations or more
away from the mean
```

is at most:

$$
\frac{1}{k^2}
$$

---

# More Useful Form

Taking the complement:

\[
P(|X-\mu|<k\sigma)
\ge
1-\frac{1}{k^2}
\]

which means:

> At least \(1-\frac{1}{k^2}\) of the observations lie within \(k\) standard deviations of the mean.

---

# Understanding the Two Forms

---

## Outside the Interval

\[
P(|X-\mu|\ge k\sigma)
\le
\frac{1}{k^2}
\]

means:

```text
Maximum proportion
outside the interval
```

---

## Inside the Interval

\[
P(|X-\mu|<k\sigma)
\ge
1-\frac{1}{k^2}
\]

means:

```text
Minimum proportion
inside the interval
```

---

# Important Interpretation

Chebyshev gives:

```text
Guaranteed Minimum Coverage
```

not the exact probability.

This is why the inequality uses:

```text
≥
```

instead of:

```text
=
```

---

# Common Results

---

## k = 2

\[
P(|X-\mu|<2\sigma)
\ge
1-\frac14
\]

\[
P(|X-\mu|<2\sigma)
\ge0.75
\]

Thus:

```text
At least 75%
of observations lie within
μ ± 2σ
```

---

## k = 3

\[
P(|X-\mu|<3\sigma)
\ge
1-\frac19
\]

\[
P(|X-\mu|<3\sigma)
\ge
\frac89
\]

\[
P(|X-\mu|<3\sigma)
\ge
88.89\%
\]

Thus:

```text
At least 88.89%
lie within μ ± 3σ
```

---

## k = 4

\[
P(|X-\mu|<4\sigma)
\ge
1-\frac1{16}
\]

\[
P(|X-\mu|<4\sigma)
\ge
93.75\%
\]

---

# Quick Reference Table

| k | Interval | Minimum Probability |
|----|-----------|-------------------|
| 2 | μ ± 2σ | ≥ 75% |
| 3 | μ ± 3σ | ≥ 88.89% |
| 4 | μ ± 4σ | ≥ 93.75% |
| 5 | μ ± 5σ | ≥ 96% |

---

# Visual Intuition

Suppose:

```text
Mean = μ
Std Dev = σ
```

Then:

```text
μ-kσ          μ          μ+kσ
 |------------|------------|
```

Chebyshev says:

```text
Most data must stay
inside this interval.
```

The farther we move from the mean:

```text
larger k
```

the larger the guaranteed coverage becomes.

---

# Example 1: Salary Analysis

Suppose salaries have:

```text
μ = $40,000
σ = $10,000
```

We do NOT know the distribution.

Question:

> What percentage of employees earn between $20k and $60k?

---

## Step 1

Observe:

\[
20k = \mu - 2\sigma
\]

\[
60k = \mu + 2\sigma
\]

Thus interval is:

\[
[\mu-2\sigma,\mu+2\sigma]
\]

---

## Step 2

Apply Chebyshev:

\[
P(|X-\mu|<2\sigma)
\ge
0.75
\]

Therefore:

\[
P(20k<X<60k)
\ge
75\%
\]

---

## Interpretation

We cannot say:

```text
Exactly 75%
```

We can only say:

```text
At least 75%
```

Actual value may be:

```text
80%
90%
95%
98%
```

Chebyshev guarantees the minimum.

---

# Example 2: Same Problem if Salaries are Normal

Now suppose:

\[
X\sim N(\mu,\sigma^2)
\]

Then:

\[
P(\mu-2\sigma<X<\mu+2\sigma)
\approx95.4\%
\]

using the Empirical Rule.

---

# Comparing Results

| Method | Requirement | Result |
|----------|-------------|----------|
| Chebyshev | Any distribution | ≥ 75% |
| Gaussian Rule | Normal distribution | ≈ 95.4% |

---

# Why 95% > 75%?

This is one of the most important concepts.

Students often ask:

> If Gaussian says 95%, why does Chebyshev say only 75%?

---

Because Chebyshev is designed for:

```text
ALL possible distributions
```

including very strange ones.

It must remain correct for:

- Normal distributions
- Highly skewed distributions
- Heavy-tailed distributions
- Weird multimodal distributions

Therefore it provides a:

```text
Worst-case guarantee
```

---

# Key Insight

Suppose actual data is normal.

Then:

\[
95.4\%
\]

lie inside:

\[
\mu\pm2\sigma
\]

Chebyshev says:

\[
\ge75\%
\]

Both statements are true.

Because:

\[
95.4\% \ge 75\%
\]

Chebyshev is not trying to give the exact answer.

It only guarantees:

```text
At least this much coverage
must exist.
```

---

# Why Is Chebyshev Conservative?

Imagine you know only:

```text
Mean = 40k
Std Dev = 10k
```

but nothing else.

You don't know whether salaries are:

```text
Normal
Skewed
Bimodal
Heavy-tailed
```

The safest statement you can make is:

```text
At least 75%
must lie within μ ± 2σ
```

Chebyshev chooses safety over precision.

# One-Line Summary

Chebyshev's Inequality is a distribution-free safety guarantee: for any random variable with finite mean and variance, at least \(1-\frac{1}{k^2}\) of observations must lie within \(k\) standard deviations of the mean. It is less precise than the Gaussian 68-95-99.7 rule but works even when the underlying distribution is completely unknown.

# Uniform Distribution (Discrete & Continuous)

# Why Do We Need Uniform Distribution?

Uniform Distribution is one of the simplest probability distributions.

The central idea is:

> Every possible outcome is equally likely.

Whenever we have:

```text
No outcome is preferred over another
```

Uniform Distribution becomes a natural model.

Examples:

- Fair dice roll
- Random card selection
- Random integer generation
- Random position in an interval
- Random waiting time within a fixed period
- Random angle selection
- Monte Carlo simulations

---

# Uniform Distribution Family

Uniform Distribution has two forms:

```text
Uniform Distribution
│
├── Discrete Uniform Distribution
│
└── Continuous Uniform Distribution
```

The distinction depends on whether the random variable is:

```text
Discrete  → countable outcomes
Continuous → infinitely many values
```

---

# Discrete vs Continuous Random Variables

## Discrete Random Variable

Can take only specific values.

Examples:

```text
Dice Roll
X ∈ {1,2,3,4,5,6}

Number of Heads
X ∈ {0,1,2,3}
```

---

## Continuous Random Variable

Can take any value in an interval.

Examples:

```text
Height
Time
Weight
Temperature
Voltage
```

For example:

```text
Height ∈ [150cm, 200cm]
```

There are infinitely many possible values.

---

# Uniform Distribution Intuition

Imagine a fair die.

Possible outcomes:

```text
1,2,3,4,5,6
```

Each outcome occurs with probability:

$$
\frac16
$$

No face is favored.

This is a Discrete Uniform Distribution.

---

Now imagine choosing a random point on a ruler between:

```text
0 and 10
```

Every interval of equal length is equally likely.

This is a Continuous Uniform Distribution.

# Important Clarification

Many students initially think:

```text
PMF and PDF are the same thing
```

Conceptually:

 Both measure likelihood.

Mathematically:

 They are different.

---

# PMF (Discrete Case)

For a discrete random variable:

$$
P(X=x)
$$

directly gives probability.

Example:

```text
Fair Dice
```

$$
P(X=4)=\frac16
$$

This is an actual probability.

---

# PDF (Continuous Case)

For a continuous variable:

$$
f(x)
$$

is NOT probability.

Instead:

$$
P(a\le X\le b)
=
\int_a^b f(x)\,dx
$$

Probability comes from area under the curve.

---

# Why PDF Is Not Probability

Suppose:

$$
X \sim U(0,10)
$$

Question:

$$
P(X=5)
$$

Answer:

$$
0
$$

because a single point has zero width.

Only intervals have probability.

Example:

$$
P(4 \le X \le 6)
=
\int_4^6 f(x)\,dx
=
0.2
$$

---

Because probability exists only at discrete points.

Whenever we hit:

Therefore:

```text
Flat → Jump → Flat → Jump
```

# CDF of Continuous Uniform

Recall:

$$
F(x)=P(X\le x)
$$

For continuous distributions:

```text
CDF is obtained by integrating PDF.
```

---

Since:

$$
f(x)=\frac1{b-a}
$$

Integrating gives:

$$
F(x)=
\begin{cases}
0,& x<a\\
\frac{x-a}{b-a},& a\le x\le b\\
1,& x>b
\end{cases}
$$

```text
1 |         /
  |       /
  |     /
0 |____/
     a   b
```

Straight line

Because probability accumulates continuously.

Every tiny interval contributes probability.

No jumps occur.

Therefore:

```text
Continuous
Differentiable
Smooth
```

# Example 1: T-Shirt Inventory Planning

---

## Problem

A company has:

```text
100,000 employees
```

and needs to order T-shirts.

Available sizes:

```text
S, M, L, XL
```

Question:

> How many XL T-shirts should be ordered?

---

## Naive Approach

Measure height of all:

```text
100,000 employees
```

and determine size.

Problem:

```text
Expensive
Time consuming
Impractical
```

---

## Statistical Approach

Take a random sample:

```text
n = 500 employees
```

Measure heights.

---

Suppose after collecting data:

```text
Mean Height = 170 cm
Standard Deviation = 5 cm
```

We model:

$$
H \sim N(170,5^2)
$$

where

```text
H = employee height
```

---

# Why Gaussian?

Human physical measurements often approximately follow:

```text
Normal Distribution
```

Examples:

- Height
- Weight
- Blood pressure
- IQ

---

# Defining XL Size

Suppose company policy says:

```text
Height > 180 cm
→ XL
```

We need:

$$
P(H > 180)
$$

---

# Converting to Z-score

Recall:

$$
Z=\frac{X-\mu}{\sigma}
$$

Therefore:

$$
Z=\frac{180-170}{5}
$$

$$
Z=2
$$

---

Using Z-table:

$$
P(Z<2)=0.9772
$$

Thus:

$$
P(H>180)
=
1-0.9772
\approx0.0228
$$

---

Meaning:

```text
2.28% employees need XL shirts
```

---

For:

```text
100,000 employees
```

Expected XL shirts:

$$
100000 \times 0.0228
=
2280
$$

---

# Business Insight

Without measuring:

```text
100,000 employees
```

we estimated:

```text
≈ 2280 XL shirts
```

using:

- Sampling
- Distribution fitting
- Probability

---

# Why PDF Helps Here

Recall:

PDF tells:

```text
Density of observations
```

Higher PDF:

```text
More people likely around that height
```

Lower PDF:

```text
Fewer people around that height
```

---

# Why CDF Helps Here

CDF tells:

$$
P(X \le x)
$$

Example:

$$
P(H \le 180)
=
0.9772
$$

Meaning:

```text
97.72% employees are
180 cm or shorter
```

Therefore:

$$
P(H>180)
=
1-CDF(180)
$$

---

# Visualization

PDF:

```text
Area under curve
to the right of 180 cm
```

gives:

$$
P(H>180)
$$

---

CDF:

```text
Height = 180 cm
```

might correspond to:

```text
97.72%
```

Thus:

```text
Remaining 2.28%
```

need XL shirts.

$$
P(S>100000)
=
P(Z>2)
\approx 0.0228
$$

---

Meaning:

```text
2.28% employees
earn above $100k
```

---

For:

```text
50,000 employees
```

Expected:

$$
50000\times0.0228
=
1140
$$

employees.
