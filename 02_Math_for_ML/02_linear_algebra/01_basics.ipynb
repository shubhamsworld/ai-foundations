# 1. Real World Problem

Suppose we have an electronic manufacturing plant in Japan.

A conveyor belt is moving electronic parts.

```
Conveyor Belt
---------------------------------------------------
| Part1 | Part2 | Part3 | Part4 | Part5 | .......
---------------------------------------------------
                      ↓
                   Scanner
```

The scanner captures information about each part:

- length
- width
- weight
- color
- temperature
- resistance
- etc.

The goal is to automatically decide:

- Is the part manufactured correctly?
- Which type of part is this?
- What will be its future performance?

This is where Machine Learning comes in.

# 3. Axis Parallel Decision Boundaries

## Vertical Boundary

Rule:

```python
if l > 10:
      Type2
else:
      Type1
```

Equation:

$$
l=10
$$

---

## Horizontal Boundary

Rule:

```python
if w > 8:
      Type2
else:
      Type1
```

Equation:

$$
w=8
$$

---

These are called:

# Axis Parallel Decision Boundaries

because they are parallel to one of the coordinate axes.

---

# Why Decision Trees Love These Boundaries

Decision Trees repeatedly ask questions like:

```python
if feature < threshold:
      left
else:
      right
```

Examples:

```python
if age < 30
if salary > 50000
if length < 10
if width > 8
```

Every such rule creates:

- Vertical line
- Horizontal line

---

Therefore:

# Decision Trees create axis-parallel boundaries.

---

Algorithms:

- Decision Tree
- Random Forest
- Gradient Boosted Trees (GBDT)
- XGBoost
- LightGBM
- CatBoost

all build piecewise axis-parallel regions.

---

# Example

```
if length < 10 and width < 8:
       Type1
else:
       Type2
```

This forms a rectangle in feature space.

---

# 4. Slanted Decision Boundary

Sometimes vertical or horizontal lines are not enough.

Suppose classes look like:

```
xxxxx
 xxxxx
  xxxxx
   ooooo
    ooooo
```

A slanted line works better.

Equation:

$$
w=-2l+30
$$

or

$$
w_1l+w_2w+b=0
$$

---

Rule:

```python
if w1*l + w2*w + b > 0:
       Type1
else:
       Type2
```

---

This is the foundation of:

- Logistic Regression
- Perceptron
- Linear SVM

---

# General Equation of a Line

$$
w_1x_1+w_2x_2+b=0
$$

where

$$
x_1=l
$$

$$
x_2=w
$$

---

# 5. Curved Decision Boundaries

Sometimes classes are arranged like this:

```
xxxxxxxx
xxxxxxxx
   ooooooo
   ooooooo
```

A straight line cannot separate them.

Need:

- curve
- parabola
- polynomial boundary

---

Algorithms:

- Polynomial Logistic Regression
- Kernel SVM
- Neural Networks

---

# 6. Circular Decision Boundary

Suppose one class lies inside another.

```
      ooooooo
    oo       oo
   o    xxxx   o
   o    xxxx   o
    oo       oo
      ooooooo
```

No line can separate them.

Need:

# Circle

Equation:

$$
(x-a)^2+(y-b)^2=r^2
$$

---

Rule:

```python
if (x-a)^2+(y-b)^2 < r^2:
       Class1
else:
       Class2
```

---

Algorithms capable of learning this:

- Kernel SVM
- Neural Networks
- RBF Networks

---

# Shapes of Decision Boundaries

| Algorithm | Boundary |
|-----------|-----------|
| Decision Tree | Vertical/Horizontal |
| Random Forest | Piecewise rectangles |
| Logistic Regression | Straight line |
| Linear SVM | Straight line |
| Polynomial SVM | Curve |
| RBF SVM | Complex curves |
| Neural Network | Arbitrary shapes |

# 7. Query Point Classification

Suppose:

$$
x_q=(l,w)
$$

For a line:

$$
w_1l+w_2w+b=0
$$

Compute:

$$
s=w_1l+w_2w+b
$$

If:

$$
s>0
$$

predict:

Type 1.

Else:

Type 2.

---

# Confidence of Prediction

The farther the point is from the boundary:

- more confident prediction.

The closer the point:

- less confident prediction.

---

# Distance from Line

For line:

$$
Ax+By+C=0
$$

distance of point:

$$
(x_0,y_0)
$$

is

$$
d=
\frac{|Ax_0+By_0+C|}
{\sqrt{A^2+B^2}}
$$

---

Large distance:

high confidence.

Small distance:

low confidence.

x_i=(x_1,x_2,\dots,x_d)

d

x_q

# 7. Decision Boundary

To classify points, we create a boundary.

This boundary divides the feature space into regions.

This boundary is called:

# Decision Boundary

---

# Example

```text
Type1 | Type2
xxxx  | ooooo
xxxx  | ooooo
```

Boundary:

```text
|
|
|
```

---

# A model is nothing but a decision boundary.

---

# 8. Vertical Decision Boundary

Rule:

```python
if length > 10:
    Type2
else:
    Type1
```

Equation:

$$
l=10
$$

This is a:

# Vertical Line

---

# Horizontal Boundary

Rule:

```python
if width > 8:
    Type2
else:
    Type1
```

Equation:

$$
w=8
$$

This is:

# Horizontal Line

---

# Axis Parallel Decision Boundaries

Vertical and horizontal boundaries are called:

# Axis Parallel Boundaries

because they are parallel to one of the coordinate axes.

---

# Algorithms that like Axis Parallel Boundaries

- Decision Tree
- Random Forest
- XGBoost
- LightGBM
- CatBoost
- GBDT

---

# Why?

Because trees repeatedly ask questions like:

```python
if age < 30
if salary > 50000
if length < 10
```

Each question creates:

- vertical line
- horizontal line

---

# Example

```python
if length < 10 and width < 8:
    Type1
else:
    Type2
```

This creates a rectangle.

---

# 9. Slanted Decision Boundary

Sometimes axis-parallel lines cannot separate the data.

Then we use:

# Slanted Line

---

Equation:

$$
w_1x_1+w_2x_2+b=0
$$

or

$$
Ax+By+C=0
$$

---

Rule:

```python
if w1*x1 + w2*x2 + b > 0:
    Type1
else:
    Type2
```

---

Algorithms:

- Logistic Regression
- Perceptron
- Linear SVM

---

# 10. Misclassified Points

# Misclassified Points

Classification error:

$$
Error

\text{Number of Misclassified Points}
$$

---

# Goal of Training

Find the boundary that minimizes:

$$
Error
$$

or

$$
Loss
$$

# Training = Optimization

```text
Try Line L1
Compute Error

Try Line L2
Compute Error

Try Line L3
Compute Error

Choose Best
```

This process is called:

# Optimization

# 11. Nonlinear Decision Boundaries

Sometimes a line cannot separate the data.

Then we use:

- curve
- circle
- ellipse
- polynomial
- complicated surfaces

---

# Circular Boundary

Equation:

$$
(x-a)^2+(y-b)^2=r^2
$$

Rule:

```python
if (x-a)**2 + (y-b)**2 < r**2:
    Type1
else:
    Type2
```

---

Algorithms:

- Kernel SVM
- Neural Networks
- RBF Networks

---

# Elliptical Boundary

Equation:

$$
\frac{(x-a)^2}{r_1^2}
+
\frac{(y-b)^2}{r_2^2}
=
1
$$

---

# 12. Geometry in Higher Dimensions

In:

## 2D

decision boundary is:

# Line

---

In:

## 3D

decision boundary is:

# Plane

---

In:

## d-dimensional space

decision boundary is:

# Hyperplane

---

# General Rule

A boundary in:

$$
d
$$

dimensions has:

$$
d-1
$$

dimensions.

---

Examples:

| Space | Boundary |
|-------|-----------|
| 2D | 1D Line |
| 3D | 2D Plane |
| 4D | 3D Hyperplane |
| dD | (d−1)-dimensional Hyperplane |

---

# Why?

Because one equation reduces one degree of freedom.

---

# Example

In 3D:

Point:

$$
(x_1,x_2,x_3)
$$

Plane:

$$
w_1x_1+w_2x_2+w_3x_3+b=0
$$

This plane separates the classes.

---

# Door Wall Analogy

In 2D:

Boundary = line.

In 3D:

Boundary = wall.

You cannot separate rooms using a line.

You need a wall.

Similarly:

Higher dimensions require:

# Hyperplanes

---

# 13. Hyperplane Equation

General equation:

$$
w^Tx+b=0
$$

where

$$
w=
[w_1,w_2,\dots,w_d]
$$

and

$$
x=
[x_1,x_2,\dots,x_d]
$$

---

Expanded form:

$$
w_1x_1+w_2x_2+\dots+w_dx_d+b=0
$$

---

# This equation is the model.

---

# Model Parameters

Parameters:

$$
w_1,w_2,\dots,w_d,b
$$

Training means learning these values.

---

# 14. Prediction Rule

Compute:

$$
s=w^Tx+b
$$

Then

```python
if s > 0:
    Class 1
else:
    Class 2
```

---

This is called:

# Decision Function

---

# 15. Signed Distance

The value

$$
w^Tx+b
$$

tells us:

1. Which side of the boundary?
2. How far from the boundary?

---

Distance:

$$
d=
\frac{|w^Tx+b|}
{\|w\|}
$$

---

Large distance:

```text
More confident prediction
```

Small distance:

```text
Less confident prediction
```

---

# 16. Normal Vector

The vector:

$$
w
$$

is perpendicular to the hyperplane.

It is called:

- normal vector
- perpendicular vector

---

The sign of

$$
w^Tx+b
$$

determines:

```text
which side of the hyperplane
```

A machine learning classifier learns a geometric object (line, plane, curve, or hyperplane) that divides the feature space into regions, and prediction is simply determining on which side of that boundary a new point lies.
