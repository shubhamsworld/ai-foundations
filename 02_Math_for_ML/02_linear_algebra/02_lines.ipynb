# Geometry for Machine Learning: Lines, Slopes, Hyperplanes and Decision Boundaries

---

# Why are we studying lines in Machine Learning?

Almost every classification algorithm starts with a simple idea:

```text
Data Points
      ↓
Find a Line
      ↓
Separate Classes
      ↓
Generalize to Higher Dimensions
      ↓
Hyperplane
```

Algorithms like:

- Perceptron
- Logistic Regression
- Linear SVM
- Neural Networks (last layer)

all rely on this geometric idea.

---

# 1. Equation of a Line

The most common form is:

$$
y=mx+c
$$

where

- \(m\) = slope of the line
- \(c\) = y-intercept

---

# Geometric Meaning

```text
          y
          ^
          |
          |        /
          |      /
          |    /
          |  /
----------|----------------> x
          |
```

---

# What is the slope?

Slope tells us:

> How fast y changes when x changes.

Mathematically,

$$
m=\frac{\Delta y}{\Delta x}
$$

or

$$
m=\frac{y_2-y_1}{x_2-x_1}
$$

---

# Example

Suppose:

```text
Point A = (1,2)
Point B = (3,6)
```

Then

$$
m=\frac{6-2}{3-1}
$$

$$
m=\frac{4}{2}
$$

$$
m=2
$$

Meaning:

```text
When x increases by 1,
y increases by 2.
```

---

# Visual Meaning

```text
rise
  |
  |____ run
```

Therefore,

$$
m=\frac{\text{Rise}}{\text{Run}}
$$

---

# Example

```text
rise = 6
run = 3
```

then

$$
m=\frac{6}{3}=2
$$

---

# 2. Y-Intercept

The intercept is where the line cuts the y-axis.

Since on the y-axis,

$$
x=0
$$

putting into

$$
y=mx+c
$$

gives

$$
y=c
$$

Therefore:

> c is the point where the line intersects the y-axis.

---

# Example

Line:

$$
y=2x+3
$$

At

$$
x=0
$$

we get

$$
y=3
$$

Therefore,

the line crosses the y-axis at

$$
(0,3)
$$

---

# Interpretation of the Parameters

For

$$
y=2x+3
$$

- slope = 2
- intercept = 3

---

# Example

For

$$
y=-3x+5
$$

- slope = -3
- intercept = 5

---

# Sign of Slope

---

## Positive Slope

$$
m>0
$$

```text
 /
/
```

As x increases, y increases.

---

## Negative Slope

$$
m<0
$$

```text
\
 \
```

As x increases, y decreases.

---

## Zero Slope

$$
m=0
$$

```text
--------
```

Horizontal line.

---

## Infinite Slope

```text
|
|
|
```

Vertical line.

Equation:

$$
x=a
$$

---

# 3. If a Point Lies on the Line

Suppose

$$
P=(x_1,y_1)
$$

lies on

$$
y=mx+c
$$

then

it must satisfy

$$
y_1=mx_1+c
$$

---

# Example

Line:

$$
y=2x+3
$$

Point:

$$
(1,5)
$$

Check:

$$
2(1)+3=5
$$

Therefore,

the point lies on the line.

---

# Example

Point:

$$
(1,7)
$$

Check:

$$
2(1)+3=5 \neq 7
$$

Therefore,

the point does not lie on the line.

---

# 4. Parallel Lines

Two lines are parallel if they have the same slope.

---

# Example

$$
L_1:y=2x+3
$$

$$
L_2:y=2x+5
$$

Both have:

$$
m=2
$$

Therefore,

they are parallel.

---

# Visual

```text
       /
      /
     /
    /
```

```text
     /
    /
   /
  /
```

Same angle with x-axis.

Never intersect.

---

# Example

These are parallel:

$$
y=5x+1
$$

$$
y=5x-7
$$

---

# Example

These are NOT parallel:

$$
y=2x+3
$$

$$
y=3x+3
$$

because

$$
2\neq3
$$

---

# Why?

Slope determines the angle of the line.

Changing only the intercept shifts the line up or down.

Changing the slope rotates the line.

---

# 5. Angle Interpretation

Suppose

the line makes an angle

$$
\theta
$$

with the positive x-axis.

Then

$$
m=\tan\theta
$$

---

# Example

If

$$
\theta=45^\circ
$$

then

$$
m=\tan45^\circ=1
$$

---

If

$$
\theta=60^\circ
$$

then

$$
m=\tan60^\circ=\sqrt3
$$

---

# Therefore

Slope is simply:

> tangent of the angle the line makes with the x-axis.

---

# 6. General Form of a Line

Another way to write a line:

$$
ax+by+c=0
$$

This is called:

# General Form

---

# Why is this useful?

Because:

1. Easier for algebra
2. Easier for machine learning
3. Easily generalizes to higher dimensions

---

# Converting to Slope-Intercept Form

Starting from

$$
ax+by+c=0
$$

Move terms:

$$
by=-ax-c
$$

Divide by \(b\):

$$
y=-\frac{a}{b}x-\frac{c}{b}
$$

Comparing with

$$
y=mx+c
$$

we get

---

# Slope

$$
m=-\frac{a}{b}
$$

---

# Intercept

$$
c_{intercept}
=
-\frac{c}{b}
$$

---

# Example

Given

$$
3x+2y-1=0
$$

Solve:

$$
2y=-3x+1
$$

$$
y=-\frac32x+\frac12
$$

Therefore,

Slope:

$$
m=-\frac32
$$

Intercept:

$$
\frac12
$$

---

# Example

Line:

$$
6x+4y=2
$$

Then

$$
4y=-6x+2
$$

$$
y=-\frac64x+\frac24
$$

$$
y=-\frac32x+\frac12
$$

Same line.

---

# Important Idea

Multiplying an equation by a constant does not change the line.

Thus,

$$
3x+2y-1=0
$$

and

$$
6x+4y=2
$$

represent the same line.

---

# Why?

Because both equations have

- same slope
- same intercept

---

# 7. Line Equation in Machine Learning

A classifier tries to learn:

$$
ax+by+c=0
$$

or

$$
w_1x_1+w_2x_2+w_0=0
$$

where

- \(w_1\) = weight of feature 1
- \(w_2\) = weight of feature 2
- \(w_0\) = bias/intercept

---

# Example

Features:

- Length
- Width

Classifier:

$$
2(length)+3(width)-5=0
$$

This line separates two classes.

---

# Prediction Rule

Compute

$$
s=w_1x_1+w_2x_2+w_0
$$

If

$$
s>0
$$

Class 1

If

$$
s<0
$$

Class 2

If

$$
s=0
$$

Point lies exactly on the boundary.

---

# 8. Generalization to 3D

A line in 2D becomes a plane in 3D.

Equation:

$$
w_1x_1+w_2x_2+w_3x_3+w_0=0
$$

---

# Example

Features:

- Length
- Width
- Weight

Classifier:

$$
2x_1+4x_2+3x_3-10=0
$$

This equation represents a plane.

---

# Generalization to d Dimensions

Suppose there are:

$$
d
$$

features.

Then

$$
w_1x_1+w_2x_2+\cdots+w_dx_d+w_0=0
$$

or

$$
\sum_{i=1}^{d}w_ix_i+w_0=0
$$

or

$$
w^Tx+b=0
$$

This is called a:

# Hyperplane

---

# Relationship Between Dimensions

| Feature Space | Decision Boundary |
|---------------|------------------|
| 2D | Line |
| 3D | Plane |
| dD | Hyperplane |

---

# Why does the boundary lose one dimension?

Because:

one equation imposes one constraint.

Hence,

```text
d dimensions
     ↓
d−1 dimensional boundary
```

---

# Machine Learning Interpretation

## 2 Features

Decision boundary:

$$
w_1x_1+w_2x_2+b=0
$$

(Line)

---

## 3 Features

Decision boundary:

$$
w_1x_1+w_2x_2+w_3x_3+b=0
$$

(Plane)

---

## d Features

Decision boundary:

$$
w^Tx+b=0
$$

(Hyperplane)

---

# The Entire Story

```text
Data Points
      ↓
Represent as vectors
      ↓
Feature Space
      ↓
Find a separating boundary
      ↓
Write its equation
      ↓
Learn parameters (weights)
      ↓
Predict using sign(wᵀx+b)
```

---

# One-Line Summary

A straight line in school mathematics:

$$
y=mx+c
$$

becomes

$$
ax+by+c=0
$$

which becomes

$$
w_1x_1+w_2x_2+w_0=0
$$

which finally becomes

$$
w^Tx+b=0
$$

and this single equation forms the foundation of:

- Perceptron
- Logistic Regression
- Linear SVM
- Neural Networks
- Linear Classification in Machine Learning.

# From Lines to Hyperplanes: The Language of Machine Learning

---

# 1. Parallel Lines

Suppose

$$
L_1:y=mx+c_1
$$

and

$$
L_2:y=mx+c_2
$$

Both have the same slope:

$$
m_1=m_2=m
$$

Therefore,

$$
L_1 \parallel L_2
$$

provided

$$
c_1 \neq c_2
$$

---

# Example

$$
L_1:y=-\frac32x+\frac14
$$

$$
L_2:y=-\frac32x+1
$$

Since

$$
m_1=m_2=-\frac32
$$

and

$$
c_1\neq c_2
$$

the lines are parallel.

---

# If both slope and intercept are same

If

$$
m_1=m_2
$$

and

$$
c_1=c_2
$$

then

the two equations represent exactly the same line.

---

# Example

$$
3x+2y-1=0
$$

and

$$
6x+4y=2
$$

represent the same line because the second equation is simply:

$$
2\times(3x+2y-1=0)
$$

---

# Important Geometric Fact

Parallel lines never intersect.

The distance between them is constant.

---

# Why Parallel Lines Matter in Machine Learning

In machine learning, many decision boundaries are represented as:

$$
w^Tx+b=0
$$

Changing only the bias term \(b\) shifts the boundary parallel to itself.

```text
w fixed
 ↓
same orientation
 ↓
parallel hyperplanes
```

This idea is heavily used in:

- Logistic Regression
- Perceptron
- SVM

---

# 2. Decision Boundaries in 2D

Suppose we have two features:

- Feature 1 → \(x_1\)
- Feature 2 → \(x_2\)

A classifier learns:

$$
M_i x + C_i
$$

or equivalently,

$$
w_1x_1+w_2x_2+w_0=0
$$

Different values of

- slope
- intercept

produce different lines.

---

# Visual Interpretation

```text
Different parameters
       ↓
Different lines
       ↓
Different decision boundaries
       ↓
Different classification errors
```

The goal of learning is:

> Find the line that separates the classes with minimum error.

---

# Why 2D is Special

In 2D:

```text
Feature Space = Plane
Decision Boundary = Line
```

because one equation removes one degree of freedom.

---

# What Happens in 3D?

Features:

$$
(x_1,x_2,x_3)
$$

Decision boundary:

$$
w_1x_1+w_2x_2+w_3x_3+w_0=0
$$

This is no longer a line.

It becomes a:

# Plane

---

# Dimensional Pattern

| Feature Space | Boundary |
|--------------|-----------|
| 2D | Line |
| 3D | Plane |
| 4D | Hyperplane |
| dD | Hyperplane |

---

# Important Observation

A line exists naturally in:

- 2D
- 3D
- 4D
- higher dimensions

but it usually cannot separate high-dimensional data.

Instead we need:

- planes
- hyperplanes

---

# General Equation in d Dimensions

Suppose there are

$$
d
$$

features:

$$
x_1,x_2,\dots,x_d
$$

Then the decision boundary is

$$
\sum_{i=1}^{d} w_i x_i + w_0 = 0
$$

which means

$$
w_1x_1+w_2x_2+\cdots+w_dx_d+w_0=0
$$

---

# What are the Parameters?

The numbers

$$
w_1,w_2,\dots,w_d
$$

are called:

# Weights

and

$$
w_0
$$

is called:

# Bias (or intercept)

---

# Vector Representation

Instead of writing:

$$
w_1x_1+w_2x_2+\cdots+w_dx_d
$$

we create two vectors.

---

# Weight Vector

$$
w=
\begin{bmatrix}
w_1\\
w_2\\
\vdots\\
w_d
\end{bmatrix}
\in \mathbb{R}^d
$$

---

# Feature Vector

$$
x=
\begin{bmatrix}
x_1\\
x_2\\
\vdots\\
x_d
\end{bmatrix}
\in \mathbb{R}^d
$$

---

# Dot Product Form

Then

$$
w^Tx
=
\begin{bmatrix}
w_1&w_2&\cdots&w_d
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2\\
\vdots\\
x_d
\end{bmatrix}
$$

which becomes

$$
w^Tx
=
w_1x_1+w_2x_2+\cdots+w_dx_d
$$

---

# Therefore

The entire decision boundary can be written compactly as

$$
\boxed{w^Tx+w_0=0}
$$

This is one of the most important equations in machine learning.

---

# Why Do We Use Matrix Form?

Because:

1. Cleaner notation
2. Faster computations
3. Efficient implementation using NumPy
4. Easily generalizes to millions of features

---

# Geometric Meaning of \(w\)

The vector

$$
w
$$

is perpendicular (normal) to the decision boundary.

---

# Why?

For any point on the boundary,

$$
w^Tx+w_0=0
$$

All points satisfy the same dot product condition.

Therefore,

\(w\) determines the orientation of the boundary.

---

# Geometric Interpretation

```text
Change w
   ↓
Rotate boundary

Change w0
   ↓
Shift boundary parallel to itself
```

This is an extremely important idea.

---

# Example

Boundary:

$$
2x_1+3x_2+4=0
$$

Weight vector:

$$
w=
\begin{bmatrix}
2\\
3
\end{bmatrix}
$$

Bias:

$$
w_0=4
$$

---

# Classification Rule

Compute

$$
s=w^Tx+w_0
$$

If

$$
s>0
$$

predict Class 1.

If

$$
s<0
$$

predict Class 2.

If

$$
s=0
$$

the point lies exactly on the decision boundary.

---

# Why This Equation Appears Everywhere

The following algorithms all start from

$$
w^Tx+b
$$

---

## Perceptron

$$
\hat y=\text{sign}(w^Tx+b)
$$

---

## Logistic Regression

$$
P(y=1|x)=\sigma(w^Tx+b)
$$

---

## Support Vector Machine

$$
w^Tx+b=0
$$

---

## Neural Networks

First layer:

$$
z=w^Tx+b
$$

then apply an activation function.

---

# Big Picture

```text
Equation of line
      ↓
y=mx+c
      ↓
General form
ax+by+c=0
      ↓
Higher dimensions
w1x1+w2x2+...+wdxd+w0=0
      ↓
Vector notation
wᵀx+b=0
      ↓
Decision boundary
      ↓
Machine Learning
```

---

# One Sentence Summary

Machine learning classification is fundamentally about learning the best

$$
w
\quad\text{and}\quad
b
$$

so that the hyperplane

$$
w^Tx+b=0
$$

separates the data as accurately as possible(minimises the misclassification error).

# Data Points as Vectors

---

# 1. A Point in 2D Space

Suppose we have two features:

- Feature 1 : \(f_1\)
- Feature 2 : \(f_2\)

A data point is

$$
x=
\begin{bmatrix}
2\\
3
\end{bmatrix}
$$

or simply

$$
x=(2,3)
$$

---

# Geometric Interpretation

The point

$$
(2,3)
$$

means:

- move 2 units along \(f_1\)
- move 3 units along \(f_2\)

and stop.

---

# In Machine Learning

The point

$$
(2,3)
$$

can represent:

- Age = 2
- Salary = 3

or

- Height = 2
- Weight = 3

or any two features.

The axes are features.

The point is one observation.

---

# Physics vs Machine Learning

In Physics:

The arrow from origin to the point represents

- displacement
- magnitude
- direction

---

# In Machine Learning:

The same arrow represents

# One data point.

The geometric properties are still useful.

---

# Why Do We Treat Data Points as Vectors?

Because vectors allow us to measure:

- distance
- similarity
- angle
- projection

These ideas power almost all ML algorithms.

---

# Example

Point:

$$
x=
\begin{bmatrix}
2\\
3
\end{bmatrix}
$$

Weight vector:

$$
w=
\begin{bmatrix}
-1\\
1
\end{bmatrix}
$$

Bias:

$$
w_0=10
$$

Decision function:

$$
w^Tx+w_0
$$

becomes

$$
[-1\;\;1]
\begin{bmatrix}
2\\
3
\end{bmatrix}
+10
$$

which is

$$
=-2+3+10
$$

$$
=11
$$

---

# Magnitude of a Vector

For

$$
x=
\begin{bmatrix}
2\\
3
\end{bmatrix}
$$

the length is

$$
||x||
=
\sqrt{2^2+3^2}
$$

$$
=
\sqrt{13}
$$

---

# General Formula

For

$$
x=
\begin{bmatrix}
x_1\\
x_2
\end{bmatrix}
$$

the magnitude is

$$
||x||
=
\sqrt{x_1^2+x_2^2}
$$

---

# In d Dimensions

For

$$
x=
\begin{bmatrix}
x_1\\
x_2\\
\vdots\\
x_d
\end{bmatrix}
$$

the Euclidean norm is

$$
||x||
=
\sqrt{
x_1^2+x_2^2+\cdots+x_d^2
}
$$

---

# Why is this Important?

Distance calculations use vector norms.

Examples:

- KNN
- K-Means
- Clustering
- PCA
- Recommendation Systems

all rely heavily on vector lengths and distances.

---

# Notation

We usually write

$$
x\in\mathbb R^d
$$

which means

- \(x\) is a vector
- it has \(d\) components
- each component is a real number.

---

# Dataset Representation

A dataset is a collection of points:

$$
D=
\{
(x_i,y_i)
\}_{i=1}^{n}
$$

---

# Meaning of Symbols

## Number of samples

$$
n
$$

## Number of features

$$
d
$$

---

# Each Data Point

$$
x_i\in\mathbb R^d
$$

means:

the \(i^{th}\) observation has

\(d\) features.

---

# Labels

For classification:

$$
y_i\in\{1,2,\dots,K\}
$$

For binary classification:

$$
y_i\in\{0,1\}
$$

or

$$
y_i\in\{-1,+1\}
$$

---

# Dataset in Plain English

The notation

$$
D=
\{
(x_i,y_i)
\}_{i=1}^{n}
$$

simply means:

```text
Dataset D consists of:

(point1,label1)
(point2,label2)
(point3,label3)
...
(pointn,labeln)
```

---

# Example Dataset

| Height | Weight | Class |
|--------|---------|--------|
| 170 | 70 | 1 |
| 160 | 55 | 0 |
| 180 | 85 | 1 |

Then

$$
x_1=
\begin{bmatrix}
170\\
70
\end{bmatrix},
\qquad
y_1=1
$$

---

# Matrix Representation of Dataset

Instead of storing points separately, we stack them.

If there are

- \(n\) samples
- \(d\) features

then

$$
X
\in
\mathbb R^{n\times d}
$$

---

# Dataset Matrix

$$
X=
\begin{bmatrix}
- & x_1^T & -\\
- & x_2^T & -\\
& \vdots & \\
- & x_n^T & -
\end{bmatrix}
$$

Each row is one data point.

---

# Example

$$
X=
\begin{bmatrix}
170 & 70\\
160 & 55\\
180 & 85
\end{bmatrix}
$$

Shape:

$$
3\times2
$$

meaning:

- 3 observations
- 2 features

---

# Big Picture

```text
Point in geometry
        ↓
Vector
        ↓
Feature vector
        ↓
Collection of vectors
        ↓
Dataset matrix X
        ↓
Machine Learning algorithms
```

---

# Most Important Idea

Machine learning can be viewed as:

```text
Data
      ↓
Points in high dimensions
      ↓
Vectors
      ↓
Linear algebra operations
      ↓
Predictions
```

This is why Linear Algebra is the language of Machine Learning.
