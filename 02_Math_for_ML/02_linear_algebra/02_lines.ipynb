# From Lines to Hyperplanes: The Language of Machine Learning

---

# 1. Parallel Lines

Suppose

$$
L_1:y=mx+c_1
$$

and

$$
L_2:y=mx+c_2
$$

Both have the same slope:

$$
m_1=m_2=m
$$

Therefore,

$$
L_1 \parallel L_2
$$

provided

$$
c_1 \neq c_2
$$

---

# Example

$$
L_1:y=-\frac32x+\frac14
$$

$$
L_2:y=-\frac32x+1
$$

Since

$$
m_1=m_2=-\frac32
$$

and

$$
c_1\neq c_2
$$

the lines are parallel.

---

# If both slope and intercept are same

If

$$
m_1=m_2
$$

and

$$
c_1=c_2
$$

then

the two equations represent exactly the same line.

---

# Example

$$
3x+2y-1=0
$$

and

$$
6x+4y=2
$$

represent the same line because the second equation is simply:

$$
2\times(3x+2y-1=0)
$$

---

# Important Geometric Fact

Parallel lines never intersect.

The distance between them is constant.

---

# Why Parallel Lines Matter in Machine Learning

In machine learning, many decision boundaries are represented as:

$$
w^Tx+b=0
$$

Changing only the bias term \(b\) shifts the boundary parallel to itself.

```text
w fixed
 ↓
same orientation
 ↓
parallel hyperplanes
```

This idea is heavily used in:

- Logistic Regression
- Perceptron
- SVM

---

# 2. Decision Boundaries in 2D

Suppose we have two features:

- Feature 1 → \(x_1\)
- Feature 2 → \(x_2\)

A classifier learns:

$$
M_i x + C_i
$$

or equivalently,

$$
w_1x_1+w_2x_2+w_0=0
$$

Different values of

- slope
- intercept

produce different lines.

---

# Visual Interpretation

```text
Different parameters
       ↓
Different lines
       ↓
Different decision boundaries
       ↓
Different classification errors
```

The goal of learning is:

> Find the line that separates the classes with minimum error.

---

# Why 2D is Special

In 2D:

```text
Feature Space = Plane
Decision Boundary = Line
```

because one equation removes one degree of freedom.

---

# What Happens in 3D?

Features:

$$
(x_1,x_2,x_3)
$$

Decision boundary:

$$
w_1x_1+w_2x_2+w_3x_3+w_0=0
$$

This is no longer a line.

It becomes a:

# Plane

---

# Dimensional Pattern

| Feature Space | Boundary |
|--------------|-----------|
| 2D | Line |
| 3D | Plane |
| 4D | Hyperplane |
| dD | Hyperplane |

---

# Important Observation

A line exists naturally in:

- 2D
- 3D
- 4D
- higher dimensions

but it usually cannot separate high-dimensional data.

Instead we need:

- planes
- hyperplanes

---

# General Equation in d Dimensions

Suppose there are

$$
d
$$

features:

$$
x_1,x_2,\dots,x_d
$$

Then the decision boundary is

$$
\sum_{i=1}^{d} w_i x_i + w_0 = 0
$$

which means

$$
w_1x_1+w_2x_2+\cdots+w_dx_d+w_0=0
$$

---

# What are the Parameters?

The numbers

$$
w_1,w_2,\dots,w_d
$$

are called:

# Weights

and

$$
w_0
$$

is called:

# Bias (or intercept)

---

# Vector Representation

Instead of writing:

$$
w_1x_1+w_2x_2+\cdots+w_dx_d
$$

we create two vectors.

---

# Weight Vector

$$
w=
\begin{bmatrix}
w_1\\
w_2\\
\vdots\\
w_d
\end{bmatrix}
\in \mathbb{R}^d
$$

---

# Feature Vector

$$
x=
\begin{bmatrix}
x_1\\
x_2\\
\vdots\\
x_d
\end{bmatrix}
\in \mathbb{R}^d
$$

---

# Dot Product Form

Then

$$
w^Tx
=
\begin{bmatrix}
w_1&w_2&\cdots&w_d
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2\\
\vdots\\
x_d
\end{bmatrix}
$$

which becomes

$$
w^Tx
=
w_1x_1+w_2x_2+\cdots+w_dx_d
$$

---

# Therefore

The entire decision boundary can be written compactly as

$$
\boxed{w^Tx+w_0=0}
$$

This is one of the most important equations in machine learning.

---

# Why Do We Use Matrix Form?

Because:

1. Cleaner notation
2. Faster computations
3. Efficient implementation using NumPy
4. Easily generalizes to millions of features

---

# Geometric Meaning of \(w\)

The vector

$$
w
$$

is perpendicular (normal) to the decision boundary.

---

# Why?

For any point on the boundary,

$$
w^Tx+w_0=0
$$

All points satisfy the same dot product condition.

Therefore,

\(w\) determines the orientation of the boundary.

---

# Geometric Interpretation

```text
Change w
   ↓
Rotate boundary

Change w0
   ↓
Shift boundary parallel to itself
```

This is an extremely important idea.

---

# Example

Boundary:

$$
2x_1+3x_2+4=0
$$

Weight vector:

$$
w=
\begin{bmatrix}
2\\
3
\end{bmatrix}
$$

Bias:

$$
w_0=4
$$

---

# Classification Rule

Compute

$$
s=w^Tx+w_0
$$

If

$$
s>0
$$

predict Class 1.

If

$$
s<0
$$

predict Class 2.

If

$$
s=0
$$

the point lies exactly on the decision boundary.

---

# Why This Equation Appears Everywhere

The following algorithms all start from

$$
w^Tx+b
$$

---

## Perceptron

$$
\hat y=\text{sign}(w^Tx+b)
$$

---

## Logistic Regression

$$
P(y=1|x)=\sigma(w^Tx+b)
$$

---

## Support Vector Machine

$$
w^Tx+b=0
$$

---

## Neural Networks

First layer:

$$
z=w^Tx+b
$$

then apply an activation function.

---

# Big Picture

```text
Equation of line
      ↓
y=mx+c
      ↓
General form
ax+by+c=0
      ↓
Higher dimensions
w1x1+w2x2+...+wdxd+w0=0
      ↓
Vector notation
wᵀx+b=0
      ↓
Decision boundary
      ↓
Machine Learning
```

---

# One Sentence Summary

Machine learning classification is fundamentally about learning the best

$$
w
\quad\text{and}\quad
b
$$

so that the hyperplane

$$
w^Tx+b=0
$$

separates the data as accurately as possible(minimises the misclassification error).

# Dot Product

Suppose

$$
x_1=
\begin{bmatrix}
2\\
3
\end{bmatrix}
,\qquad
x_2=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

Then

$$
x_1^Tx_2
=
[2\;\;3]
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

$$
=
2\times3+3\times4
$$

$$
=
18
$$

---

# General Formula

For

$$
x_1,x_2\in \mathbb R^d
$$

the dot product is

$$
x_1^Tx_2
=
\sum_{i=1}^{d}
x_{1i}x_{2i}
$$

---

# Geometric Meaning of Dot Product

The dot product can also be written as

$$
x_1^Tx_2
=
||x_1||
\,||x_2||
\cos\theta
$$

where

- \(||x_1||\) = length of first vector
- \(||x_2||\) = length of second vector
- \(\theta\) = angle between them.

This formula is extremely important in Machine Learning.

---

# Why Dot Product is Important

Dot product measures:

- alignment
- similarity
- angle between vectors.

---

# Cosine Similarity

From

$$
x_1^Tx_2
=
||x_1||
||x_2||
\cos\theta
$$

we get

$$
\cos\theta
=
\frac{x_1^Tx_2}
{||x_1||\,||x_2||}
$$

This is called:

# Cosine Similarity

---

# Interpretation

### Same direction

$$
\theta=0^\circ
$$

$$
\cos\theta=1
$$

Maximum similarity.

---

### Perpendicular

$$
\theta=90^\circ
$$

$$
\cos\theta=0
$$

No correlation.

---

### Opposite directions

$$
\theta=180^\circ
$$

$$
\cos\theta=-1
$$

Completely opposite.

---

# Why Cosine Similarity is Used in ML

Used in:

- Search engines
- Recommendation systems
- NLP
- Sentence embeddings
- Document similarity
- Information retrieval

---

# Example

Suppose

$$

\end{bmatrix}
,

\end{bmatrix}

$$

$$

=
\sqrt{13}
$$

$$
||x_2||
=
5
$$

Therefore,

$$
\cos\theta
=
\frac{18}
{\sqrt{13}\times5}
\approx0.998
$$

Hence the vectors are almost pointing in the same direction.

# Hyperplane

A hyperplane is one dimension lower than the ambient space.

| Space | Hyperplane |
|-------|------------|
| 2D | line |
| 3D | plane |
| dD | (d-1)-dimensional hyperplane |

---

# Hyperplane Through Origin

If

$$
w_0=0
$$

then

$$
w^Tx=0
$$

This hyperplane passes through the origin.

---

# Why?

Because

$$
x=0
$$

satisfies

$$
w^T0=0
$$

Therefore,

the origin lies on the hyperplane.

---

# Extremely Important Geometric Fact

For every point on

$$
w^Tx=0
$$

we have

$$
w^Tx
=
||w||
||x||
\cos\theta
=
0
$$

Since

$$
||w||\neq0
$$

and

$$
||x||\neq0
$$

we must have

$$
\cos\theta=0
$$

which means

$$
\theta=90^\circ
$$

Therefore,

# \(w\) is perpendicular to every vector \(x\) lying on the hyperplane.

or

$$
w\perp x
\qquad
\forall x \in \{x:w^Tx=0\}
$$

---

# This is why \(w\) is called the Normal Vector.

It is normal (perpendicular) to the entire hyperplane.

---

# Visual Intuition

```text
              w
              ↑
              |
              |
--------------+-------------- Hyperplane
            x →
```

Every vector on the hyperplane is perpendicular to \(w\).

---

# What Happens When Bias Exists?

The equation becomes

$$
w^Tx+w_0=0
$$

The hyperplane shifts.

It no longer has to pass through the origin.

But

# \(w\) is STILL perpendicular to the hyperplane.

The bias only translates the hyperplane.

It does not rotate it.

---

# Classification Interpretation

For any point \(x\),

If

$$
w^Tx+w_0>0
$$

then

\(x\) lies on one side.

If

$$
w^Tx+w_0<0
$$

then

\(x\) lies on the other side.

If

$$
w^Tx+w_0=0
$$

then

\(x\) lies exactly on the decision boundary.

# Geometric Interpretation of a Hyperplane

Consider the hyperplane

$$
\Pi_d=\{x\in\mathbb R^d : w^Tx+w_0=0\}
$$

where

$$
w=
\begin{bmatrix}
w_1\\
w_2\\
\vdots\\
w_d
\end{bmatrix}
\neq 0
$$

is called the

- weight vector
- normal vector
- perpendicular vector to the hyperplane.

---

# Important Observation

The vector

$$
w
$$

is perpendicular to the hyperplane:

$$
w\perp \Pi_d
$$

This is why \(w\) is called the **normal vector**.

---

# Why is \(w\) perpendicular?

Take any point

$$
x
$$

on the hyperplane.

Then

$$
w^Tx+w_0=0
$$

If the hyperplane passes through the origin:

$$
w_0=0
$$

then

$$
w^Tx=0
$$

Using the dot product formula,

$$
w^Tx
=
||w||\,||x||\cos\theta
$$

Therefore,

$$
||w||\,||x||\cos\theta=0
$$

Since

$$
||w||\neq 0
$$

and

$$
||x||\neq 0
$$

we get

$$
\cos\theta=0
$$

which implies

$$
\theta=90^\circ
$$

Hence,

$$
w\perp x
$$

for every point \(x\) on the hyperplane.

---

# Very Important Statement

When

$$
w^Tx=0
$$

it does **NOT** mean

> \(w\) is perpendicular to every vector in \(\mathbb R^d\).

It means:

> \(w\) is perpendicular only to those vectors \(x\) that lie on the hyperplane.

---

# Example

Suppose

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix}
$$

Then

$$
w^Tx=0
$$

becomes

$$
x_1+x_2=0
$$

or

$$
x_2=-x_1
$$

which is the line

```text
      /
     /
----/-------
   /
  /
```

Every point on this line satisfies

$$
w^Tx=0
$$

Therefore,

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix}
$$

is perpendicular to this line.

---

# But \(x\) Need Not Be Perpendicular to \(w\)

Consider

$$
x=
\begin{bmatrix}
1\\
2
\end{bmatrix}
$$

Then

$$
w^Tx
=
[1\;\;1]
\begin{bmatrix}
1\\
2
\end{bmatrix}
=
3
\neq 0
$$

Therefore,

$$
x
\not\perp w
$$

because \(x\) does not lie on the hyperplane.

---

# Hyperplane Through the Origin

If

$$
w_0=0
$$

then

$$
\Pi_d=\{x:w^Tx=0\}
$$

passes through the origin.

Why?

Because

$$
x=0
$$

satisfies

$$
w^T0=0
$$

Therefore,

$$
0\in \Pi_d
$$

---

# Hyperplane Not Passing Through the Origin

If

$$
w_0\neq0
$$

then

$$
w^Tx+w_0=0
$$

does not necessarily contain the origin.

The bias term

$$
w_0
$$

simply shifts the hyperplane.

It changes its position but does not change its orientation.

The vector \(w\) remains perpendicular to the hyperplane.

---

# Geometric Meaning of \(w_0\)

The bias term determines how far the hyperplane is shifted from the origin.

```text
                Hyperplane
       ---------------------------
                     |
                     |
                     | distance
                     |
                     O (origin)
```

---

# Distance from Origin to Hyperplane

For

$$
w^Tx+w_0=0
$$

the perpendicular distance from the origin is

$$
d(O,\Pi_d)
=
\frac{|w_0|}{||w||}
$$

---

# Why?

Let

$$
a
$$

be the shortest distance from the origin to the hyperplane.

Since the shortest distance is always perpendicular to the hyperplane,

its direction must be along \(w\).

Therefore,

$$
a=\frac{|w_0|}{||w||}
$$

---

# Interpretation

---

## If

$$
w_0=0
$$

then

$$
d(O,\Pi_d)=0
$$

The hyperplane passes through the origin.

---

## If

$$
|w_0|
$$

increases,

the hyperplane moves farther away from the origin.

---

## If

$$
||w||
$$

increases,

the distance decreases.

---

# Signed Distance

Sometimes we use the signed distance:

$$
\frac{w_0}{||w||}
$$

instead of

$$
\frac{|w_0|}{||w||}
$$

because the sign tells us on which side of the hyperplane the origin lies.

---

# Distance of Any Point from a Hyperplane

Suppose

$$
x
$$

is any point in \(\mathbb R^d\).

The perpendicular distance from \(x\) to the hyperplane

$$
w^Tx+w_0=0
$$

is

$$
d(x,\Pi_d)
=

$$

---

# Signed Distance of a Point

The signed distance is

$$
\frac{w^Tx+w_0}{||w||}
$$

---

# Interpretation

---

### Positive

$$
w^Tx+w_0>0
$$

Point lies on one side.

---

### Negative

$$
w^Tx+w_0<0
$$

Point lies on the other side.

---

### Zero

$$
w^Tx+w_0=0
$$

Point lies exactly on the hyperplane.

---

# Why This Formula Appears Everywhere in ML

The quantity

$$
w^Tx+w_0
$$

appears in

- Perceptron
- Logistic Regression
- Linear SVM
- Neural Networks

because it tells us

1. which side of the boundary the point lies on,
2. how far the point is from the boundary,
3. confidence of classification.

---

# SVM Connection

Support Vector Machines maximize

$$

$$

- large \(||w||\) → smaller margin
- small \(||w||\) → larger margin.

# Hyperplanes, Normal Vectors, Distances and Their Geometric Meaning

---

# 1. What is a Hyperplane?

A hyperplane in \( \mathbb{R}^d \) is defined as

$$
\Pi_d=\{x\in\mathbb{R}^d : w^Tx+w_0=0\}
$$

where

$$
w=
\begin{bmatrix}
w_1\\
w_2\\
\vdots\\
w_d
\end{bmatrix}
\neq 0
$$

and

$$
w_0\in\mathbb{R}.
$$

---

# Examples

### In \( \mathbb{R}^1 \)

A hyperplane is just a point.

Example:

$$
x=3
$$

---

### In \( \mathbb{R}^2 \)

A hyperplane is a line.

Example:

$$
2x_1+3x_2-6=0
$$

---

### In \( \mathbb{R}^3 \)

A hyperplane is an ordinary plane.

Example:

$$
x+y+z-1=0
$$

---

### In \( \mathbb{R}^{100} \)

A hyperplane is a 99-dimensional object.

We cannot visualize it, but all the mathematics remains exactly the same.

---

# Dimension of a Hyperplane

A hyperplane in \( \mathbb R^d \) always has dimension

$$
d-1.
$$

Examples:

| Space | Hyperplane | Dimension |
|-------|-------------|------------|
| \(R^1\) | Point | 0 |
| \(R^2\) | Line | 1 |
| \(R^3\) | Plane | 2 |
| \(R^d\) | Hyperplane | \(d-1\) |

---

# 2. The Vector \(w\)

The vector \(w\) is called:

- Weight vector
- Normal vector
- Perpendicular vector
- Orthogonal vector to the hyperplane.

---

# Why is \(w\) called a Normal Vector?

Because

$$
w\perp \Pi_d
$$

which means

> \(w\) is perpendicular to every direction lying inside the hyperplane.

---

# Hyperplane Through Origin

If

$$
w_0=0
$$

then

$$
\Pi_d=\{x:w^Tx=0\}
$$

passes through the origin.

Why?

Because

$$
x=0
$$

satisfies

$$
w^T0=0.
$$

Therefore

$$
0\in \Pi_d.
$$

---

# Hyperplane Not Passing Through Origin

If

$$
w_0\neq 0
$$

then

$$
w^Tx+w_0=0
$$

usually does not contain the origin.

The bias term \(w_0\) simply shifts the hyperplane.

It changes the position of the hyperplane but not its orientation.

---

# 3. Why is \(w\) Perpendicular to the Hyperplane?

Take any point

$$
x
$$

on the hyperplane passing through the origin.

Then

$$
w^Tx=0
$$

Using the dot product formula,

$$
w^Tx
=
||w||\,||x||\cos\theta
$$

therefore,

$$
||w||\,||x||\cos\theta=0.
$$

Since

$$
||w||\neq 0
$$

and

$$
||x||\neq 0
$$

we obtain

$$
\cos\theta=0
$$

which gives

$$
\theta=90^\circ.
$$

Hence

$$
w\perp x.
$$

---

# Important Clarification

The statement

$$
w^Tx=0
$$

does NOT mean

> \(w\) is perpendicular to every vector in \(R^d\).

It only means

> \(w\) is perpendicular to vectors lying on the hyperplane.

---

# Example

Suppose

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix}
$$

Then

$$
w^Tx=0
$$

becomes

$$
x_1+x_2=0
$$

or

$$
x_2=-x_1.
$$

This is a line.

Every point on this line is perpendicular to

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix}.
$$

---

# But \(x\) Need Not Be Perpendicular to \(w\)

Take

$$
x=
\begin{bmatrix}
1\\
2
\end{bmatrix}
$$

Then

$$
w^Tx
=
[1\;1]
\begin{bmatrix}
1\\
2
\end{bmatrix}
=
3
\neq0.
$$

Hence

$$
x
\not\perp w.
$$

because \(x\) does not lie on the hyperplane.

---

# 4. Difference of Two Points on a Hyperplane

Suppose

$$
x_1,x_2\in\Pi_d.
$$

Then

$$
w^Tx_1+w_0=0
$$

and

$$
w^Tx_2+w_0=0.
$$

Subtracting,

$$
w^Tx_2-w^Tx_1=0
$$

gives

$$
w^T(x_2-x_1)=0.
$$

Therefore

$$
x_2-x_1\perp w.
$$

---

# Important Geometric Meaning

The vector

$$
x_2-x_1
$$

is a direction vector lying inside the hyperplane.

Therefore,

> Every direction inside the hyperplane is perpendicular to \(w\).

---

# Why \(w\) Need Not Be Perpendicular to a Point?

A point is just a location.

Perpendicularity is defined between vectors.

To talk about orthogonality, we first convert points into vectors.

The correct statement is:

> \(w\) is perpendicular to every direction vector inside the hyperplane.

NOT

> \(w\) is perpendicular to every point on the hyperplane.

---

# Example

Suppose

$$
P=
\begin{bmatrix}
1\\
2
\end{bmatrix},
\qquad
Q=
\begin{bmatrix}
3\\
4
\end{bmatrix}.
$$

Then

$$
\overrightarrow{PQ}
=
Q-P
=
\begin{bmatrix}
2\\
2
\end{bmatrix}.
$$

The subtraction of two points gives a vector.

Points themselves are not vectors.

---

# 5. Unit Normal Vector

Sometimes we only care about direction and not length.

So we normalize \(w\).

The unit normal vector is

$$
\hat w
=
\frac{w}{||w||}.
$$

Properties:

$$
||\hat w||=1
$$

and

$$
\hat w
$$

points in exactly the same direction as \(w\).

---

# Why Use Unit Vectors?

Because they simplify geometric formulas.

Distance formulas become cleaner.

Many ML algorithms use the unit normal vector implicitly.

---

# 10. Half-Spaces Induced by a Hyperplane

Consider the hyperplane

$$
\Pi_d=\{x\in\mathbb R^d : w^Tx+w_0=0\}.
$$

A hyperplane divides the entire space into **two regions** called **half-spaces**.

```text
             Positive Half-space
                    (+)
                     |
                     |
---------------------|--------------------  Hyperplane
                     |
                     |
             Negative Half-space
                    (-)
```

The hyperplane itself is the boundary between these two regions.

---

# Positive Half-Space

The set of all points satisfying

$$
w^Tx+w_0>0
$$

is called the positive half-space:

$$
H^+
=
\{x\in\mathbb R^d :
w^Tx+w_0>0
\}.
$$

---

# Negative Half-Space

The set of all points satisfying

$$
w^Tx+w_0<0
$$

is called the negative half-space:

$$
H^-
=
\{x\in\mathbb R^d :
w^Tx+w_0<0
\}.
$$

---

# Hyperplane Itself

The hyperplane consists of points satisfying

$$
w^Tx+w_0=0.
$$

Thus,

$$
\mathbb R^d
=
H^+
\cup
\Pi_d
\cup
H^-.
$$

---

# Important Geometric Interpretation

The quantity

$$
w^Tx+w_0
$$

does two things simultaneously:

1. Tells us on which side of the hyperplane the point lies.
2. Tells us how far the point is from the hyperplane (up to scaling).

This quantity is sometimes called:

- score
- decision function
- activation
- discriminant function.

---

# Which Side Does a Point Lie On?

Suppose

$$
f(x)=w^Tx+w_0.
$$

Then:

### Case 1

$$
f(x)>0
$$

Point lies in the positive half-space.

---

### Case 2

$$
f(x)<0
$$

Point lies in the negative half-space.

---

### Case 3

$$
f(x)=0
$$

Point lies exactly on the hyperplane.

---

# Example

Consider

$$
x_1+x_2-2=0.
$$

Here,

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix},
\qquad
w_0=-2.
$$

---

## Point A

$$
A=(3,2)
$$

Compute:

$$
f(A)=3+2-2=3>0
$$

Therefore,

$$
A\in H^+.
$$

---

## Point B

$$
B=(0,0)
$$

Compute:

$$
f(B)=0+0-2=-2<0
$$

Therefore,

$$
B\in H^-.
$$

---

## Point C

$$
C=(1,1)
$$

Compute:

$$
f(C)=1+1-2=0
$$

Therefore,

$$
C
\in
\Pi_d.
$$

---

# Why Does the Sign Change?

Because the hyperplane acts like a boundary.

Crossing the boundary changes the sign of

$$
w^Tx+w_0.
$$

This is exactly what happens in classification.

```text
Negative Side      Hyperplane       Positive Side

      -                  0                  +
```

---

# Signed Distance

Earlier we derived the distance:

$$
d(x,\Pi_d)
=
\frac{|w^Tx+w_0|}{||w||}.
$$

If we remove the absolute value:

$$
\boxed{
\frac{w^Tx+w_0}{||w||}
}
$$

we obtain the **signed distance**.

---

# Interpretation

### Positive

Point lies in positive half-space.

### Negative

Point lies in negative half-space.

### Zero

Point lies on the hyperplane.

---

# Why Is It Called Signed Distance?

Because:

- magnitude gives distance
- sign gives direction.

---

# Geometric Picture

```text
              Positive Side

                    x1
                    *
                   /
                  /
                 /
---------------Hyperplane----------------
                \
                 \
                  \
                   *
                   x2

              Negative Side
```

If

$$
x_1
$$

lies above the hyperplane,

then

$$
w^Tx_1+w_0>0.
$$

If

$$
x_2
$$

lies below the hyperplane,

then

$$
w^Tx_2+w_0<0.
$$

---

# Why Does the Normal Vector Determine Positive and Negative Sides?

The vector

$$
w
$$

points toward the positive half-space.

By convention,

the direction of

$$
w
$$

defines the positive side.

Reversing

$$
w
$$

changes the labels of the half-spaces.

---

# Example

Suppose

$$
w=
\begin{bmatrix}
1\\
0
\end{bmatrix},
\qquad
w_0=-2.
$$

The hyperplane is

$$
x_1-2=0.
$$

or

$$
x_1=2.
$$

Positive half-space:

$$
x_1>2.
$$

Negative half-space:

$$
x_1<2.
$$

---

# Machine Learning Interpretation

This idea is the foundation of linear classification.

=

$$

$$

=

$$
