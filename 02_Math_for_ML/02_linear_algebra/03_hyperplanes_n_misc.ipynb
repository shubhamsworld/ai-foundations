# 13. Hyperplanes from a Geometric Perspective

A hyperplane in $\mathbb R^d$ is

$$
\Pi_d:\; w^Tx+w_0=0
$$

To completely specify a hyperplane, we need only two things:

1. A direction perpendicular to the hyperplane.
2. The distance of the hyperplane from the origin.

---

# 1. Normal Vector

The vector

$$
w
$$

is perpendicular to the hyperplane:

$$
\boxed{
w \perp \Pi_d
}
$$

Every vector lying on the hyperplane is orthogonal to $w$.

---

# Why is the Normal Unique?

The direction perpendicular to the hyperplane is unique up to scaling.

If

$$
w
$$

is normal to the hyperplane, then

$$
\alpha w
$$

for any nonzero scalar $\alpha$ is also normal.

Therefore,

$$
w
\quad\text{and}\quad
\alpha w
$$

represent exactly the same direction.

---

# Unit Normal Vector

Often we normalize:

$$
\boxed{
\hat w=
\frac{w}{\|w\|}
}
$$

This is called the **unit normal vector**.

---

# 2. Distance from the Origin

The perpendicular distance from the origin to

$$
\Pi_d:\;w^Tx+w_0=0
$$

is

$$
\boxed{
d(0,\Pi_d)
=
\frac{|w_0|}{\|w\|}
}
$$

---

# Signed Distance

The signed distance is

$$
\boxed{
d_s(0,\Pi_d)
=
-\frac{w_0}{\|w\|}
}
$$

The sign indicates on which side of the hyperplane the origin lies.

---

# Geometric Interpretation

Thus a hyperplane can be completely represented by:

- normal direction:

$$
\hat w
$$

- signed distance:

$$
-\frac{w_0}{\|w\|}
$$

---

# Hyperplane Representation

Therefore,

$$
\boxed{
\Pi_d
\equiv
\left(
\hat w,
-\frac{w_0}{\|w\|}
\right)
}
$$

---

# Important Observation

The parameters

$$
(w,w_0)
$$

are **not unique**.

If we multiply both by any nonzero scalar $\alpha$,

$$
(\alpha w,\alpha w_0)
$$

we obtain the same hyperplane:

$$
(\alpha w)^Tx+\alpha w_0
=
\alpha(w^Tx+w_0)
=
0.
$$

Hence,

$$
(w,w_0)
\sim
(\alpha w,\alpha w_0)
$$

represent the same hyperplane.

---

# Why Normalize?

Because only

$$
\frac{w_0}{\|w\|}
$$

has geometric meaning.

The individual values of

$$
w
\quad\text{and}\quad
w_0
$$

do not.

---

# Example

Consider

$$
x_1+x_2-2=0.
$$

Then

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix},
\qquad
w_0=-2.
$$

Distance from origin:

$$
d=
\frac{|-2|}{\sqrt2}
=
\sqrt2.
$$

Unit normal:

$$
\hat w=
\frac1{\sqrt2}
\begin{bmatrix}
1\\
1
\end{bmatrix}.
$$

---

# Hyperplanes Divide Space

Every hyperplane partitions space into:

---

## Positive Half-Space

$$
\boxed{
H^+
=
\{x:w^Tx+w_0>0\}
}
$$

---

## Hyperplane

$$
\boxed{
\Pi_d
=
\{x:w^Tx+w_0=0\}
}
$$

---

## Negative Half-Space

$$
\boxed{
H^-
=
\{x:w^Tx+w_0<0\}
}
$$

---

# Classification Interpretation

For binary classification:

$$
f(x)=w^Tx+w_0
$$

If

$$
f(x)>0
$$

predict:

$$
+1
$$

If

$$
f(x)<0
$$

predict:

$$
-1
$$

If

$$
f(x)=0
$$

the point lies on the decision boundary.

---

# 14. Distance Between Parallel Hyperplanes

Suppose

$$
\Pi_1:\; w^Tx+w_1=0
$$

and

$$
\Pi_2:\; w^Tx+w_2=0.
$$

Since both have the same normal vector,

$$
\boxed{
\Pi_1 \parallel \Pi_2
}
$$

---

# Distance Formula

The perpendicular distance between them is

$$
\boxed{
d(\Pi_1,\Pi_2)
=
\frac{|w_2-w_1|}
{\|w\|}
}
$$

---

# Derivation

Distance of

$$
\Pi_1
$$

from origin:

$$
d_1=
\frac{|w_1|}
{\|w\|}
$$

Distance of

$$
\Pi_2
$$

from origin:

$$
d_2=
\frac{|w_2|}
{\|w\|}
$$

Therefore,

$$
d=
|d_2-d_1|
=
\frac{|w_2-w_1|}
{\|w\|}
$$

---

# Geometric Meaning

Changing only the bias term shifts the hyperplane along the normal direction.

Changing

$$
w
$$

changes the orientation.

Changing

$$
w_0
$$

moves the hyperplane parallel to itself.

---

# Example

Hyperplanes:

$$
x_1+x_2-1=0
$$

and

$$
x_1+x_2+3=0
$$

Then

$$
w=
\begin{bmatrix}
1\\
1
\end{bmatrix}
$$

and

$$
\|w\|=\sqrt2.
$$

Distance:

$$
d=
\frac{|3-(-1)|}
{\sqrt2}
=
2\sqrt2.
$$

---

# Connection to SVM

The two margin hyperplanes in SVM are

$$
w^Tx+w_0=1
$$

and

$$
w^Tx+w_0=-1.
$$

Rewriting,

$$
w^Tx+w_0-1=0
$$

and

$$
w^Tx+w_0+1=0.
$$

Thus,

$$
w_1=w_0-1,
\qquad
w_2=w_0+1.
$$

Distance:

$$
\boxed{
\text{Margin}
=
\frac{|(w_0+1)-(w_0-1)|}
{\|w\|}
=
\frac{2}{\|w\|}
}
$$

---

# One-Line Summary

A hyperplane is completely determined by:

1. its normal direction

$$
\hat w=\frac{w}{\|w\|}
$$

2. its signed distance from the origin

$$
-\frac{w_0}{\|w\|}
$$

and the distance between two parallel hyperplanes is

$$
\boxed{
\frac{|w_2-w_1|}
{\|w\|}
}
$$

which leads directly to the SVM margin:

$$
\boxed{
\text{Margin}
=
\frac{2}{\|w\|}
}
$$

# Vectors, Unit Vectors, Hyperplanes and SVM Geometry
---

# 1. What is a Vector?

A vector is an object that has:

1. **Magnitude (Length)**
2. **Direction**

Examples:

In 2D:

$$
\mathbf{w}=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

In 3D:

$$
\mathbf{w}=
\begin{bmatrix}
1\\
2\\
3
\end{bmatrix}
$$

In Machine Learning, vectors usually live in:

$$
\mathbb{R}^d
$$

where

- $d=2$ → 2D space
- $d=3$ → 3D space
- $d=1000$ → 1000-dimensional feature space

---

# 2. Components of a Vector

A vector in $d$ dimensions:

$$
\mathbf{w}
=
\begin{bmatrix}
w_1\\
w_2\\
\vdots\\
w_d
\end{bmatrix}
$$

or

$$
\mathbf{w}
=
[w_1,w_2,\dots,w_d]^T
$$

---

# Example

$$
\mathbf{w}
=
\begin{bmatrix}
1\\
2\\
3
\end{bmatrix}
$$

means:

- Move 1 unit along x-axis
- Move 2 units along y-axis
- Move 3 units along z-axis

---

# 3. Magnitude (Norm) of a Vector

Length of a vector:

$$
\|\mathbf{w}\|
=
\sqrt{
w_1^2+w_2^2+\dots+w_d^2
}
$$

This is called the:

- Euclidean norm
- L2 norm

---

# Example

For

$$
\mathbf{w}
=
\begin{bmatrix}
1\\
2\\
3
\end{bmatrix}
$$

Magnitude:

$$
\|\mathbf{w}\|
=
\sqrt{1^2+2^2+3^2}
=
\sqrt{14}
$$

---

# Geometric Meaning

Magnitude tells:

> How far the vector tip is from the origin.

---

# Example

For

$$
\mathbf{w}
=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

$$
\|\mathbf{w}\|
=
\sqrt{9+16}
=
5
$$

This is exactly Pythagoras theorem.

---

# 4. Unit Vector

A unit vector has:

$$
\|\hat{\mathbf{w}}\|=1
$$

It keeps the direction but removes the magnitude.

---

# Formula

$$
\hat{\mathbf{w}}
=
\frac{\mathbf{w}}
{\|\mathbf{w}\|}
$$

---

# Example

Given

$$
\mathbf{w}
=
\begin{bmatrix}
1\\
2\\
3
\end{bmatrix}
$$

Magnitude:

$$
\|\mathbf{w}\|
=
\sqrt{14}
$$

Therefore,

$$
\hat{\mathbf{w}}
=
\frac{1}{\sqrt{14}}
\begin{bmatrix}
1\\
2\\
3
\end{bmatrix}
=
\begin{bmatrix}
\frac{1}{\sqrt{14}}\\
\frac{2}{\sqrt{14}}\\
\frac{3}{\sqrt{14}}
\end{bmatrix}
$$

---

# Why Unit Vector?

Suppose

Vector A:

$$
[100,200]
$$

Vector B:

$$
[1,2]
$$

Both point in the same direction.

Machine learning often cares about:

- direction
- orientation

rather than magnitude.

Hence we normalize.

---

# Proof that Unit Vector has Length 1

$$
\|\hat{\mathbf{w}}\|
=
\left\|
\frac{\mathbf{w}}
{\|\mathbf{w}\|}
\right\|
$$

$$
=
\sqrt{
\left(
\frac{w_1}{\|\mathbf{w}\|}
\right)^2
+\cdots+
\left(
\frac{w_d}{\|\mathbf{w}\|}
\right)^2
}
$$

$$
=
\sqrt{
\frac{
w_1^2+\cdots+w_d^2
}
{
\|\mathbf{w}\|^2
}
}
$$

$$
=
\sqrt{
\frac{
\|\mathbf{w}\|^2
}{
\|\mathbf{w}\|^2
}
}
=
1
$$

---

# 5. Dot Product

For

$$
\mathbf{a}
=
[a_1,\dots,a_d]^T
$$

and

$$
\mathbf{b}
=
[b_1,\dots,b_d]^T
$$

dot product is:

$$
\mathbf{a}^T\mathbf{b}
=
a_1b_1+a_2b_2+\dots+a_db_d
$$

---

# Geometric Meaning

$$
\mathbf{a}^T\mathbf{b}
=
\|\mathbf{a}\|
\|\mathbf{b}\|
\cos\theta
$$

where

$\theta$

is the angle between vectors.

---

# Important Cases

## Same direction

$$
\theta=0^\circ
$$

$$
\cos\theta=1
$$

Maximum dot product.

---

## Perpendicular

$$
\theta=90^\circ
$$

$$
\cos\theta=0
$$

Therefore

$$
\mathbf{a}^T\mathbf{b}=0
$$

Orthogonal vectors.

---

## Opposite direction

$$
\theta=180^\circ
$$

$$
\cos\theta=-1
$$

Dot product becomes negative.

---

# 6. Orthogonal (Perpendicular) Vectors

Two vectors are orthogonal if:

$$
\mathbf{a}^T\mathbf{b}=0
$$

Example:

$$
\mathbf{a}
=
\begin{bmatrix}
1\\
0
\end{bmatrix}
,
\quad
\mathbf{b}
=
\begin{bmatrix}
0\\
1
\end{bmatrix}
$$

because

$$
1(0)+0(1)=0
$$

---

# 7. What is a Hyperplane?

A hyperplane is a flat surface of dimension:

$$
d-1
$$

inside a

$$
d
$$

dimensional space.

---

# Examples

| Space | Hyperplane |
|-------|------------|
| 2D | Line |
| 3D | Plane |
| 4D | 3D object |
| dD | (d−1)-dimensional object |

---

# Hyperplane Equation

$$
\mathbf{w}^T\mathbf{x}+w_0=0
$$

where

$$
\mathbf{w}
=
[w_1,w_2,\dots,w_d]^T
$$

is called:

- normal vector
- perpendicular vector

and

$$
w_0
$$

is called:

- bias
- intercept
- offset

---

# Example

In 2D:

$$
2x_1+3x_2+4=0
$$

This is a line.

---

# Example

In 3D:

$$
2x_1+3x_2+x_3+5=0
$$

This is a plane.

---

# Why is w called the Normal Vector?

Every vector lying on the hyperplane is perpendicular to

$$
\mathbf{w}
$$

Hence:

$$
\mathbf{w}
\perp
\Pi_d
$$

---

# Geometric Interpretation

The vector

$$
\mathbf{w}
$$

tells:

1. which side is positive
2. which side is negative
3. orientation of hyperplane

---

# Half Spaces

Hyperplane divides space into:

## Positive half-space

$$
\mathbf{w}^T\mathbf{x}+w_0>0
$$

## Negative half-space

$$
\mathbf{w}^T\mathbf{x}+w_0<0
$$

## On hyperplane

$$
\mathbf{w}^T\mathbf{x}+w_0=0
$$

---

# Example

Suppose

$$
x_1+x_2-2=0
$$

Point:

$$
(3,3)
$$

gives

$$
3+3-2=4>0
$$

Positive side.

Point:

$$
(0,0)
$$

gives

$$
0+0-2=-2<0
$$

Negative side.

---

# Distance from Origin to Hyperplane

Hyperplane:

$$
\mathbf{w}^T\mathbf{x}+w_0=0
$$

Distance:

$$
d
=
\frac{|w_0|}
{\|\mathbf{w}\|}
$$

---

# Derivation

Unit normal:

$$
\hat{\mathbf{w}}
=
\frac{\mathbf{w}}
{\|\mathbf{w}\|}
$$

Projection along normal direction gives:

$$
d
=
\frac{|w_0|}
{\|\mathbf{w}\|}
$$

---

# Example

Hyperplane:

$$
2x_1+3x_2+6=0
$$

Then

$$
\mathbf{w}
=
\begin{bmatrix}
2\\
3
\end{bmatrix}
$$

Magnitude:

$$
\|\mathbf{w}\|
=
\sqrt{13}
$$

Distance:

$$
d
=
\frac{6}
{\sqrt{13}}
$$

---

# Distance of Any Point from Hyperplane

For point

$$
\mathbf{x}
$$

distance is

$$
d
=
\frac{
|\mathbf{w}^T\mathbf{x}+w_0|
}{
\|\mathbf{w}\|
}
$$

---

# Signed Distance

Sometimes we need the side information too.

Then:

$$
d_s
=
\frac{
\mathbf{w}^T\mathbf{x}+w_0
}{
\|\mathbf{w}\|
}
$$

Positive:

point lies on positive side.

Negative:

point lies on negative side.

Zero:

point lies on hyperplane.

---

# Distance Between Parallel Hyperplanes

Suppose

$$
\Pi_1:
\mathbf{w}^T\mathbf{x}+w_1=0
$$

and

$$
\Pi_2:
\mathbf{w}^T\mathbf{x}+w_2=0
$$

Same normal vector:

therefore they are parallel.

Distance:

$$
d
=
\frac{
|w_2-w_1|
}{
\|\mathbf{w}\|
}
$$

---

# Example

$$
2x_1+3x_2+5=0
$$

and

$$
2x_1+3x_2+9=0
$$

Distance:

$$
d
=
\frac{|9-5|}
{\sqrt{13}}
=
\frac{4}{\sqrt{13}}
$$

---

# Why This Formula Makes Sense

Changing only the bias

$$
w_0
$$

moves the hyperplane along the normal direction.

Changing

$$
\mathbf{w}
$$

changes the orientation.

---

# Connection to SVM

SVM looks for a hyperplane:

$$
\mathbf{w}^T\mathbf{x}+w_0=0
$$

that separates classes.

---

# Decision Rule

Predict:

$$
y=+1
$$

if

$$
\mathbf{w}^T\mathbf{x}+w_0>0
$$

Predict:

$$
y=-1
$$

if

$$
\mathbf{w}^T\mathbf{x}+w_0<0
$$

---

# Margin Hyperplanes

Positive margin:

$$
\mathbf{w}^T\mathbf{x}+w_0=1
$$

Negative margin:

$$
\mathbf{w}^T\mathbf{x}+w_0=-1
$$

Distance between these two hyperplanes:

$$
\frac{2}
{\|\mathbf{w}\|}
$$

This is called the:

- margin
- geometric margin

---

# Important Insight

Large margin means:

small

$$
\|\mathbf{w}\|
$$

Therefore SVM solves:

$$
\min
\frac12
\|\mathbf{w}\|^2
$$

subject to

$$
y_i
(
\mathbf{w}^T\mathbf{x_i}+w_0
)
\ge 1
$$

---

# Entire Geometry Connection

```
Vector
    ↓
Magnitude
    ↓
Unit Vector
    ↓
Dot Product
    ↓
Orthogonality
    ↓
Normal Vector
    ↓
Hyperplane
    ↓
Distance from Hyperplane
    ↓
Parallel Hyperplanes
    ↓
Margin
    ↓
Support Vector Machine
```

# Interview Questions

### Why divide by ||w|| in SVM?

Because:

$$
\mathbf{w}^T\mathbf{x}+w_0
$$

is not a distance.

Dividing by

$$
\|\mathbf{w}\|
$$

converts it into Euclidean distance.

---

### Why is w perpendicular to the hyperplane?

Because every displacement vector on the hyperplane satisfies:

$$
\mathbf{w}^T\mathbf{v}=0
$$

Therefore:

$$
\mathbf{w}
\perp
\mathbf{v}
$$

for all vectors lying on the hyperplane.

---

### Why does increasing ||w|| decrease margin?

Because:

$$
\text{Margin}
=
\frac{2}{\|\mathbf{w}\|}
$$

Larger denominator means smaller distance.

---
